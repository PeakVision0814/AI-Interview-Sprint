## 1. 理论：DFS vs BFS

请在 Notebook 的 Markdown 单元格中记录：

* **DFS (深度优先搜索)**:
    * **工具**: 栈 (Stack) / 递归 (Recursion)。
    * **行为**: 不撞南墙不回头。一条路走到黑，到底了再回退。
    * **场景**: 我们前几天做的所有题 (最大深度、翻转、LCA、路径总和) 都是 DFS。


* **BFS (广度优先搜索)**:
    * **工具**: **队列 (Queue)**。
    * **行为**: 像水波纹一样扩散。先看离我距离为 1 的所有点，再看距离为 2 的所有点...
    * **场景**: 层序遍历、最短路径 (Shortest Path)。
    * **口诀**: **"先进先出" (FIFO)**。

## 2. 核心算法：LC 102. 二叉树的层序遍历

**题目**: [102. Binary Tree Level Order Traversal](https://leetcode.cn/problems/binary-tree-level-order-traversal/)
**描述**: 给你一个二叉树，请你返回其按 **层序遍历** 得到的节点值。 （即逐层地，从左到右访问所有节点）。

**示例**:

```text
    3
   / \
  9  20
    /  \
   15   7

```

**输出**: `[[3], [9,20], [15,7]]`

### 2.1 代码模板 (标准 BFS)

这是 BFS 的**万能模板**，请务必背下来。

````python
import collections

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right

class Solution:
    def levelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
        if not root:
            return []
        
        # 1. 初始化队列 (Queue)
        # 使用 deque (双端队列) 为了高效 popleft()
        queue = collections.deque([root])
        result = []
        
        # 2. 循环直到队列为空
        while queue:
            level_nodes = [] # 存放这一层的节点值
            
            # 3. 关键点：锁定当前层的长度！
            # 这一步决定了我们怎么把不同层的节点分开
            current_level_size = len(queue)
            
            # 4. 遍历当前层的每一个节点
            for _ in range(current_level_size):
                # 出队 (FIFO)
                node = queue.popleft()
                level_nodes.append(node.val)
                
                # 入队 (把下一层的孩子加到队尾)
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
            
            # 把这一层的结果打包放进大结果里
            result.append(level_nodes)
            
        return result
````

### 2.2 深度解析 (为什么要那个 for 循环？)

很多初学者写 BFS 容易漏掉 `for _ in range(len(queue))`，结果输出变成了一维数组 `[3, 9, 20, 15, 7]`，而不是二维的 `[[3], [9, 20]...]`。

* **不加 for 循环**: 队列是一个持续流动的管道，你分不清哪个节点是第几层的，只知道它们出来的顺序。
* **加了 for 循环**: 相当于在每一层开始前，**拍了一张快照**。“现在的队列里有 2 个元素（9和20），这俩肯定是同一层的。好，我只处理这 2 个，新加进来的孩子（15和7）是下一层的，暂时别动。”

## 3. 图解演示 (Trace)

以 `[3, 9, 20, 15, 7]` 为例：

1. **Init**: `Q = [3]`, `res = []`
2. **Round 1**:
* `size = 1`。
* Pop `3`。Add `9, 20` to Q。
* `level = [3]`。 `res = [[3]]`。
* `Q` 现在是 `[9, 20]`。


3. **Round 2**:
* `size = 2` (锁定！这一轮只处理 2 个)。
* Pop `9`。Add nothing (没孩子)。
* Pop `20`。Add `15, 7` to Q。
* `level = [9, 20]`。 `res = [[3], [9, 20]]`。
* `Q` 现在是 `[15, 7]`。


4. **Round 3**:
* `size = 2`。
* Pop `15`... Pop `7`...
* `level = [15, 7]`。 `res` 更新。


5. **End**.
