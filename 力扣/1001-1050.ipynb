## [1047. 删除字符串中的所有相邻重复项](https://leetcode.cn/problems/remove-all-adjacent-duplicates-in-string/)

简单  相关标签  ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业



提示



给出由小写字母组成的字符串 `s`，**重复项删除操作**会选择两个相邻且相同的字母，并删除它们。

在 `s` 上反复执行重复项删除操作，直到无法继续删除。

在完成所有重复项删除操作后返回最终的字符串。答案保证唯一。

 

**示例：**

```
输入："abbaca"
输出："ca"
解释：
例如，在 "abbaca" 中，我们可以删除 "bb" 由于两字母相邻且相同，这是此时唯一可以执行删除操作的重复项。之后我们得到字符串 "aaca"，其中又只有 "aa" 可以执行重复项删除操作，所以最后的字符串为 "ca"。
```

````python
class Solution:
    def removeDuplicates(self, s: str) -> str:
        result = []
        for char in s:
            if result and result[-1] == char:
                result.pop()
            else:
                result.append(char)
        
        return "".join(result)

solution = Solution()
input_s = input("请输入字符串s：")
output_result = solution.removeDuplicates(input_s)
print(f"删除相邻重复项后的字符串为：{output_result}")
````

**算法思路:**

这道题的核心是重复地删除相邻的相同字符，直到没有相邻重复项为止。一种直观的想法是不断地遍历字符串，找到相邻重复项并删除，然后重新开始遍历。但是，在字符串中直接进行插入和删除操作效率较低，并且在删除后需要小心处理索引的变化。

我们可以换一种思路：利用一个**栈（或者说一个列表模拟栈的行为）**来处理。栈是一种后进先出（LIFO）的数据结构。在这个问题中，我们可以利用栈的这个特性来方便地检查和删除相邻的重复项。

具体来说，我们遍历输入的字符串 `s` 的每个字符。对于每一个字符，我们将其与**当前栈顶的元素**进行比较：

1. **如果栈是空的**，或者**当前字符与栈顶元素不同**，说明当前字符不会立即与前面的字符形成相邻重复项，我们将当前字符压入栈中。
2. **如果栈不为空**，并且**当前字符与栈顶元素相同**，说明当前字符与栈顶元素是相邻重复的。在这种情况下，我们不需要将当前字符压入栈，而是将栈顶元素弹出，相当于删除了这对相邻的重复项。

重复这个过程，直到遍历完输入字符串中的所有字符。最后，栈中剩下的元素就是经过所有删除操作后最终的字符串。

**Python 代码实现及语法解释:**

```python
class Solution:
    def removeDuplicates(self, s: str) -> str:
        # 初始化一个空列表，用来模拟栈的行为
        # 我们将处理后的字符暂时存放在这个列表中
        result = []

        # 遍历输入的字符串 s 中的每一个字符
        for char in s:
            # 检查条件：
            # 1. 结果列表 result 是否不为空 (result 非空为 True)
            # 2. 结果列表 result 的最后一个元素 (result[-1]) 是否与当前字符 char 相同
            # 只有当这两个条件都满足时，才执行 if 块的代码
            if result and result[-1] == char:
                # 如果找到相邻重复项，将结果列表的最后一个元素弹出
                # result.pop() 会删除并返回列表末尾的元素
                # 在这里我们只关心删除操作
                result.pop()
            else:
                # 如果结果列表为空，或者当前字符与结果列表的最后一个元素不同
                # 说明当前字符不会与前面的字符形成相邻重复项
                # 将当前字符添加到结果列表的末尾 (模拟入栈操作)
                # result.append(char) 会在列表末尾添加元素
                result.append(char)

        # 当循环结束后，结果列表 result 中存放的就是最终的字符序列
        # "".join(result) 将列表中的所有字符连接起来，形成一个字符串
        # "" 表示连接时使用的分隔符是空字符串，即直接连接字符
        return "".join(result)

# 创建 Solution 类的实例
solution = Solution()

# 获取用户输入的字符串
input_s = input("请输入字符串s：")

# 调用 removeDuplicates 方法处理输入的字符串
output_result = solution.removeDuplicates(input_s)

# 打印处理后的结果
print(f"删除相邻重复项后的字符串为：{output_result}")
```

**Python 语法解释:**

* **`class Solution:`**: 定义一个名为 `Solution` 的类。在面向对象编程中，类是创建对象的蓝图。这里我们将算法的逻辑封装在一个类中。
* **`def removeDuplicates(self, s: str) -> str:`**: 定义一个名为 `removeDuplicates` 的方法，它是 `Solution` 类的一部分。
    * `self`: 表示方法所属的对象本身。在调用类的方法时会自动传递。
    * `s: str`: 定义一个参数 `s`，并使用类型提示 `str` 表明它应该是一个字符串。
    * `-> str`: 使用类型提示表明该方法将返回一个字符串。
* **`result = []`**: 创建一个空的列表，并将其赋值给变量 `result`。在 Python 中，列表使用方括号 `[]` 定义。列表是可变的，可以动态地添加、删除和修改元素。这里我们将列表作为栈来使用。
* **`for char in s:`**: 这是一个 `for` 循环，用于遍历字符串 `s` 中的每一个字符。在每一次迭代中，当前字符会被赋值给变量 `char`。
* **`if result and result[-1] == char:`**: 这是一个条件判断语句。
    * `result`: 在布尔上下文中，非空的列表被视为 `True`，空列表被视为 `False`。所以 `if result` 检查 `result` 列表是否不为空。
    * `result[-1]`: 这是列表的索引操作。`-1` 表示列表中的最后一个元素。
    * `result[-1] == char`: 比较列表的最后一个元素是否等于当前字符 `char`。
    * `and`: 逻辑运算符，只有当两个条件都为真时，整个 `if` 条件才为真。
* **`result.pop()`**: 这是一个列表的方法。`pop()` 方法删除并返回列表的最后一个元素。在这里，我们利用它来模拟栈的“出栈”操作，移除相邻的重复项。
* **`else:`**: 如果 `if` 条件不满足，则执行 `else` 块中的代码。
* **`result.append(char)`**: 这是一个列表的方法。`append(item)` 方法在列表的末尾添加一个元素 `item`。在这里，我们利用它来模拟栈的“入栈”操作。
* **`return "".join(result)`**: 返回处理后的字符串。
    * `"".join(list)`: 这是一个字符串方法。它将列表 `list` 中的所有元素连接起来，形成一个字符串。连接时使用 `""` 作为分隔符，即不加任何分隔符直接连接。
* **`solution = Solution()`**: 创建 `Solution` 类的一个实例（对象），并将其赋值给变量 `solution`。
* **`input("请输入字符串s：")`**: 这是一个内置函数，用于从用户那里获取输入。括号内的字符串是给用户的提示信息。用户输入的内容会被作为字符串返回。
* **`print(f"...")`**: 这是一个内置函数，用于将内容输出到控制台。`f"..."` 是一个 f-string（格式化字符串字面量），可以在字符串中方便地嵌入变量的值，使用大括号 `{}` 将变量名括起来即可。

**总结:**

这段代码采用了一种高效的基于栈（或者说列表模拟栈）的算法来解决删除相邻重复项的问题。通过遍历输入字符串，并根据当前字符与栈顶元素的关系进行入栈或出栈操作，最终栈中剩下的元素就是符合题目要求的字符串。代码使用了 Python 中常用的列表操作（`append`、`pop`、索引 `[-1]`）和字符串操作 (`join`)，以及基本的控制流语句 (`for` 循环、`if/else` 条件判断)。理解列表作为栈的使用是解决这类问题的关键。
