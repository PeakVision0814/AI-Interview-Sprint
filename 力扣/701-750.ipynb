## [704. 二分查找](https://leetcode.cn/problems/binary-search/)

简单  相关标签  ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业



给定一个 `n` 个元素有序的（升序）整型数组 `nums` 和一个目标值 `target` ，写一个函数搜索 `nums` 中的 `target`，如果 `target` 存在返回下标，否则返回 `-1`。

你必须编写一个具有 `O(log n)` 时间复杂度的算法。


**示例 1:**

```
输入: nums = [-1,0,3,5,9,12], target = 9
输出: 4
解释: 9 出现在 nums 中并且下标为 4
```

**示例 2:**

```
输入: nums = [-1,0,3,5,9,12], target = 2
输出: -1
解释: 2 不存在 nums 中因此返回 -1
```

````python
class Solution:
    def search(self, nums: List[int], target: int) -> int:
        left = 0
        right = len(nums) - 1

        while left <= right:
            mid = (left + right) // 2

            if nums[mid] == target:
                return mid

            elif nums[mid] < target:
                left = mid + 1
            
            elif nums[mid] > target:
                right = mid - 1

        return -1
````

**这就是我们要学的这个算法的核心思想：二分查找（Binary Search）**。

它就像是你在翻电话簿时，不是从头开始翻，而是先翻到中间。

1.  **看中间这个人是谁：** 如果中间这个人就是你要找的人，那太好了，你一下子就找到了！
2.  **如果中间这个人不是你要找的人：**
    *   如果中间这个人的名字在你要找的人**前面**（比如你想找 "张三"，但翻到中间是 "李四"），那说明你要找的人肯定在电话簿的**后半部分**。
    *   如果中间这个人的名字在你要找的人**后面**（比如你想找 "张三"，但翻到中间是 "王五"），那说明你要找的人肯定在电话簿的**前半部分**。
3.  **缩小范围：** 无论在哪一半，你都把注意力集中到那一半，然后**重复上面的步骤**，再次找到这一半的中间，继续比较。
4.  **直到找到或者找不到：** 你不断地缩小查找的范围，直到找到你要找的人，或者范围已经小到没有可以再找的地方了，那就说明你要找的人不在电话簿里。

这个过程是不是很像在玩“猜数字”游戏，你每次都通过猜测来排除掉一半不可能的数字？

**好，现在我们来看代码，它就是用编程语言把上面的“二分查找”过程实现出来。**

这个代码叫做 `search`，它的工作就是在一个叫做 `nums` 的列表（你可以暂时理解为一串有序的数字电话簿）里，查找一个叫做 `target` 的数字（你要找的那个人的电话号码）。如果找到了，它会告诉你那个数字在列表里的位置（索引），如果没找到，就告诉你找不到（返回 -1）。

```python
class Solution:
    def search(self, nums: List[int], target: int) -> int:
        # 这部分代码就像是创建了一个工具，叫做 Solution，
        # 里面的 search 函数就是我们的查找工具
        # 你可以暂时不用管 class Solution 和 def search 这两行，
        # 重点看 search 函数里面是怎么工作的

        # 1. 准备工作：设定查找的范围
        left = 0
        right = len(nums) - 1

        # Imagine you have a list of numbers: [-1, 0, 3, 5, 9, 12]
        # nums is [-1, 0, 3, 5, 9, 12]
        # len(nums) is 6 (because there are 6 numbers)
        # left is 0 (the index of the first number, -1)
        # right is 6 - 1 = 5 (the index of the last number, 12)
        # So, initially, our search range is from index 0 to index 5.

        # 'left' 代表当前查找范围的左边最开始的索引
        # 'right' 代表当前查找范围的右边最结束的索引
        # 刚开始，查找范围就是整个列表，所以 left 是第一个元素的索引 0，right 是最后一个元素的索引 (列表长度 - 1)

        # 2. 开始循环查找：只要查找范围还有东西，就继续找
        while left <= right:
            # 这个 while 循环的意思是：只要 left 的索引小于或等于 right 的索引，就一直执行里面的代码。
            # 如果 left 大于 right 了，说明我们的查找范围已经“空”了，找不到目标数字，循环就结束了。

            # 3. 找到中间位置
            mid = (left + right) // 2
            # 'mid' 代表当前查找范围的中间那个元素的索引
            # (left + right) // 2 就是计算 left 和 right 的中间值，// 是整数除法，保证结果是整数索引
            # Example: If left is 0 and right is 5, mid = (0 + 5) // 2 = 5 // 2 = 2.
            # So the middle index is 2. In the list [-1, 0, 3, 5, 9, 12], the element at index 2 is 3.

            # 4. 比较中间元素和目标值
            if nums[mid] == target:
                # 你这里之前的代码写成了 nums(mid)，这是一个常见的初学者错误！
                # 在 Python 里，当你想要获取列表（nums）里某个位置（mid 索引）的元素时，要用方括号 []
                # nums[mid] 的意思是“去 nums 这个列表里，找到第 mid 个位置的元素”。
                # 如果这个中间元素 nums[mid] 正好等于我们要找的 target 数字，那说明我们找到了！
                return mid
                # 找到目标了，就返回它所在的索引 mid。函数运行到这里就结束了。

            elif nums[mid] < target:
                # 如果中间的元素 nums[mid] 比我们要找的 target 小
                # Example: target is 9, nums[mid] is 3. Since 3 < 9, target must be in the right half.
                # 这说明我们要找的 target 肯定在中间元素的右边部分（因为列表是排序好的！）
                left = mid + 1
                # 所以，我们就把查找范围的左边界 (left) 移动到 mid 的右边一个位置 (mid + 1)。
                # 下一次循环，我们就在右边那个缩小了的范围里继续找。

            elif nums[mid] > target:
                # 如果中间的元素 nums[mid] 比我们要找的 target 大
                # Example: target is 3, nums[mid] is 5. Since 5 > 3, target must be in the left half.
                # 这说明我们要找的 target 肯定在中间元素的左边部分。
                right = mid - 1
                # 所以，我们就把查找范围的右边界 (right) 移动到 mid 的左边一个位置 (mid - 1)。
                # 下一次循环，我们就在左边那个缩小了的范围里继续找。

        # 5. 如果循环结束还没找到
        return -1
        # 如果 while 循环结束了（说明 left 已经大于 right，查找范围空了），还没通过 return mid 返回，
        # 那就说明我们要找的 target 数字不在这个列表里。
        # 约定俗成地，我们返回 -1 来表示没有找到。
```
**总结一下，这个代码就是用二分查找的思路来找一个数字：**

1.  设定最开始的查找范围（整个列表）。
2.  在一个循环里，不断计算当前范围的中间位置。
3.  比较中间位置的元素和目标数字。
4.  如果找到了，返回位置。
5.  如果没找到，根据比较结果，把查找范围缩小到左边或右边一半。
6.  重复步骤 2-5，直到找到或者范围为空。
7.  如果范围为空还没找到，就返回 -1。

**Python 语法小提示：**

*   `nums: List[int]`：这叫做类型提示，它告诉我们 `nums` 应该是一个由整数组成的列表。对于初学者可以暂时忽略。
*   `-> int`：这也叫类型提示，它告诉我们这个函数最后会返回一个整数。对于初学者可以暂时忽略。
*   `len(nums)`：这是一个内置函数，用来获取列表 `nums` 的长度（有多少个元素）。
*   `nums[mid]`：这是访问列表元素的方法，`nums` 是列表名字，`mid` 是你要访问的元素的索引（位置）。
*   `while left <= right:`：这是一个循环结构，只要 `left <= right` 这个条件为真，就会一直执行循环里面的代码块。
*   `if ... elif ... else ...`：这是一个条件判断结构。先判断 `if` 后面的条件，如果为真就执行 `if` 下面的代码。如果 `if` 条件为假，就判断 `elif` 后面的条件，为真则执行 `elif` 下面的代码。你这里的代码没有 `else` 部分，因为 `elif nums[mid] > target:` 已经覆盖了所有 `nums[mid] < target` 之外的情况（除了等于的情况，已经由第一个 `if` 处理了）。
*   `return`：当函数执行到 `return` 语句时，它会立即停止执行，并将后面的值作为函数的结果返回。

## [707. 设计链表](https://leetcode.cn/problems/design-linked-list/)

中等  相关标签  ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业



你可以选择使用单链表或者双链表，设计并实现自己的链表。

单链表中的节点应该具备两个属性：`val` 和 `next` 。`val` 是当前节点的值，`next` 是指向下一个节点的指针/引用。

如果是双向链表，则还需要属性 `prev` 以指示链表中的上一个节点。假设链表中的所有节点下标从 **0** 开始。

实现 `MyLinkedList` 类：

- `MyLinkedList()` 初始化 `MyLinkedList` 对象。
- `int get(int index)` 获取链表中下标为 `index` 的节点的值。如果下标无效，则返回 `-1` 。
- `void addAtHead(int val)` 将一个值为 `val` 的节点插入到链表中第一个元素之前。在插入完成后，新节点会成为链表的第一个节点。
- `void addAtTail(int val)` 将一个值为 `val` 的节点追加到链表中作为链表的最后一个元素。
- `void addAtIndex(int index, int val)` 将一个值为 `val` 的节点插入到链表中下标为 `index` 的节点之前。如果 `index` 等于链表的长度，那么该节点会被追加到链表的末尾。如果 `index` 比长度更大，该节点将 **不会插入** 到链表中。
- `void deleteAtIndex(int index)` 如果下标有效，则删除链表中下标为 `index` 的节点。

 

**示例：**

```
输入
["MyLinkedList", "addAtHead", "addAtTail", "addAtIndex", "get", "deleteAtIndex", "get"]
[[], [1], [3], [1, 2], [1], [1], [1]]
输出
[null, null, null, null, 2, null, 3]

解释
MyLinkedList myLinkedList = new MyLinkedList();
myLinkedList.addAtHead(1);
myLinkedList.addAtTail(3);
myLinkedList.addAtIndex(1, 2);    // 链表变为 1->2->3
myLinkedList.get(1);              // 返回 2
myLinkedList.deleteAtIndex(1);    // 现在，链表变为 1->3
myLinkedList.get(1);              // 返回 3
```


````python
class ListNode:
    def __init__(self, val = 0, next = None):
        self.val = val
        self.next = next

class MyLinkedList:

    def __init__(self):
        self.dummy_head = ListNode()
        self.size = 0

    def get(self, index: int) -> int:
        if index < 0 or index >= self.size:
            return -1
        current = self.dummy_head.next
        for i in range(index):
            current = current.next
        
        return current.val

    def addAtHead(self, val: int) -> None:
        new_node = ListNode(val)
        new_node.next = self.dummy_head.next
        self.dummy_head.next = new_node
        self.size += 1        

    def addAtTail(self, val: int) -> None:
        new_node = ListNode(val)
        current = self.dummy_head 

        while current.next:
            current = current.next

        current.next = new_node

        self.size += 1

    def addAtIndex(self, index: int, val: int) -> None:
        if index >self.size:
            return 0
        if index < 0:
            index = 0

        new_node = ListNode(val)
        current = self.dummy_head

        for _ in range(index):   
            current = current.next
        new_node.next = current.next
        current.next = new_node

        self.size +=1
        

    def deleteAtIndex(self, index: int) -> None:
        if index < 0 or index >= self.size:
            return

        current = self.dummy_head
        for _ in range(index):
            current = current.next
        current.next = current.next.next

        self.size -= 1

# Your MyLinkedList object will be instantiated and called as such:
# obj = MyLinkedList()
# param_1 = obj.get(index)
# obj.addAtHead(val)
# obj.addAtTail(val)
# obj.addAtIndex(index,val)
# obj.deleteAtIndex(index)
````

**链表是什么？**

想象一下你有很多张写着信息的纸条（节点），每张纸条除了信息，还告诉你下一张纸条在哪里。把这些纸条串起来，就像一个链条，这就是一个链表。

*   **节点 (ListNode):** 每张纸条就是一个节点。它包含两个东西：
    *   `val`: 纸条上的信息（也就是你想要存的数据，比如一个数字）。
    *   `next`: 指向下一张纸条的箭头或者地址。如果这是最后一张纸条，它就没有箭头，或者箭头指向“无”（None）。

*   **链表 (MyLinkedList):** 就是管理这些纸条的整体。它需要知道第一张纸条在哪里，有多少张纸条等等。

**你的代码比喻：**

你的代码使用了**虚拟头节点 (dummy_head)**。这就像你在第一张纸条前面放了一个特殊的、不放信息的“引子纸条”。这个引子纸条不包含实际数据，它的唯一作用就是指向真正的第一张纸条。这样做的好处是，无论链表是空的还是有数据的，对第一个节点的处理方式都会更统一、更简单。

*   `dummy_head`: 就是那个特殊的“引子纸条”，它总是在最前面。它自己没有 `val`（或者 `val` 是默认值 0），它的 `next` 指向链表的第一个**实际**节点。
*   `size`: 记录链表里有多少张**实际**的纸条。

**现在，我们来一句一句解释你的代码：**

```python
class ListNode:

        self.val = val
        self.next = next
```
*   `class ListNode:`: 这就像在说，“我们要定义一个叫做‘节点’的东西的‘蓝图’”。
*   `def __init__(self, val = 0, next = None):`: 这是定义节点在被“创建”时要做的事情。`__init__` 是一个特殊的函数，当你说 `ListNode()` 的时候，它就会被调用。
    *   `self`: 代表正在被创建的这个节点自己。
    *   `val = 0`: 创建节点时，可以给它一个值（信息）。如果你不给，默认就是 0。`self.val = val` 把你给的值存到这个节点的 `val` 部分。
    *   `next = None`: 创建节点时，可以指定它的下一个节点。如果你不指定，默认就是 `None`（没有下一个）。`self.next = next` 把你指定的下一个节点（或 None）存到这个节点的 `next` 部分。

```python
class MyLinkedList:

    def __init__(self):
        self.dummy_head = ListNode()
        self.size = 0
```
*   `class MyLinkedList:`: 这是定义一个叫做“我的链表”的东西的“蓝图”。
*   `def __init__(self):`: 这是定义一个链表在被“创建”时要做的事情。
    *   `self.dummy_head = ListNode()`: 创建一个上面说的“引子纸条”，并把它存到链表的 `dummy_head` 这个地方。这个引子纸条默认 `val` 是 0，`next` 是 `None`（因为它刚创建时，链表是空的，它指向的第一个实际节点是 None）。
    *   `self.size = 0`: 刚创建时，链表里没有任何实际的纸条，所以实际纸条的数量是 0。

```python
    def get(self, index: int) -> int:
        """
        根据索引获取节点的值
        index: 你想获取哪个位置的纸条 (从 0 开始计数)
        返回: 纸条上的值，如果索引无效则返回 -1
        """
        if index < 0 or index >= self.size:
            return -1 # 如果索引小于 0 或者大于等于实际纸条的数量，说明这个位置不存在，返回 -1

        # 从虚拟头节点的下一个节点开始遍历，因为虚拟头节点没有实际数据
        current = self.dummy_head.next
        # 从第一个实际节点 (索引 0) 开始，我们要走到索引为 index 的那个节点
        for i in range(index):
            # 向后移动一步，走到下一个纸条
            current = current.next

        # 循环结束后，current 正好指向了索引为 index 的那个纸条
        return current.val # 返回这个纸条上的值

```
*   `def get(self, index: int) -> int:`: 定义一个叫做 `get` 的功能，它可以根据你给的 `index`（位置），找到对应纸条上的 `val`。
*   `if index < 0 or index >= self.size:`: 检查你给的 `index` 有没有意义。链表的索引从 0 开始，到 `size - 1` 结束。如果你要找的位置小于 0 或者大于等于实际纸条总数，那就找不到，所以返回 -1。
*   `current = self.dummy_head.next`: 创建一个临时的“现在看的是哪张纸条”的指针 `current`。我们从虚拟头节点的下一个开始看，因为那是第一个**实际**纸条（索引为 0）。
*   `for i in range(index):`: 这是一个循环，会执行 `index` 次。
    *   `current = current.next`: 在循环里，每次都把 `current` 指针向后移动一步，指向下一张纸条。
*   `return current.val`: 循环结束时，`current` 就指向了我们要找的那个纸条，所以返回它的 `val` 值。

```python
    def addAtHead(self, val: int) -> None:
        """
        在链表的最前面 (第一个实际节点之前) 插入一个新节点
        val: 新节点的值
        """
        new_node = ListNode(val) # 创建一张新的纸条，上面写着 val 的信息

        # 把新纸条插入到虚拟头节点和它原来指向的第一个实际节点之间
        new_node.next = self.dummy_head.next # 新纸条的箭头指向原来虚拟头节点指向的那张纸条
        self.dummy_head.next = new_node     # 虚拟头节点的箭头现在指向新纸条

        self.size += 1 # 实际纸条的数量增加 1

```
*   `def addAtHead(self, val: int) -> None:`: 定义一个叫做 `addAtHead` 的功能，用来在链表的最前面加一张新纸条。`-> None` 表示这个函数不返回任何东西。
*   `new_node = ListNode(val)`: 创建一张带有 `val` 信息的新纸条。
*   `new_node.next = self.dummy_head.next`: 这是关键一步。新纸条的 `next` 指针（箭头）被设置为指向原来虚拟头节点指向的那张纸条。想象一下，新纸条准备要排在第一个位置，它自然要指向原来的第一个纸条。
*   `self.dummy_head.next = new_node`: 虚拟头节点的 `next` 指针（箭头）现在指向我们刚刚创建的新纸条。这样，虚拟头节点就直接连到了新纸条，新纸条就成为了链表的第一个实际节点。
*   `self.size += 1`: 链表里多了一张实际纸条，所以总数加一。

```python
    def addAtTail(self, val: int) -> None:
        """
        在链表的最后面插入一个新节点
        val: 新节点的值
        """
        new_node = ListNode(val) # 创建一张新的纸条，上面写着 val 的信息

        # 需要找到链表的最后一张纸条
        current = self.dummy_head # 从虚拟头节点开始找

        # 循环，只要 current 的下一个节点不是 None (也就是 current 还不是最后一个纸条)，就继续往后走
        while current.next:
            current = current.next

        # 循环结束后，current 就指向了链表的最后一张纸条 (它的 next 是 None)

        # 把新纸条连到最后一张纸条后面
        current.next = new_node # 最后一张纸条的箭头现在指向新纸条 (新纸条的 next 默认是 None，它成了新的最后一张纸条)

        self.size += 1 # 实际纸条的数量增加 1

```
*   `def addAtTail(self, val: int) -> None:`: 定义一个叫做 `addAtTail` 的功能，用来在链表的最末尾加一张新纸条。
*   `new_node = ListNode(val)`: 创建一张带有 `val` 信息的新纸条。
*   `current = self.dummy_head`: 创建一个临时指针 `current`，从虚拟头节点开始。我们最终需要找到最后一张纸条，然后修改它的 `next`。从 `dummy_head` 开始遍历比较方便。
*   `while current.next:`: 这是一个循环。只要 `current` 指向的纸条后面还有纸条（`current.next` 不是 `None`），就继续执行循环。
    *   `current = current.next`: 把 `current` 指针向后移动一步。
*   循环结束后，`current` 就指向了链表的最后一个**实际**纸条（或者如果链表是空的，它就是 `self.dummy_head`）。
*   `current.next = new_node`: 把 `current` 指向的那个纸条（也就是最后一张纸条）的 `next` 指针（箭头）指向新创建的 `new_node`。这样，新纸条就接在了最后面。
*   `self.size += 1`: 链表里多了一张实际纸条，所以总数加一。

```python
    def addAtIndex(self, index: int, val: int) -> None:
        """
        在链表中索引为 index 的位置之前插入一个新节点
        val: 新节点的值
        index: 插入的位置 (0 表示头部，self.size 表示尾部)
        """
        # 检查索引是否有效
        # 如果索引大于实际纸条总数，无法插入
        if index >self.size:
            return # 不做任何事情

        # 如果索引小于 0，视为在头部插入 (和 addAtHead 一样)
        if index < 0:
            index = 0

        new_node = ListNode(val) # 创建新纸条

        # 找到需要在哪个位置的前面插入。
        # 也就是说，我们要找到索引为 index - 1 的那个纸条。
        # 从虚拟头节点 (相当于索引 -1) 开始
        current = self.dummy_head

        # 循环 index 次，让 current 走到索引 index - 1 的位置
        # 比如 index 是 0，循环 0 次，current 还在 dummy_head (索引 -1)
        # 比如 index 是 1，循环 1 次，current 走到 dummy_head.next (索引 0)
        # 比如 index 是 size，循环 size 次，current 走到最后一个实际节点 (索引 size - 1)
        for _ in range(index):
            current = current.next

        # 现在 current 指向了索引为 index - 1 的纸条

        # 把新纸条插入到 current 和 current.next 之间
        new_node.next = current.next # 新纸条的箭头指向 current 原来的下一个纸条
        current.next = new_node     # current 的箭头现在指向新纸条

        self.size +=1 # 实际纸条的数量增加 1

```
*   `def addAtIndex(self, index: int, val: int) -> None:`: 定义一个叫做 `addAtIndex` 的功能，用来在指定 `index` 位置插入新纸条。
*   `if index > self.size:`: 如果 `index` 大于链表总数，说明插入位置无效，直接返回。
*   `if index < 0:`: 如果 `index` 小于 0，我们约定它就是在头部插入（索引 0），所以把 `index` 改成 0。
*   `new_node = ListNode(val)`: 创建新纸条。
*   `current = self.dummy_head`: 临时指针从虚拟头节点开始。
*   `for _ in range(index): current = current.next`: 循环 `index` 次，把 `current` 移动到索引为 `index - 1` 的那个纸条。
*   `new_node.next = current.next`: 新纸条的 `next` 指向 `current` 原来指向的下一个纸条。
*   `current.next = new_node`: `current` 的 `next` 现在指向新纸条。这样，新纸条就插在了 `current` 和它原来的下一个纸条之间。
*   `self.size +=1`: 链表总数加一。

```python
    def deleteAtIndex(self, index: int) -> None:
        """
        删除链表中索引为 index 的节点
        index: 要删除哪个位置的纸条 (从 0 开始计数)
        """
        # 检查索引是否有效
        # 如果索引小于 0 或者大于等于实际纸条总数，说明这个位置不存在，不能删除
        if index < 0 or index >= self.size:
            return # 索引无效，不做任何事情

        # 需要找到要删除节点的前一个节点
        # 从虚拟头节点开始
        current = self.dummy_head

        # 循环 index 次，让 current 走到索引 index - 1 的位置
        # 比如 index 是 0，循环 0 次，current 还在 dummy_head (索引 -1)，这是要删除第一个实际节点的前一个
        # 比如 index 是 1，循环 1 次，current 走到 dummy_head.next (索引 0)，这是要删除索引 1 节点的前一个
        for _ in range(index):
            current = current.next

        # 现在 current 指向了要删除节点的前一个节点

        # 执行删除操作：跳过要删除的节点
        # current.next 是要删除的节点
        # current.next.next 是要删除节点的下一个节点
        current.next = current.next.next # 把 current 的 next 指针直接指向要删除节点的下一个节点

        self.size -= 1 # 实际纸条的数量减少 1

```
*   `def deleteAtIndex(self, index: int) -> None:`: 定义一个叫做 `deleteAtIndex` 的功能，用来删除指定 `index` 位置的纸条。
*   `if index < 0 or index >= self.size:`: 检查索引是否有效。有效范围是 `[0, self.size - 1]`。无效就返回。
*   `current = self.dummy_head`: 临时指针从虚拟头节点开始。
*   `for _ in range(index): current = current.next`: 循环 `index` 次，把 `current` 移动到索引为 `index - 1` 的那个纸条（也就是要删除节点的前一个）。
*   `current.next = current.next.next`: 这是删除的关键。`current.next` 是要删除的节点。通过把 `current` 的 `next` 指针指向 `current.next.next`（要删除节点的下一个节点），就像跳过了要删除的那个节点，把它从链条里移除了。
*   `self.size -= 1`: 链表总数减一。
