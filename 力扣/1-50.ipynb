## 1. 两数之和

给定一个整数数组 `nums` 和一个整数目标值 `target`，请你在该数组中找出 **和为目标值** *`target`* 的那 **两个** 整数，并返回它们的数组下标。

你可以假设每种输入只会对应一个答案，并且你不能使用两次相同的元素。

你可以按任意顺序返回答案。

 

**示例 1：**

```
输入：nums = [2,7,11,15], target = 9
输出：[0,1]
解释：因为 nums[0] + nums[1] == 9 ，返回 [0, 1] 。
```

**示例 2：**

```
输入：nums = [3,2,4], target = 6
输出：[1,2]
```

**示例 3：**

```
输入：nums = [3,3], target = 6
输出：[0,1]
```

 

**提示：**

- `2 <= nums.length <= 104`
- `-109 <= nums[i] <= 109`
- `-109 <= target <= 109`
- **只会存在一个有效答案**

````python
from typing import List

class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        hashmap = {}
        for i, num in enumerate(nums):
            complement = target - num
            if complement in hashmap:
                return [hashmap[complement], i]
            hashmap[num] = i
        return []
  
solution = Solution()

# 手动输入数组
nums_input = input("请输入数组（用逗号分隔）: ")
nums = list(map(int, nums_input.split(',')))  # 将输入的字符串转成整数列表
target = int(input("请输入目标值: "))

print("结果:", solution.twoSum(nums, target))
````


🍎 **整体思路**
我们需要找到数组中的两个数，它们加起来等于目标值。**关键技巧：用字典记住每个数字的「值」和「位置」**，例如 `{2:0, 7:1}` 表示数字2在位置0，数字7在位置1。这样当我们检查一个新数字时，可以直接查字典看是否有配对（比如看到7时，找有没有9-7=2）。

📝 **代码逐行解释**
```python
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        hashmap = {}  # 创建一个空字典，用来存【数字的值】->【该数字的索引】
        for i, num in enumerate(nums):  # 遍历数组，i是索引，num是当前数字的值
            complement = target - num    # 计算缺口的数（比如target=9，当前数是2，缺口是7）
            if complement in hashmap:    # 查字典：缺口的数是否已经存过了？
                return [hashmap[complement], i]  # 如果有，直接返回两个索引
            hashmap[num] = i   # 如果没找到，把当前数和索引存入字典
        return []  # 理论上题目保证有解，所以这行不会执行
```

🔍 **以示例1（nums = [2,7,11,15], target = 9）模拟代码运行**
| 当前循环 | i的值 | num的值 | complement=9-num | 检查字典 | 执行操作 |
|---------|------|--------|------------------|---------|---------|
| 第一次   | 0    | 2      | 9-2=7            | 字典为空 | 存入 `{2:0}` |
| 第二次   | 1    | 7      | 9-7=2            | 字典中存在2（对应索引0） | 返回 `[0, 1]` |

👉 **最终结果**：返回`[0,1]`，正确 ✅。

🤔 **难点答疑**
1. 为什么要用字典而不是直接遍历两次？
- **两重循环暴力法**虽然也能解决问题，但检查每个数时都要遍历一遍剩下的所有数，时间复杂度是$O(n^2)$。
- **字典法**通过空间换时间：用字典的快速查找能力（检查哈希表的时间复杂度是$O(1)$），总时间复杂度降到$O(n)$。

2. 为什么不会重复使用同一个元素？
比如数组`[3,3]`，target=6：
1. 第一次循环，`i=0`，`num=3`，计算`complement=6-3=3`，此时字典是空的，存入`{3:0}`；
2. 第二次循环，`i=1`，`num=3`，计算`complement=6-3=3`，检查字典是否存在3。**此时`3`已存在（索引0）**，直接返回`[0,1]`。

代码中判断的是字典中是否存在「已经遍历过」的元素，而当前循环的是下一个数，所以不会被算成同一个元素。

🐍 **Python语法小课堂**
- **字典（Dictionary）**：用大括号 `{}` 定义，存储键值对。例如 `hashmap[num] = i` 表示键是 `num`，值是 `i`。
- **enumerate函数**：用于同时获取数组的「索引」和「值」。例如遍历 `nums = [2,7,11]` 时，第一次得到 `i=0, num=2`，第二次 `i=1, num=7`。
- **return语句**：当找到两个数时直接返回一个列表 `[索引1, 索引2]`，立即结束函数。


⚡ **时间复杂度分析**
- **时间复杂度**：$O(n)$。最坏情况下遍历所有元素一次。
- **空间复杂度**：$O(n)$。字典最多需要存储 n 个元素。

## 3. 无重复字符的最长字串

给定一个字符串 `s` ，请你找出其中不含有重复字符的 **最长 子串** 的长度。


**示例 1:**

```
输入: s = "abcabcbb"
输出: 3 
解释: 因为无重复字符的最长子串是 "abc"，所以其长度为 3。
```

**示例 2:**

```
输入: s = "bbbbb"
输出: 1
解释: 因为无重复字符的最长子串是 "b"，所以其长度为 1。
```

**示例 3:**

```
输入: s = "pwwkew"
输出: 3
解释: 因为无重复字符的最长子串是 "wke"，所以其长度为 3。
     请注意，你的答案必须是 子串 的长度，"pwke" 是一个子序列，不是子串。
```

````python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        dic, res, i = {}, 0, -1
        for j in range(len(s)):
            if s[j] in dic:
                i = max(dic[s[j]], i) # 更新左指针 i
            dic[s[j]] = j # 哈希表记录
            res = max(res, j - i) # 更新结果
        return res
````

这个问题是要求在给定的字符串中找到最长无重复字符的子串的长度。我们采用滑动窗口的算法来解决，同时利用哈希表优化效率。以下是详细的思路和代码解释：

**思路解析**

1. **滑动窗口**：使用两个指针 `i` 和 `j` 分别表示窗口的左边界（上一次重复字符的位置）和右边界（当前遍历位置）。窗口内的字符保证不重复。
2. **哈希表记录位置**：用一个字典 `dic` 记录每个字符最后一次出现的位置。
3. **更新左边界**：当遇到重复字符时，将左边界 `i` 更新为**当前字符上次位置**和**原 `i`** 中的较大值，确保窗口内无重复。
4. **维护最大长度**：每次更新窗口后，计算当前窗口长度 `j - i`，并与历史最大值 `res` 比较。

**代码逐行解释**

```python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        dic, res, i = {}, 0, -1          # 初始化：哈希表、结果、左指针
        for j in range(len(s)):           # 遍历字符串，j为右指针
            if s[j] in dic:               # 若当前字符已存在
                i = max(dic[s[j]], i)     # 更新左指针到重复字符的上次位置
            dic[s[j]] = j                 # 记录/更新当前字符的位置
            res = max(res, j - i)         # 计算当前窗口长度，更新结果
        return res
```

**关键点解析**

- **哈希表 `dic`**：快速查找字符是否存在，并保存其最新位置。
- **左指针 `i`**：初始为 `-1` 以正确处理索引 `0`（如 `j=0` 时窗口长度为 `0 - (-1) = 1`）。
- **去重逻辑**：当字符重复时，`i = max(dic[s[j]], i)` 确保左指针只能向右滑动，避免回退。

**示例分析**

以 `s = "abcabcbb"` 为例：
- **`a` 首次出现**（索引0）：存入字典，窗口长度 `0 - (-1) = 1`，更新 `res = 1`。
- **`b` 首次出现**（索引1）：窗口长度 `2`（`1 - (-1)`），`res = 2`。
- **`c` 首次出现**（索引2）：同理，`res = 3`。
- **后续重复字符**：更新 `i` 并维护最大长度，最终得到结果 `3`。

**时间复杂度**

- **O(n)**：每个字符最多被访问两次（`j` 遍历一次，`i` 不回溯），线性时间复杂度。

通过这种算法，可以高效找到最长无重复子串的长度。

## [9. 回文数](https://leetcode.cn/problems/palindrome-number/)

相关标签

![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业


给你一个整数 `x` ，如果 `x` 是一个回文整数，返回 `true` ；否则，返回 `false` 。

回文数是指正序（从左向右）和倒序（从右向左）读都是一样的整数。

- 例如，`121` 是回文，而 `123` 不是。

````python
class Solution:
    def isPalindrome(self, x: int) -> bool:
        # 1. 优化初始判断：负数和末尾为0的正数直接排除
        #    (x % 10 == 0 and x != 0) 比 (x > 0 and x % 10 == 0) 更严谨
        if x < 0 or (x % 10 == 0 and x != 0):
            return False

        rev = 0
        original = x # 依然保留 original 用于最终比较

        # 2. 核心反转逻辑 (保持不变，因为已经很完美了)
        while x > 0:
            rev = rev * 10 + x % 10
            x = x // 10

        # 3. 优化返回值，直接返回比较结果
        return original == rev

# --- 以下是交互式测试部分 ---
sol = Solution()

try:
    # 1. 获取用户输入
    input_str = input("请输入一个整数: ")

    # 2. 【关键】将输入的字符串转换为整数
    input_int = int(input_str)

    # 3. 调用函数并传入整数
    output_result = sol.isPalindrome(input_int)

    # 4. 打印结果
    print(f"整数 {input_int} 是回文数吗? -> {output_result}")

except ValueError:
    # 这是一个“异常处理”，如果用户输入的不是数字，程序不会崩溃
    print("错误：请输入一个有效的整数。")
````

现在我们明白了每一行代码的语法含义，让我们把它们串起来，看看您的**解题思路**是什么。这是一个非常经典的\*\*“整数反转”\*\*算法。

**核心思想**

> 要判断一个数字是不是回文，最直接的方法就是把它**整个颠倒过来**，然后看看**颠倒后的数字**和**原来的数字**是否相等。

您的代码完美地执行了这个思想，分为三个步骤：

**步骤一：预处理 (处理特殊情况)**

这一步就像一个俱乐部门口的保安，先把明显不合格的人拦在外面，提高效率。

  * **`if x < 0:`**：负数（如 `-121`）不可能是回文数，因为它的颠倒（`121-`）和它本身不相等。直接返回 `False`。
  * **`elif x > 0 and x % 10 == 0:`**：一个正数如果个位数是 `0`（如 `120`），那它也不可能是回文数。因为如果它是回文，它的最高位也必须是 `0`（`021`），但一个多位数不能以 `0` 开头。所以也直接返回 `False`。
  * **`elif x == 0:`**：`0` 是一个回文数。直接返回 `True`。

**步骤二：核心引擎 (反转数字的 `while` 循环)**

这是整个算法的心脏。让我们用 `x = 121` 作为例子，看看这个循环是如何工作的。

  * **初始状态**: `original = 121`, `rev = 0`, `x = 121`

  * **第一次循环**:

      * `while x > 0` (121 \> 0) 条件为 True，进入循环。
      * `rev = rev * 10 + x % 10`  =\> `rev = 0 * 10 + 121 % 10` =\> `rev = 0 + 1` =\> `rev` 变为 **`1`**。
      * `x = x // 10` =\> `x = 121 // 10` =\> `x` 变为 **`12`**。
      * **解读**：我们成功地“剥下”了 `121` 的个位数 `1`，并把它“安装”到了 `rev` 的个位上。

  * **第二次循环**:

      * `while x > 0` (12 \> 0) 条件为 True，进入循环。
      * `rev = rev * 10 + x % 10` =\> `rev = 1 * 10 + 12 % 10` =\> `rev = 10 + 2` =\> `rev` 变为 **`12`**。
      * `x = x // 10` =\> `x = 12 // 10` =\> `x` 变为 **`1`**。
      * **解读**：我们剥下了 `12` 的个位数 `2`，并把它安装到了 `rev` 的末尾。

  * **第三次循环**:

      * `while x > 0` (1 \> 0) 条件为 True，进入循环。
      * `rev = rev * 10 + x % 10` =\> `rev = 12 * 10 + 1 % 10` =\> `rev = 120 + 1` =\> `rev` 变为 **`121`**。
      * `x = x // 10` =\> `x = 1 // 10` =\> `x` 变为 **`0`**。
      * **解读**：我们剥下了最后的 `1`，并完成了反转。

  * **循环结束**:

      * `while x > 0` (0 \> 0) 条件为 False，循环停止。

**步骤三：最终裁决**

  * 循环结束后，我们得到 `rev = 121`。
  * 执行最后的 `if original == rev:`，也就是 `if 121 == 121:`。
  * 条件为 `True`，所以函数返回 `True`。

**总结**

它首先通过巧妙的预处理排除了大量无效情况，然后通过一个经典的 `while` 循环和数学运算（取余和整除）实现了数字的反转，最后通过一次简单的比较得出最终结论。这是一个值得深入理解和学习的经典算法。

## 11. 盛水最多的容器

给定一个长度为 `n` 的整数数组 `height` 。有 `n` 条垂线，第 `i` 条线的两个端点是 `(i, 0)` 和 `(i, height[i])` 。

找出其中的两条线，使得它们与 `x` 轴共同构成的容器可以容纳最多的水。

返回容器可以储存的最大水量。

**说明：**你不能倾斜容器。

 

**示例 1：**

![img](https://aliyun-lc-upload.oss-cn-hangzhou.aliyuncs.com/aliyun-lc-upload/uploads/2018/07/25/question_11.jpg)

```
输入：[1,8,6,2,5,4,8,3,7]
输出：49 
解释：图中垂直线代表输入数组 [1,8,6,2,5,4,8,3,7]。在此情况下，容器能够容纳水（表示为蓝色部分）的最大值为 49。
```

**示例 2：**

```
输入：height = [1,1]
输出：1
```

````python
class Solution:
    def maxArea(self, height: List[int]) -> int:
        max_area = 0
        left = 0
        right = len(height) - 1
        
        while left < right:
            current_height = min(height[left], height[right])
            current_width = right - left
            current_area = current_height * current_width
            max_area = max(max_area, current_area)
            
            if height[left] < height[right]:
                left += 1
            else:
                right -= 1
        
        return max_area
````

**问题理解**
假设有一排不同高度的柱子，你需要找到两根柱子，使得它们和底部（x轴）组成的容器能装最多的水。容器装水的量由两个因素决定：
1. **两根柱子的距离**（越远越好）
2. **较矮柱子的高度**（越高越好）

比如：柱子高度为 `[1, 7, 3]`，两根柱子距离为 2（索引 0 和 2），但高度只能按较矮的 1 计算，所以水量是 `1 * 2 = 2`。


**为什么用双指针？**
想象一下：如果我们从最左和最右的柱子开始（即最大宽度），逐步向中间缩小范围。每次移动时，只移动较矮的那根柱子，看看能不能找到更高的柱子。这样可以在减少宽度的同时，尽量保持高度，从而找到更大的水量。

**核心思路**：宽度从最大开始，优先寻找更高的柱子，弥补宽度减少的损失。


**算法步骤**
1. **初始化两个指针**：左指针在数组开头（`left=0`），右指针在数组末尾（`right=len(height)-1`）。
2. **计算当前水量**：水量 = 较矮柱子的高度 × 指针间的距离。
3. **移动较矮的指针**：如果左柱子矮，左指针右移；如果右柱子矮，右指针左移。
4. **重复步骤 2-3**，直到左右指针相遇。
5. **记录过程中的最大水量**。


**代码逐行解释**
```python
class Solution:
    def maxArea(self, height: List[int]) -> int:
        max_area = 0         # 记录最大水量
        left = 0             # 左指针，初始指向第一个柱子
        right = len(height) - 1  # 右指针，初始指向最后一个柱子

        while left < right:  # 只要左右指针没相遇，就继续循环
            # 计算当前两根柱子的水量
            current_height = min(height[left], height[right])  # 较矮柱子的高度
            current_width = right - left  # 两根柱子的距离
            current_area = current_height * current_width  # 当前水量

            # 更新最大水量
            if current_area > max_area:
                max_area = current_area

            # 移动较矮的指针
            if height[left] < height[right]:
                left += 1   # 左柱子矮，左指针右移
            else:
                right -= 1  # 右柱子矮，右指针左移

        return max_area  # 返回最大水量
```


**举个具体例子**
以示例 1 为例：`height = [1, 8, 6, 2, 5, 4, 8, 3, 7]`，数组索引从 0 到 8。

1. **初始状态**：
   - `left=0`（高度1），`right=8`（高度7）
   - 水量 = 1 × 8 = 8
   - 因为左边矮，移动左指针到1。

2. **下一步**：
   - `left=1`（高度8），`right=8`（高度7）
   - 水量 = 7 × 7 = 49（更新最大值）
   - 因为右边矮，移动右指针到7。

3. **继续移动**：
   - `left=1`（高度8），`right=7`（高度3）
   - 水量 = 3 × 6 = 18（不更新最大值）
   - 右边矮，继续移动右指针...

经过多次移动后，最终找到的最大水量是 49。


**为什么这个算法是对的？**
- **不会错过最优解**：每次移动较矮的柱子，是为了寻找更高的柱子。即使移动后宽度变小，但高度可能大幅增加，从而总水量更大。
- **时间复杂度 O(n)**：只需遍历一次数组，非常高效。


**总结**
- **双指针从两端向中间移动**，逐步缩小范围。
- **每次移动较矮的指针**，尝试找到更高的柱子。
- **实时更新最大水量**，最终得到答案。

## 15. 三数之和

给你一个整数数组 `nums` ，判断是否存在三元组 `[nums[i], nums[j], nums[k]]` 满足 `i != j`、`i != k` 且 `j != k` ，同时还满足 `nums[i] + nums[j] + nums[k] == 0` 。请你返回所有和为 `0` 且不重复的三元组。

**注意**：答案中不可以包含重复的三元组。

 

**示例 1：**

```
输入：nums = [-1,0,1,2,-1,-4]
输出：[[-1,-1,2],[-1,0,1]]
解释：
nums[0] + nums[1] + nums[2] = (-1) + 0 + 1 = 0 。
nums[1] + nums[2] + nums[4] = 0 + 1 + (-1) = 0 。
nums[0] + nums[3] + nums[4] = (-1) + 2 + (-1) = 0 。
不同的三元组是 [-1,0,1] 和 [-1,-1,2] 。
注意，输出的顺序和三元组的顺序并不重要。
```

**示例 2：**

```
输入：nums = [0,1,1]
输出：[]
解释：唯一可能的三元组和不为 0 。
```

**示例 3：**

```

输出：[[0,0,0]]
解释：唯一可能的三元组和为 0 。
```

````python
# 定义您的 Solution 类和 threeSum 方法
from typing import List
import json

class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        n = len(nums)
        res = []
        if(not nums or n < 3):
            return []
        nums.sort()
        for i in range(n):
            if(nums[i] > 0):
                break # 优化：如果当前元素大于0，后面的和肯定大于0

            if(i > 0 and nums[i] == nums[i-1]):
                continue # 跳过重复的元素

            L = i + 1
            R = n - 1
            while(L < R):
                current_sum = nums[i] + nums[L] + nums[R]
                if(current_sum == 0):
                    res.append([nums[i], nums[L], nums[R]])
                    # 跳过重复的元素
                    while(L < R and nums[L] == nums[L+1]):
                        L += 1
                    while(L < R and nums[R] == nums[R-1]):
                        R -= 1
                    # 移动双指针继续寻找
                    L += 1
                    R -= 1
                elif(current_sum > 0):
                    R -= 1
                else:
                    L += 1
        return res

# 在一个新的 cell 中添加测试逻辑
solution = Solution()

print("请按提示输入要测试的整数列表 (例如: [-1, 0, 1, 2, -1, -4])")
print("输入 'quit' 或 'exit' 退出测试。")

while True:
    try:
        # 在 Jupyter 中运行此 cell 时，input() 会在输出区域显示输入框
        user_input = input("请输入整数列表: ")
        if user_input.lower() in ['quit', 'exit']:
            break

        # 尝试将输入的字符串解析为 Python 列表
        nums = json.loads(user_input)

        # 确保输入是列表
        if not isinstance(nums, list):
            print("输入无效：请输入一个整数列表。")
            continue

        # 确保列表中的元素是整数
        if not all(isinstance(x, int) for x in nums):
             print("输入无效：列表中的所有元素必须是整数。")
             continue

        # 调用您的函数
        result = solution.threeSum(nums)

        # 打印结果
        print(f"Input: {nums}")
        print(f"Result: {result}")
        print("-" * 20)

    except json.JSONDecodeError:
        print("输入格式错误：请确保输入是有效的JSON格式列表 (例如: [1, 2, 3])。")
    except Exception as e:
        print(f"发生错误: {e}")

print("测试结束。")
````

```python

res = []
```

- **作用**：如果 `nums` 是空数组或长度小于 3，直接返回空列表（无法构成三元组）。
- **关键点**：`not nums` 检查是否为空列表，`n < 3` 判断长度是否不足，避免后续计算浪费时间。

**2. 对数组排序**


- **作用**：将数组排序，让后续计算更容易控制三元组的组合和去重。
- **关键点**：排序后的数组允许我们用“双指针法”高效找到三元组，并方便检测重复元素。

**3. 遍历数组中的每个元素作为第一个数**

```python
for i in range(n):
    # 后续逻辑
```

- **作用**：逐个尝试将 `nums[i]` 作为三元组的第一个数（固定一个数，然后找另外两个数）。
- **疑问**：为什么外层循环从 `i=0` 开始？因为我们要穷举所有可能的第一个数。


**4. **去重第一个数**：避免重复的三元组**

```python
if i > 0 and nums[i] == nums[i-1]:
    continue
```

- **作用**：如果当前 `nums[i]` 和前一个数相同，直接跳过本轮循环（避免重复）。
- **关键点**：只有当 `i>0` 时才检查，确保第一个元素不会被跳过。例如当 `i=0`时不会触发这条语句。
- **例子**：
  - 假设 `nums = [-1, -1, 0, 1]`，在遍历到第二个 `-1` 时会因为 `nums[i] == nums[i-1]` 而跳过，这样就不会重复计算以两个 `-1` 为第一个数的情况。

**5. 初始化双指针**

```python

R = n - 1     # 右指针在数组末尾
```

- **作用**：双指针 `L`（左）和 `R`（右）从当前 `i` 的两侧开始，寻找和为 `0 - nums[i]` 的第二和第三个数。
- **关键点**：初始位置保证 `L` 永远在 `R` 左边，避免重复计算。


**6. 使用双指针寻找符合条件的三元组**

```python
while L < R:
    s = nums[i] + nums[L] + nums[R]
    if s == 0:
        # found a valid triplet
    elif s > 0:
        R -= 1     # 和太大，右指针左移减小值
    else:
        L += 1     # 和太小，左指针右移增大值
```

**情况分三类处理**：

**a. 当前和 `s` 等于 0（找到一个有效解）：**

```python
res.append([nums[i], nums[L], nums[R]])
# 跳过与当前L相同的元素，避免重复
while L < R and nums[L] == nums[L+1]:
    L += 1
while L < R and nums[R] == nums[R-1]:
    R -= 1
L += 1
R -= 1
```

- **作用**：将找到的有效三元组存入结果列表。
- **去重处理**：
  - 移动指针时跳过与当前 `nums[L]` 和 `nums[R]` 相同的元素，防止后续出现相同的三元组。
  - 例如：如果 `nums[L]` 和 `nums[L+1]` 相同，直接跳过 `L += 1`，避免重复计算类似 `[a, b, c]` 和 `[a, b, c]`。

- **关键点**：移动指针后，`L` 和 `R` 的最终位置会改变，保证下次循环时尝试新的数值。

**b. 当前和 `s` 大于 0**：

```python
R -= 1  # 减小总和（因为数组是排序的，右边的数大，左移可以减小R）
```

- 因为三数之和太大，要减少右指针的值，让总和更小。

**c. 当前和 `s` 小于 0**：

```python
L += 1  # 增大总和（左边的数小，右移可以增大L）
```

- 因为三数之和太小，要增大左指针的值，让总和更大。

**完整流程举例说明**

假设输入 `nums = [-1, 0, 1, 2, -1, -4]`，排序后为 `[-4, -1, -1, 0, 1, 2]`。

**步骤示例**

**当 `i = 0` (第一个元素是 -4)**

- 三元组的第一数是 `nums[i]=-4`，因为 `-4` 比 `0` 小，可以继续找。
- L=1 (元素是 -1)， R=5 (元素是 2)。
- 此时 `s = -4 + (-1) + 2 = -3 <0` → 移动左指针到 L=2。
- 重复…直到 L=4 (元素为1)，此时总和为 `-4 +1+2 =-1` 仍然小于0，左指针继续移动。
- 最终 `[ -4, ... ]` 的情况找不到有效解。

**当 `i=1` (第一个元素是 -1)**

- 此时检查 `nums[1] == nums[0]`，但 `-1 != -4`，不跳过。
- 开始双指针遍历：
  - L=2(R=-1), R=5(2) → s=-1 + (-1) + 2 =0 → 找到了！
    - 添加 `[-1, -1, 2]` 到结果。
    - 跳过重复的L（下一位置是 L=2 的下一个元素 L+1=3 → 当前 nums[L]是 -1，nums[L+1]是0 → 无重复，直接移动L+1， R-1 → L=3， R=4）
    - 后续可能找到其他三元组。

**关键优化点总结**

1. **排序**：将数组排序后，简化双指针的移动逻辑。
2. **去重**：
   - 如果当前元素和前一个元素相同（如 `nums[i] == nums[i-1]`），则跳过本轮，避免重复的第一个元素。
   - 在找到一个有效三元组后，跳过左右指针处重复的元素，避免相同组合被多次添加。
3. **双指针移动逻辑**：
   - 根据当前三数之和的大小，选择缩小右指针（减少总和）或扩大左指针（增加总和）。


**时间复杂度**

- 排序时间复杂度为 **O(n log n)**。
- 外层循环遍历每个元素 **O(n)**，内层双指针遍历每个元素对 **O(n)**。
- 总时间复杂度为 **O(n²)**，这是当前最优解法的复杂度。


**总结代码逻辑**

1. 排序数组，使得双指针法可用。
2. 遍历每个可能的第一个元素作为固定值。
3. 对于每个固定值，使用双指针寻找另外两个元素，调整指针移动方向并解决重复问题。
4. 通过跳过相同元素，避免结果重复。

## [18. 四数之和](https://leetcode.cn/problems/4sum/)

中等  相关标签  ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业



给你一个由 `n` 个整数组成的数组 `nums` ，和一个目标值 `target` 。请你找出并返回满足下述全部条件且**不重复**的四元组 `[nums[a], nums[b], nums[c], nums[d]]` （若两个四元组元素一一对应，则认为两个四元组重复）：

- `0 <= a, b, c, d < n`
- `a`、`b`、`c` 和 `d` **互不相同**
- `nums[a] + nums[b] + nums[c] + nums[d] == target`

你可以按 **任意顺序** 返回答案 。

 

**示例 1：**

```
输入：nums = [1,0,-1,0,-2,2], target = 0
输出：[[-2,-1,1,2],[-2,0,0,2],[-1,0,0,1]]
```

**示例 2：**

```
输入：nums = [2,2,2,2,2], target = 8
输出：[[2,2,2,2]]
```

````python
class Solution(object):
    def fourSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[List[int]]
        """
        nums.sort()  # 第一步：对列表进行排序
        n = len(nums)
        result = [] # 用于存储最终结果的列表

        # 第一层循环：固定第一个数字 i
        for i in range(n - 3):
            # 跳过重复的第一个数字
            if i > 0 and nums[i] == nums[i - 1]:
                continue

            # 第二层循环：固定第二个数字 j
            for j in range(i + 1, n - 2):
                # 跳过重复的第二个数字
                if j > i + 1 and nums[j] == nums[j - 1]:
                    continue

                # 双指针查找另外两个数字
                left = j + 1
                right = n - 1

                while left < right:
                    current_sum = nums[i] + nums[j] + nums[left] + nums[right]

                    if current_sum == target:
                        # 找到一组解，添加到结果列表中
                        result.append([nums[i], nums[j], nums[left], nums[right]])

                        # 跳过重复的第三个数字 (left 指针)
                        while left < right and nums[left] == nums[left + 1]:
                            left += 1
                        # 跳过重复的第四个数字 (right 指针)
                        while left < right and nums[right] == nums[right - 1]:
                            right -= 1

                        # 移动双指针继续查找
                        left += 1
                        right -= 1

                    elif current_sum < target:
                        
                        # 和太小，移动 left 指针向右
                        left += 1
                    else:
                        # 和太大，移动 right 指针向左
                        right -= 1

        return result
    
````

**算法思路：**

这个算法的核心思想与我们之前讨论的标准排序 + 双指针方法一致：

1.  **排序 (Sorting):** 首先对输入的数字列表 `nums` 进行排序。排序是这个算法的基础，它使得相同数字相邻，方便我们跳过重复项；同时，它也使得我们可以使用双指针技术来高效地查找剩余的两个数字。
2.  **固定前两个数字 (Fixing Two Pointers):** 使用两层嵌套的 `for` 循环来固定住前两个数字。外层循环 (`i`) 选择第一个数字 `nums[i]`，内层循环 (`j`) 选择第二个数字 `nums[j]`。由于不能重复使用同一个数字，并且为了避免重复的组合，内层循环 `j` 总是从 `i + 1` 开始。
3.  **双指针查找后两个数字 (Two-Pointer Search):** 在固定了 `nums[i]` 和 `nums[j]` 后，问题转化为在剩余的子数组（从 `nums[j + 1]` 到 `nums[n - 1]`）中找到两个数字 `nums[left]` 和 `nums[right]`，使得 `nums[i] + nums[j] + nums[left] + nums[right]` 等于给定的 `target`。这就是一个典型的“两数之和”问题，可以使用双指针来解决。
    *   初始化左指针 `left` 指向子数组的起始位置 (`j + 1`)。
    *   初始化右指针 `right` 指向子数组的结束位置 (`n - 1`)。
    *   在一个 `while left < right` 的循环中：
        *   计算当前四个数字的和 `current_sum = nums[i] + nums[j] + nums[left] + nums[right]`。
        *   **如果 `current_sum` 等于 `target`：** 找到了一个符合条件的四元组。将其添加到结果列表中。然后，为了寻找下一组可能的解，我们需要移动 `left` 和 `right` 指针。同时，为了避免添加重复的四元组，我们需要跳过与当前 `nums[left]` 和 `nums[right]` 相同的相邻元素。
        *   **如果 `current_sum` 小于 `target`：** 说明当前的和太小了，需要增大和。由于数组已经排序，我们可以通过将左指针 `left` 向右移动来选择一个更大的数字。
        *   **如果 `current_sum` 大于 `target`：** 说明当前的和太大了，需要减小和。我们可以通过将右指针 `right` 向左移动来选择一个更小的数字。
4.  **去重 (Handling Duplicates):** 在固定 `i` 和 `j` 的循环中，以及在双指针移动后，都需要检查当前选择的数字是否与前一个数字相同。如果相同，则跳过，以避免重复处理相同的组合，从而确保最终结果中没有重复的四元组。

    *   之所以是 `n - 3`，是因为我们需要固定 `i`、`j`、`left`、`right` 四个位置，至少需要四个元素。如果 `i` 超过 `n - 4`，后面就不够三个元素了。
8.  `if i > 0 and nums[i] == nums[i - 1]: continue`:
    *   **去重**：检查当前元素 `nums[i]` 是否与前一个元素 `nums[i - 1]` 相同。
    *   `i > 0`: 这个条件是必需的，以确保 `i - 1` 的索引是有效的，即不是列表的第一个元素。
    *   `nums[i] == nums[i - 1]`: 判断当前元素是否与前一个元素相等。
    *   `continue`: 如果条件为真，跳过当前循环的剩余部分，直接进入 `for i in range(...)` 的下一次迭代。这 effectively 跳过了处理以重复数字开头的组合。
9.  `for j in range(i + 1, n - 2):`:
    *   开始第二个 `for` 循环。`range(i + 1, n - 2)` 生成一个从 `i + 1` 到 `n - 3` 的整数序列。
    *   `j` 将依次取这个序列中的每个值。
    *   `i + 1`: 确保 `j` 始终在 `i` 的后面，避免重复使用同一个位置的数字，并且固定了第一个数字后，第二个数字可以从它后面开始选。
    *   `n - 2`: 确保后面至少还有两个元素 (`left` 和 `right`) 可供选择。
10. `if j > i + 1 and nums[j] == nums[j - 1]: continue`:
    *   **去重**：检查当前元素 `nums[j]` 是否与前一个元素 `nums[j - 1]` 相同。
    *   `j > i + 1`: 这个条件是必需的，它确保 `j - 1` 不是紧挨着 `i` 的位置（因为紧挨着 `i` 的第一个 `j` 不需要和它前面的数比较）。只有当 `j` 不是 `i + 1` 时，才需要和 `nums[j-1]` 比较来判断是否是重复的第二个数字。
    *   `nums[j] == nums[j - 1]`: 判断当前元素是否与前一个元素相等。
    *   `continue`: 如果条件为真，跳过当前循环的剩余部分，直接进入 `for j in range(...)` 的下一次迭代。这 effectively 跳过了处理以重复数字作为第二个数字的组合。
11. `left = j + 1`:
    *   初始化左指针 `left`，使其指向紧跟在 `j` 后面的元素索引。
12. `right = n - 1`:
    *   初始化右指针 `right`，使其指向列表的最后一个元素的索引。
13. `while left < right:`:
    *   开始双指针的 `while` 循环。当左指针 `left` 的索引小于右指针 `right` 的索引时，循环继续。这意味着 `left` 和 `right` 指向的是不同的元素，并且 `left` 在 `right` 的左边。
14. `current_sum = nums[i] + nums[j] + nums[left] + nums[right]`:
    *   计算当前由 `nums[i]`, `nums[j]`, `nums[left]`, `nums[right]` 四个数字组成的和。
15. `if current_sum == target:`:
    *   如果计算出的和 `current_sum` 等于目标值 `target`。
16. `result.append([nums[i], nums[j], nums[left], nums[right]])`:
    *   将当前找到的这四个数字组成的列表添加到 `result` 列表中。
17. `while left < right and nums[left] == nums[left + 1]: left += 1`:
    *   **去重**：当找到一个解后，为了避免添加重复的四元组，我们需要跳过 `nums[left]` 的重复项。
    *   `left < right`: 确保指针没有交叉。
    *   `nums[left] == nums[left + 1]`: 检查当前 `nums[left]` 是否与它右边的元素相同。
    *   `left += 1`: 如果相同，将 `left` 指针向右移动一位。
    *   这个循环会持续移动 `left` 指针，直到遇到一个与当前 `nums[left]` 不同的元素，或者 `left` 和 `right` 相遇。
18. `while left < right and nums[right] == nums[right - 1]: right -= 1`:
    *   **去重**：同样，为了避免添加重复的四元组，我们需要跳过 `nums[right]` 的重复项。
    *   `left < right`: 确保指针没有交叉。
    *   `nums[right] == nums[right - 1]`: 检查当前 `nums[right]` 是否与它左边的元素相同。
    *   `right -= 1`: 如果相同，将 `right` 指针向左移动一位。
    *   这个循环会持续移动 `right` 指针，直到遇到一个与当前 `nums[right]` 不同的元素，或者 `left` 和 `right` 相遇。
19. `left += 1`:
    *   在找到一个解并处理完重复项后，将 `left` 指针向右移动一位，继续寻找下一组可能的解。
20. `right -= 1`:
    *   同时，将 `right` 指针向左移动一位，继续寻找下一组可能的解。
21. `elif current_sum < target:`:
    *   如果计算出的和 `current_sum` 小于目标值 `target`。
22. `left += 1`:
    *   说明当前的和太小，需要增大。由于数组已排序，将左指针 `left` 向右移动，以包含一个更大的数字。
23. `else:`:
    *   如果 `current_sum` 不等于 `target` 且不小于 `target`，那么它一定大于 `target`。
24. `right -= 1`:
    *   说明当前的和太大，需要减小。由于数组已排序，将右指针 `right` 向左移动，以包含一个更小的数字。
25. `return result`:
    *   在所有循环结束后，返回存储了所有符合条件的四元组的列表 `result`。

**总结：**

这段代码 elegantly 实现了四数之和的排序 + 双指针算法。它通过排序、嵌套循环固定前两个数字，然后使用双指针在剩余部分高效查找后两个数字，并通过巧妙的去重逻辑避免了结果的重复。这是一个非常经典的算法应用示例。

## [19. 删除链表的倒数第 N 个结点](https://leetcode.cn/problems/remove-nth-node-from-end-of-list/)

中等  相关标签  ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业

给你一个链表，删除链表的倒数第 `n` 个结点，并且返回链表的头结点。

 

**示例 1：**

![img](https://assets.leetcode.com/uploads/2020/10/03/remove_ex1.jpg)

```
输入：head = [1,2,3,4,5], n = 2
输出：[1,2,3,5]
```

**示例 2：**

```
输入：head = [1], n = 1
输出：[]
```

**示例 3：**

```
输入：head = [1,2], n = 1
输出：[1]
```

````python
from typing import Optional
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
        
class Solution:
    def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
        dummy = ListNode(0)
        dummy.next = head

        fast = dummy
        slow = dummy
        
        for _ in range(n + 1):
            if fast == None:
                return

            fast = fast.next

        while fast:
            slow = slow.next
            fast = fast.next
        
        slow.next = slow.next.next

        return dummy.next
````

**想象一下链表：**

首先，把链表想象成一串糖葫芦，每个糖葫芦是一个节点（`ListNode`），用竹签串起来，竹签就是 `next` 指针，连接着前一个糖葫芦和后一个糖葫芦。每个糖葫芦里有馅（`val`），就是节点的值。链表的开头就是第一个糖葫芦，结尾就是最后一个糖葫芦，最后一个糖葫芦的竹签没有连接到任何东西（`next` 是 `None`）。

**我们的任务：**

我们的任务是，给你一串糖葫芦（链表），告诉你从后面数第 `n` 个糖葫芦，请你把这个糖葫芦拿掉，然后把剩下的糖葫芦重新串好。

**代码解释：**

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
```
这部分是定义糖葫芦（节点）的结构，它有一个值 (`val`) 和一个指向下一个糖葫芦的竹签 (`next`)。

```python
class Solution:
    def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
```
这是我们要解决问题的函数，输入是第一个糖葫芦 (`head`) 和要去掉的是倒数第几个 (`n`)，输出是去掉糖葫芦后的第一个糖葫芦。

```python
        dummy = ListNode(0)
        dummy.next = head
```
这一步很关键！我们创造了一个 **假糖葫芦** (`dummy`)，它自己没有值（或者值设为 0，不重要），它的竹签 (`next`) 连接到了第一个真糖葫芦 (`head`)。
**为什么需要假糖葫芦？**
想象一下，如果我们要去掉的是第一个真糖葫芦（倒数第 `n` 个恰好是第一个）。如果没有这个假糖葫芦，我们很难找到第一个糖葫芦的“前一个”糖葫芦来修改它的竹签。有了假糖葫芦，不管我们去掉的是哪个真糖葫芦，我们都可以通过修改某个糖葫芦的 `next` 来完成，包括修改假糖葫芦的 `next` 如果要去掉的是第一个真糖葫芦。这个假糖葫芦就像是一个临时的起点，让我们的操作更方便统一。

```python
        fast = dummy
        slow = dummy
```
我们拿出了两个“手指”，一个叫 `fast`，一个叫 `slow`。一开始，这两个手指都指向我们的假糖葫芦 (`dummy`)。

```python
        for _ in range(n + 1):
            fast = fast.next
```
现在，我们让 `fast` 这个手指 **先** 向前移动 `n + 1` 步（也就是指向 `n + 1` 个糖葫芦）。
* 第一次移动：`fast` 从 `dummy` 指向 `head`。
* 第二次移动：`fast` 从 `head` 指向 `head.next`。
* ...
* 第 `n+1` 次移动：`fast` 指向了比 `dummy` 领先 `n + 1` 个位置的糖葫芦。
**为什么是 `n + 1` 步？**
想象一下，`slow` 指针和 `fast` 指针就像两个人跑步。`fast` 先跑了 `n + 1` 步。现在，`fast` 比 `slow` 领先了 `n + 1` 个糖葫芦的距离。

```python
        while fast:
            slow = slow.next
            fast = fast.next
```
接下来，我们让 `slow` 和 `fast` 这两个手指 **一起** 向前移动，每次都移动一个糖葫芦。这个过程一直持续，直到 `fast` 指到没有糖葫芦的地方 (`None`)。
**这个过程的意义是？**
还记得 `fast` 比 `slow` 领先 `n + 1` 步吗？当 `fast` 到达链表的末尾（`None`）时，`slow` 会因为它和 `fast` 之间的固定距离，正好移动到链表倒数第 `n + 1` 个糖葫芦的位置。
为什么是倒数第 `n + 1` 个？
从 `dummy` 开始，`fast` 移动了 `n+1` 步到达了某个位置。然后 `slow` 和 `fast` 一起移动。当 `fast` 到达 `None` 时，它总共移动了从 `dummy` 到 `None` 的距离。`slow` 移动的距离比 `fast` 少了 `n+1` 步。所以，如果 `fast` 从 `dummy` 到达 `None` 的总距离是 `TotalSteps`，那么 `slow` 移动了 `TotalSteps - (n+1)` 步。从 `dummy` 开始移动 `TotalSteps - (n+1)` 步，就指向了链表倒数第 `n + 1` 个节点。

**重要点：`slow` 指向的是要删除节点的前一个节点！**

因为我们要删除倒数第 `n` 个节点，而 `slow` 指向的是倒数第 `n+1` 个节点，所以 `slow` 指向的正是要删除的节点的前一个节点。

```python
        slow.next = slow.next.next
```
现在，`slow` 指向的是我们要去掉的糖葫芦的前一个糖葫芦。`slow.next` 就是我们要去掉的糖葫芦。`slow.next.next` 就是我们要去掉的糖葫芦的下一个糖葫芦。
这一行代码的意思是：把 `slow` 指向的糖葫芦的竹签 (`slow.next`)，从原来指向要去掉的糖葫芦，改成指向要去掉的糖葫芦的下一个糖葫芦 (`slow.next.next`)。
这样，要去掉的糖葫芦就被“跳过”了，它就不在链表里了。

```python
        return dummy.next
```
最后，我们返回假糖葫芦 `dummy` 的竹签所连接的那个糖葫芦 (`dummy.next`)。因为假糖葫芦是连接到整个链表开头的，所以 `dummy.next` 就是我们处理后的新链表的第一个糖葫芦。

**整个过程就像：**

1. 拿出一个“假”的起点。
2. 放两个手指在起点上。
3. 一个手指（`fast`）先向前跳 `n + 1` 步。
4. 两个手指（`slow` 和 `fast`）一起向前走，直到快的手指走到头。
5. 慢的手指停在了要去掉的糖葫芦的前面。
6. 把慢手指指向的糖葫芦的竹签，直接连接到它后面那个糖葫芦的后面那个糖葫芦。
7. 返回从假起点连接到的第一个真糖葫芦。

## [20. 有效的括号](https://leetcode.cn/problems/valid-parentheses/)

简单  相关标签  ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业



给定一个只包括 `'('`，`')'`，`'{'`，`'}'`，`'['`，`']'` 的字符串 `s` ，判断字符串是否有效。

有效字符串需满足：

1. 左括号必须用相同类型的右括号闭合。
2. 左括号必须以正确的顺序闭合。
3. 每个右括号都有一个对应的相同类型的左括号。

 

**示例 1：**

**输入：**s = "()"

**输出：**true

**示例 2：**

**输入：**s = "()[]{}"

**输出：**true

**示例 3：**

**输入：**s = "(]"

**输出：**false

**示例 4：**

**输入：**s = "([])"

**输出：**true

 

**提示：**

- `1 <= s.length <= 104`
- `s` 仅由括号 `'()[]{}'` 组成

````python
# 解法一
class Solution:
    def isValid(self, s: str) -> bool:
        stack = []
        for char in s:
            if char == '(':
                stack.append(')')
            elif char == '[':
                stack.append(']')
            elif char == '{':
                stack.append('}')
            # 合并处理闭合括号的逻辑
            elif char in [')', ']', '}']:
                if not stack:  # 如果栈为空
                    return False # 没有匹配的开括号

                # 如果栈不为空，检查栈顶元素是否匹配
                if stack[-1] == char:
                    stack.pop()
                else:
                    return False

        # 循环结束后，检查栈是否为空
        if stack:
            return False
        return True
````

````python
# 解法二
class Solution:
    def isValid(self, s: str) -> bool:
        """
        使用栈来解决括号匹配问题。
        """
        stack = []
        mapping = {
            ")": "(",
            "}": "{",
            "]": "["
        }
        for char in s:
            if char in mapping:
                top_element = stack.pop() if stack else '#'

                if mapping[char] != top_element:
                    return False
            
            else:
                stack.append(char)


        return not stack
    
solution = Solution()

test_input = input("请输入括号：")
result = solution.isValid(test_input)
print(result)
````

**算法思路：**

这道题目是一个典型的使用 **栈（Stack）** 来解决的问题。栈是一种后进先出（LIFO）的数据结构，非常适合处理需要“匹配”或者“配对”的场景，比如括号匹配。

我们的算法思路如下：

1.  **初始化一个空的栈。** 这个栈将用来存储我们在遍历字符串时遇到的开括号（或者为了方便匹配，存储对应的闭合括号）。
2.  **遍历输入的字符串 `s` 中的每一个字符。**
3.  **如果当前字符是开括号 (`'('`, `'['`, `'{'`)：** 我们将这个开括号对应的 **闭合括号** 压入栈中。这样做的目的是，当我们后续遇到一个闭合括号时，我们期望栈顶的元素就是与这个闭合括号相匹配的那个。
    *   遇到 `'('`，压入 `')'`
    *   遇到 `'['`，压入 `']'`
    *   遇到 `'{'`，压入 `'}'`
4.  **如果当前字符是闭合括号 (`')'`, `']'`, `'}'`)：** 这是需要进行匹配检查的时候。
    *   **首先，检查栈是否为空。** 如果栈已经为空，但是我们遇到了一个闭合括号，这说明没有对应的开括号与它匹配，字符串是无效的，直接返回 `False`。
    *   **如果栈不为空，检查栈顶元素。** 将栈顶元素弹出 (`stack.pop()`)，并检查它是否 **等于** 当前的闭合括号。
        *   如果相等，说明找到了匹配的开括号，继续遍历下一个字符。
        *   如果不相等，说明括号类型不匹配，字符串是无效的，返回 `False`。
5.  **遍历完整个字符串后：**
    *   **如果栈是空的：** 说明所有的开括号都找到了对应的闭合括号并成功匹配弹出，字符串是有效的，返回 `True`。
    *   **如果栈不是空的：** 说明还有一些开括号没有找到对应的闭合括号，字符串是无效的，返回 `False`。

**Python 代码解析：**

```python
class Solution:
    def isValid(self, s: str) -> bool:
        # 初始化一个空列表作为栈
        stack = []

        # 遍历输入的字符串 s
        for char in s:
            # 如果当前字符是开括号
            if char == '(':
                # 将对应的闭合括号压入栈
                stack.append(')')
            elif char == '[':
                stack.append(']')
            elif char == '{':
                stack.append('}')
            # 如果当前字符是闭合括号
            elif char in [')', ']', '}']:
                # 在进行匹配检查之前，先检查栈是否为空
                # Python语法：not stack 在 stack 为空时为 True
                if not stack:
                    # 如果栈为空，但遇到闭合括号，说明没有匹配的开括号，返回 False
                    return False

                # 如果栈不为空，进行匹配检查
                # stack.pop() 取出栈顶元素并从栈中移除
                top_element = stack.pop()
                # 检查取出的栈顶元素是否等于当前的闭合括号
                if top_element != char:
                    # 如果不等于，说明括号不匹配，返回 False
                    return False

        # 遍历完整个字符串后
        # 检查栈是否为空
        # Python语法：if stack: 在 stack 不为空时为 True
        if stack:
            # 如果栈不为空，说明有未匹配的开括号，返回 False
            return False
        # 如果栈为空，说明所有括号都匹配成功，返回 True
        return True
```

**Python 语法解释：**

*   `class Solution:`: 定义一个类，在 LeetCode 等在线判题平台中，通常会将解题代码放在一个类的成员方法中。
*   `def isValid(self, s: str) -> bool:`: 定义一个方法 `isValid`。
    *   `self`: 指代类的实例，在调用类的成员方法时会自动传递。
    *   `s: str`: 参数 `s`，类型注解为字符串 (`str`)。
    *   `-> bool`: 返回值类型注解为布尔值 (`bool`)。
*   `stack = []`: 初始化一个空列表。在 Python 中，列表可以很方便地作为栈来使用。
*   `for char in s:`: 这是一个 `for` 循环，遍历字符串 `s` 中的每一个字符，并将当前字符赋值给变量 `char`。
*   `stack.append(item)`: 列表的方法，用于将 `item` 添加到列表的末尾（模拟栈的压入操作）。
*   `elif char in [')', ']', '}']:`: `elif` 是 `else if` 的缩写。`in` 关键字用于检查一个元素是否在一个序列（如列表、字符串、元组）中。这里判断 `char` 是否是 ')', ']', 或 '}' 中的一个。
*   `if not stack:`: 这是一个布尔判断。在 Python 中，空的列表 (`[]`)、空的字符串 (`''`)、`0`、`None` 等的布尔值都被视为 `False`。非空的列表、字符串等的布尔值被视为 `True`。所以 `not stack` 在 `stack` 为空时为 `True`，在 `stack` 不为空时为 `False`。
*   `return False` / `return True`: 立即结束当前函数的执行，并返回指定的布尔值。
*   `stack.pop()`: 列表的方法，用于移除并返回列表的最后一个元素（模拟栈的弹出操作）。如果在空列表上调用 `pop()` 会引发 `IndexError` 错误，这也是为什么我们需要在 `pop()` 之前检查栈是否为空。
*   `stack[-1]`: 访问列表的最后一个元素。`-1` 是 Python 中列表索引的一种方式，表示倒数第一个元素。同样，如果在空列表上访问 `stack[-1]` 也会引发 `IndexError`。
*   `if stack:`: 循环结束后，再次利用列表的布尔值进行判断。如果 `stack` 不为空（即布尔值为 `True`），表示还有未匹配的开括号。如果 `stack` 为空（即布尔值为 `False`），表示所有括号都匹配成功。

**总结：**

通过使用栈，我们能够有效地跟踪开括号的出现顺序。当遇到闭合括号时，检查栈顶元素是否是对应的开括号（或者说，是否是你之前压入的对应的闭合括号），从而判断括号是否匹配。最后，检查栈是否为空来确保所有的开括号都找到了匹配的闭合括号。这个方法的时间复杂度是 O(n)，其中 n 是字符串的长度，因为我们只遍历一次字符串。空间复杂度是 O(n)，最坏情况下（例如字符串只有开括号），栈的大小会与字符串长度成正比。

给你一个链表，两两交换其中相邻的节点，并返回交换后链表的头节点。你必须在不修改节点内部的值的情况下完成本题（即，只能进行节点交换）。

 

**示例 1：**

![img](https://assets.leetcode.com/uploads/2020/10/03/swap_ex1.jpg)

```
输出：[2,1,4,3]
```

```

输出：[]
```

````python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def swapPairs(self, head: Optional[ListNode]) -> Optional[ListNode]:
        dummy = ListNode(0)
        dummy.next = head
        
        prev = dummy

        while prev.next and prev.next.next:
            first = prev.next
            second = prev.next.next

            prev.next = second

            first.next = second.next

            second.next = first

            prev = first

        return dummy.next
````

3.  **创建虚拟头节点 (`dummy`)**

    想象一下，如果我们要交换链表的前两个节点（原先的 `head` 和 `head.next`）。如果没有虚拟头节点，我们需要特殊处理，因为新的头节点不再是原来的 `head`。有了虚拟头节点，我们总是从 `dummy.next` 开始操作，无论原始链表的头节点是否被交换，新的头节点总是可以通过 `dummy.next` 获取。这使得代码逻辑更统一，无需特殊处理第一个节点对。

5.  **循环遍历和交换 (`while` 循环)**

    ```python
    # Iterate through the list in pairs.
    while prev.next and prev.next.next:
        # ... 交换操作 ...

    ```

    *   **`while prev.next and prev.next.next:`**: 这是一个 `while` 循环。循环会一直执行，只要 `prev.next` **不为空** (`prev.next` 存在，意味着 `prev` 后面至少还有一个节点) **并且** `prev.next.next` **不为空** (`prev.next` 后面还有一个节点，意味着 `prev` 后面至少有两个节点，构成一对)。
        *   这个条件确保我们总是有至少两个节点可以进行交换。如果只剩一个节点或零个节点，循环就会停止。

    *   **循环内部的变量定义:**

        ```python
        # 'first' is the first node of the pair.
        first = prev.next
        # 'second' is the second node of the pair.
        second = prev.next.next
        ```

        *   **`first = prev.next`**: 将 `prev` 的下一个节点赋值给变量 `first`。这是当前要交换的第一个节点。
        *   **`second = prev.next.next`**: 将 `prev` 的下下一个节点赋值给变量 `second`。这是当前要交换的第二个节点。

    *   **执行交换操作:** 这是最关键的部分，通过修改节点的 `next` 指针来实现节点的重新连接。

        ```python
        # Perform the swap:
        # 1. Link 'prev' to the second node.
        prev.next = second
        # 2. Link the first node to the node after the second node.
        first.next = second.next
        # 3. Link the second node to the first node.
        second.next = first
        ```

        我们来分解这三个步骤，想象一下初始状态： `prev -> first -> second -> third ...` (其中 `third` 是 `second.next`)

        1.  **`prev.next = second`**: 将 `prev` 的 `next` 指针从 `first` 改为指向 `second`。现在连接变成了：`prev -> second ...`
        2.  **`first.next = second.next`**: 将 `first` 的 `next` 指针从 `second` 改为指向 `second` 原先指向的节点 (`third`)。现在连接变成了：`prev -> second ...` 和 `first -> third ...`
        3.  **`second.next = first`**: 将 `second` 的 `next` 指针从 `first` 原先指向的节点 (`third`) 改为指向 `first`。现在连接变成了：`prev -> second -> first -> third ...`

        通过这三步，我们就成功地将 `first` 和 `second` 两个节点交换了位置，并且链表的连接关系也正确地调整了。

    *   **更新 `prev` 指针:**

        ```python
        # Move 'prev' to the first node (which is now after the swapped pair)
        # to prepare for the next iteration.
        prev = first
        ```

        *   **`prev = first`**: 将 `prev` 指针移动到刚刚交换完的 `first` 节点上。记住，在交换后，`first` 现在位于 `second` 后面。下一轮循环需要交换的是 `second` 和 `first` 后面的一对节点，而这一对节点的前一个节点就是现在的 `first`。所以，我们将 `prev` 更新为 `first`，为下一轮循环做准备。

6.  **返回结果**

    ```python
    # Return the head of the modified list (which is after the dummy node).
    return dummy.next
    ```

    *   **`return dummy.next`**: 循环结束后，链表的所有节点都完成了两两交换（如果可以交换的话）。虚拟头节点 `dummy` 的下一个节点就是整个新链表的头节点。我们返回 `dummy.next` 作为结果。

**总结一下整个流程：**

1.  创建一个虚拟头节点 `dummy`，让它指向原始链表的头节点。
2.  创建一个指针 `prev`，初始化为 `dummy`。
3.  进入一个循环，条件是 `prev` 后面至少还有两个节点（可以构成一对进行交换）。
4.  在循环内：
    *   找到当前要交换的两个节点 `first` 和 `second`。
    *   通过修改 `prev`、`first`、`second` 的 `next` 指针，重新连接节点，完成 `first` 和 `second` 的位置交换。
    *   将 `prev` 移动到刚刚交换后的第一个节点 (`first`)，为下一轮交换做准备。
5.  循环结束。
6.  返回 `dummy.next`，即交换后链表的新的头节点。

## [27. 移除元素](https://leetcode.cn/problems/remove-element/)

简单  相关标签 ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业



给你一个数组 `nums` 和一个值 `val`，你需要 **[原地](https://baike.baidu.com/item/原地算法)** 移除所有数值等于 `val` 的元素。元素的顺序可能发生改变。然后返回 `nums` 中与 `val` 不同的元素的数量。

假设 `nums` 中不等于 `val` 的元素数量为 `k`，要通过此题，您需要执行以下操作：

- 更改 `nums` 数组，使 `nums` 的前 `k` 个元素包含不等于 `val` 的元素。`nums` 的其余元素和 `nums` 的大小并不重要。
- 返回 `k`。

**用户评测：**

评测机将使用以下代码测试您的解决方案：

```
int[] nums = [...]; // 输入数组
int val = ...; // 要移除的值
int[] expectedNums = [...]; // 长度正确的预期答案。
                            // 它以不等于 val 的值排序。

int k = removeElement(nums, val); // 调用你的实现

assert k == expectedNums.length;
sort(nums, 0, k); // 排序 nums 的前 k 个元素
for (int i = 0; i < actualLength; i++) {
    assert nums[i] == expectedNums[i];
}
```

如果所有的断言都通过，你的解决方案将会 **通过**。

````python
class Solution:
    def removeElement(self, nums: List[int], val: int) -> int:
        slow, fast = 0, 0
        while fast < len(nums):
            if nums[fast] != val:
                nums[slow] = nums[fast]
                slow += 1
            fast += 1
        return slow
    
solution = Solution()
input_nums = input("请输入nums")
nums = [int(x) for x in input_nums.split()]
input_val = input("请输入val")
val = int(input_val)
output_slow = solution.removeElement(nums, val)
print(output_slow)
````

**算法解释：**

这段代码实现了一个原地移除数组中特定值的算法，使用了经典的**双指针**技术。

1.  **目标:** 算法的目的是将数组 `nums` 中所有值不等于 `val` 的元素移动到数组的前面部分，并返回这些元素的数量。同时，算法要求**原地**修改数组，这意味着不能创建一个新的数组来存储结果。
2.  **指针定义:**
    *   `fast`: 这是一个“快指针”，它负责遍历整个原始数组。它从数组的开头（索引 0）开始，逐个访问每个元素，直到数组的末尾。
    *   `slow`: 这是一个“慢指针”，它指向处理后数组中下一个应该放置不等于 `val` 的元素的位置。它也从数组的开头（索引 0）开始。
3.  **遍历过程:**
    *   `fast` 指针向前移动，检查它当前指向的元素 `nums[fast]`。
    *   **如果 `nums[fast]` 不等于 `val`:** 这意味着这是一个需要保留的元素。算法将这个元素移动到 `slow` 指针指向的位置 (`nums[slow] = nums[fast]`)。这样做是因为 `slow` 指针总是指向处理区域的末尾，即下一个空闲位置。移动后，`slow` 指针需要向前移动一位 (`slow += 1`)，为下一个需要保留的元素腾出空间。
    *   **如果 `nums[fast]` 等于 `val`:** 这意味着这是一个需要移除的元素。在这种情况下，我们不做任何操作，`slow` 指针保持不变。等于 `val` 的元素会被 `fast` 指针“跳过”，不会被复制到 `slow` 指针指向的位置，从而达到了移除的效果。
4.  **结果:** 当 `fast` 指针遍历完整个数组后，`slow` 指针的位置就指示了不等于 `val` 的元素有多少个，以及它们在数组中占据了前多少个位置。因此，`slow` 的最终值就是符合条件的元素的数量 `k`。数组的前 `slow` 个元素就是所有不等于 `val` 的元素。

简而言之，`fast` 指针负责找到需要保留的元素，`slow` 指针负责将这些保留的元素按顺序放入数组的前部。

    *   `val: int`: 这是参数 `val` 的类型提示，表明 `val` 应该是一个整数 (`int`)。
    *   `-> int`: 这是函数的返回类型提示，表明函数应该返回一个整数 (`int`)。

## [28. 找出字符串中第一个匹配项的下标](https://leetcode.cn/problems/find-the-index-of-the-first-occurrence-in-a-string/)

简单   相关标签  ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业



给你两个字符串 `haystack` 和 `needle` ，请你在 `haystack` 字符串中找出 `needle` 字符串的第一个匹配项的下标（下标从 0 开始）。如果 `needle` 不是 `haystack` 的一部分，则返回 `-1` 。

 

**示例 1：**

```
输入：haystack = "sadbutsad", needle = "sad"
输出：0
解释："sad" 在下标 0 和 6 处匹配。
第一个匹配项的下标是 0 ，所以返回 0 。
```

**示例 2：**

```
输入：haystack = "leetcode", needle = "leeto"
输出：-1
解释："leeto" 没有在 "leetcode" 中出现，所以返回 -1 。
```

````python
class Solution():
    def strStr(self, haystack:str, needle:str) -> int:
        hay = len(haystack)
        need = len(needle)
        for i in range(hay - need + 1):
            substring = haystack[i: i + need]

            if substring == needle:
                return i
        return -1
solution = Solution()
input_haystack = input("请输入haystack：")
input_needle = input("请输入needle：")
output_xiabiao = solution.strStr(input_haystack, input_needle)
print(output_xiabiao)
````

给你两个字符串 `haystack` 和 `needle` ，请你在 `haystack` 字符串中找出 `needle` 字符串的第一个匹配项的下标（下标从 0 开始）。如果 `needle` 不是 `haystack` 的一部分，则返回 `-1` 。

**示例**

输入: `haystack = "sadbutsad", needle = "sad"`
输出: `0`
解释: "sad" 在 "sadbutsad" 中出现了，下标为 0 。

输入: `haystack = "leetcode", needle = "leeto"`
输出: `-1`
解释: "leeto" 没有在 "leetcode" 中出现。

**解题思路**

对于初学者来说，最直观的思路就是 **暴力匹配（Brute Force）**。这种方法模拟了我们在一个文本中查找一个短语的最自然的方式：从文本的开头开始，逐个位置检查是否存在目标短语。

具体来说，暴力匹配的步骤如下：

1. **遍历 `haystack` 的每一个可能的起始位置。** 一个可能的起始位置是指 `haystack` 中可以容纳下整个 `needle` 字符串的起始位置。如果 `haystack` 的长度是 `n`，`needle` 的长度是 `m`，那么可能的起始位置是从下标 `0` 到 `n - m`。因为如果从 `n - m + 1` 开始，剩余的字符数不足 `m`，无法容纳 `needle`。
2. **在每一个可能的起始位置，截取 `haystack` 中从该位置开始、长度为 `m` 的子字符串。**
3. **比较截取的子字符串是否与 `needle` 完全相等。**
4. **如果相等，说明找到了第一个匹配项，立即返回当前的起始位置（下标）。**
5. **如果遍历完所有可能的起始位置都没有找到匹配项，则说明 `needle` 不在 `haystack` 中，返回 `-1`。**

**特殊情况处理**

* 如果 `needle` 是空字符串 `""`，根据题目约定，空字符串是任何字符串的子串，它的第一个匹配项的下标是 0。我们需要在代码中处理这种情况。
* 如果 `needle` 的长度大于 `haystack` 的长度，那么 `needle` 不可能在 `haystack` 中出现，可以直接返回 -1。

**Python 代码实现**

```python
class Solution:
    def strStr(self, haystack: str, needle: str) -> int:
        # 1. 特殊情况处理：如果 needle 是空字符串，返回 0
        if not needle:
            return 0

        # 获取字符串长度
        haystack_len = len(haystack)
        needle_len = len(needle)

        # 2. 特殊情况处理：如果 needle 的长度大于 haystack，返回 -1
        if needle_len > haystack_len:
            return -1

        # 3. 遍历 haystack 的每一个可能的起始位置
        # 可能的起始位置范围是 [0, haystack_len - needle_len]
        for i in range(haystack_len - needle_len + 1):
            # 4. 截取 haystack 中从当前位置 i 开始，长度为 needle_len 的子字符串
            substring = haystack[i : i + needle_len]

            # 5. 比较截取的子字符串是否与 needle 相等
            if substring == needle:
                # 如果相等，返回当前起始位置 i
                return i

        # 6. 如果循环结束都没有找到匹配项，返回 -1
        return -1
```

**代码解释**

* `if not needle:`：检查 `needle` 是否为空。在 Python 中，空字符串被视为 `False`。
* `haystack_len = len(haystack)` 和 `needle_len = len(needle)`：使用内置函数 `len()` 获取字符串长度。
* `if needle_len > haystack_len:`：检查 `needle` 是否比 `haystack` 长。
* `for i in range(haystack_len - needle_len + 1):`：使用 `range()` 函数生成从 0 到 `haystack_len - needle_len` 的整数序列，作为 `haystack` 中子字符串的起始下标。
* `substring = haystack[i : i + needle_len]`：使用 Python 的切片语法 `[start : end]` 截取 `haystack` 中从 `i` 开始到 `i + needle_len` 之前（不包含 `i + needle_len`）的子字符串。这段子字符串的长度正好是 `needle_len`。
* `if substring == needle:`：直接比较两个字符串是否相等。
* `return i`：如果相等，返回当前的起始下标 `i`。
* `return -1`：如果循环结束都没有找到匹配项，说明 `needle` 不在 `haystack` 中，返回 -1。

**复杂度分析**

* **时间复杂度：** 在最坏的情况下，我们需要对 `haystack` 中的每一个可能的起始位置都进行一次与 `needle` 的比较。假设 `haystack` 的长度为 `n`，`needle` 的长度为 `m`。可能的起始位置大约有 `n - m + 1` 个。每次比较都需要遍历 `needle` 的 `m` 个字符。因此，总的时间复杂度大约是 O((n - m + 1) * m)，简化为 **O(n * m)**。
* **空间复杂度：** 我们只使用了几个变量来存储字符串长度和循环变量，以及一个临时变量 `substring`。临时变量 `substring` 的最大长度是 `m`。因此，空间复杂度是 **O(m)**，或者如果我们不考虑存储输出字符串的空间，可以认为是 **O(1)**（因为 `substring` 的空间可以被重用）。

**总结**

暴力匹配方法是解决字符串匹配问题最简单直观的方式。虽然它的时间复杂度相对较高，但在 `haystack` 和 `needle` 长度不太大的情况下，它通常能够满足需求，并且非常容易理解和实现。对于初学者来说，这是学习字符串匹配算法的一个很好的起点。在理解了暴力匹配的基础上，可以进一步学习更高效的算法，如 KMP 算法。

## [34. 在排序数组中查找元素的第一个和最后一个位置](https://leetcode.cn/problems/find-first-and-last-position-of-element-in-sorted-array/)

中等



给你一个按照非递减顺序排列的整数数组 `nums`，和一个目标值 `target`。请你找出给定目标值在数组中的开始位置和结束位置。

如果数组中不存在目标值 `target`，返回 `[-1, -1]`。

你必须设计并实现时间复杂度为 `O(log n)` 的算法解决此问题。

 

**示例 1：**

```
输入：nums = [5,7,7,8,8,10], target = 8
输出：[3,4]
```

**示例 2：**

```
输入：nums = [5,7,7,8,8,10], target = 6
输出：[-1,-1]
```

**示例 3：**

```
输入：nums = [], target = 0
输出：[-1,-1]
```


````python
class Solution:
    def searchRange(self, nums: List[int], target: int) -> List[int]:
        def getRightBorder(nums:List[int], target:int) -> int:
            left, right = 0, len(nums)-1
            rightBoder = -2 # 记录一下rightBorder没有被赋值的情况
            while left <= right:
                middle = left + (right-left) // 2
                if nums[middle] > target:
                    right = middle - 1
                else: # 寻找右边界，nums[middle] == target的时候更新left
                    left = middle + 1
                    rightBoder = left
    
            return rightBoder
        
        def getLeftBorder(nums:List[int], target:int) -> int:
            left, right = 0, len(nums)-1 
            leftBoder = -2 # 记录一下leftBorder没有被赋值的情况
            while left <= right:
                middle = left + (right-left) // 2
                if nums[middle] >= target: #  寻找左边界，nums[middle] == target的时候更新right
                    right = middle - 1
                    leftBoder = right
                else:
                    left = middle + 1
            return leftBoder
        leftBoder = getLeftBorder(nums, target)
        rightBoder = getRightBorder(nums, target)
        # 情况一
        if leftBoder == -2 or rightBoder == -2: return [-1, -1]
        # 情况三
        if rightBoder -leftBoder >1: return [leftBoder + 1, rightBoder - 1]
        # 情况二
        return [-1, -1]
````

**这段代码是做什么的？**

想象你有一串已经排好顺序的数字，比如 `[1, 2, 3, 3, 3, 4, 5]`。现在你想知道某个特定的数字（我们称之为 `target`）在这个数字串里第一次出现和最后一次出现的位置分别是哪里。如果这个数字不存在，你就需要告诉我。

这段代码就是用来解决这个问题的。它接受一个排序好的数字列表（`nums`）和一个你想找的数字（`target`），然后返回一个包含两个数字的列表：第一个数字是 `target` 第一次出现的位置的索引（也就是它在这个列表里的“门牌号”），第二个数字是 `target` 最后一次出现的位置的索引。如果 `target` 不存在，它就返回 `[-1, -1]`。

**为什么需要一个特殊的算法？**

你可能会想，我直接一个个看数字不就行了吗？没错，对于短的数字列表，这样做当然可以。但如果列表非常长，比如有一百万个数字，一个个看就会非常慢。因为数字是排好顺序的，我们可以利用这个特点来更快地找到目标数字，这也就是我们所说的“算法”。

这段代码使用了“二分查找”这个算法。想象你在猜一个1到100之间的数字，你的朋友告诉你“大了”或“小了”。你每次都能排除掉一半的数字，这样很快就能猜到。二分查找就像这样，它不断地把搜索范围缩小一半，直到找到目标数字或者确定目标数字不存在。

**代码结构概览**

这段代码主要由三个部分组成：

1.  一个主函数 `searchRange`：这是你调用的时候会执行的函数，它负责调用另外两个辅助函数，并根据它们的结果给出最终答案。
2.  一个辅助函数 `getRightBorder`：这个函数用来找到 `target` 最后一次出现位置的“右边界”。
3.  一个辅助函数 `getLeftBorder`：这个函数用来找到 `target` 第一次出现位置的“左边界”。

**详细解析 `getRightBorder` 函数**

这个函数的目标是找到 `target` 的右边界。听起来有点抽象，对吧？我们来具体看看：

*   `nums: List[int]`：这是一个参数，表示我们要在这个数字列表里找。`List[int]` 告诉我们这个参数是一个列表，里面装的是整数。
*   `target: int`：这是我们要找的那个数字。`int` 告诉我们这是一个整数。
*   `-> int`：这表示这个函数最后会返回一个整数。

函数体内的代码：

*   `left, right = 0, len(nums)-1`: 这行代码初始化了两个变量，`left` 和 `right`。它们代表当前搜索范围的左右两端的索引。一开始，我们搜索整个列表，所以 `left` 是第一个数字的索引（0），`right` 是最后一个数字的索引（列表长度减1）。
*   `rightBoder = -2`: 这里初始化了一个变量 `rightBoder`，用来记录找到的右边界。为什么初始化为 `-2` 呢？这是一个“标记”，表示目前还没有找到有效的右边界。你可以把它想象成一个初始的、不可能出现的值，用来区分找到的边界。
*   `while left <= right:`: 这是一个循环。只要 `left` 小于或等于 `right`，就说明我们的搜索范围还有效，继续循环。
*   `middle = left + (right-left) // 2`: 这行代码计算出当前搜索范围中间位置的索引。`// 2` 是整数除法，确保结果是整数索引。为什么要用 `left + (right-left) // 2` 而不是 `(left + right) // 2` 呢？主要是为了防止 `left + right` 太大导致计算溢出（虽然在 Python 中不太常见，但这是一个好的编程习惯）。
*   `if nums[middle] > target:`: 如果中间位置的数字比 `target` 大，说明 `target` 如果存在的话，肯定在中间位置的左边。所以我们将搜索范围的右边界 `right` 移动到 `middle - 1`。
*   `else: # 寻找右边界，nums[middle] == target的时候更新left`: 如果中间位置的数字小于或等于 `target`，说明 `target` 可能在中间位置的右边（包括中间位置本身）。我们想要找到 **右边界**，所以即使 `nums[middle]` 等于 `target`，我们也不能确定这就是最后一个 `target`，可能后面还有。所以我们将搜索范围的左边界 `left` 移动到 `middle + 1`，并 **同时** 更新 `rightBoder = left`。为什么更新为 `left` 呢？因为在我们跳出循环时，`left` 会指向第一个 **大于** `target` 的元素的索引。那么 `left - 1` 就是最后一个小于或等于 `target` 的元素的索引，也就是 `target` 的右边界。我们在这里提前将 `rightBoder` 设置为 `left`，这样在循环结束后，`rightBoder` 的值恰好就是我们要找的右边界索引的 **下一个** 位置。
*   `return rightBoder`: 循环结束时，返回 `rightBoder` 的值。

**详细解析 `getLeftBorder` 函数**

这个函数的目标是找到 `target` 的左边界。和右边界类似，我们来看看：

*   `nums: List[int]`：同上，数字列表。
*   `target: int`：同上，要找的数字。
*   `-> int`：同上，返回一个整数。

函数体内的代码：

*   `left, right = 0, len(nums)-1`: 和 `getRightBorder` 一样，初始化搜索范围。
*   `leftBoder = -2`: 同样，初始化一个标记变量 `leftBoder`。
*   `while left <= right:`: 同上，循环条件。
*   `middle = left + (right-left) // 2`: 同上，计算中间位置。
*   `if nums[middle] >= target: #  寻找左边界，nums[middle] == target的时候更新right`: 如果中间位置的数字大于或等于 `target`，说明 `target` 如果存在的话，可能在中间位置的左边（包括中间位置本身）。我们想要找到 **左边界**，所以即使 `nums[middle]` 等于 `target`，我们也不能确定这就是第一个 `target`，可能前面还有。所以我们将搜索范围的右边界 `right` 移动到 `middle - 1`，并 **同时** 更新 `leftBoder = right`。为什么更新为 `right` 呢？和右边界类似，在我们跳出循环时，`right` 会指向第一个 **小于** `target` 的元素的索引。那么 `right + 1` 就是第一个大于或等于 `target` 的元素的索引，也就是 `target` 的左边界。我们在这里提前将 `leftBoder` 设置为 `right`，这样在循环结束后，`leftBoder` 的值恰好就是我们要找的左边界索引的 **前一个** 位置。
*   `else: left = middle + 1`: 如果中间位置的数字小于 `target`，说明 `target` 如果存在的话，肯定在中间位置的右边。所以我们将搜索范围的左边界 `left` 移动到 `middle + 1`。
*   `return leftBoder`: 循环结束时，返回 `leftBoder` 的值。

**回到主函数 `searchRange`**

现在我们理解了两个辅助函数的作用，回到主函数看看它们是如何被使用的：

*   `leftBoder = getLeftBorder(nums, target)`: 调用 `getLeftBorder` 函数，找到左边界（前一个）的索引，并将结果存到 `leftBoder` 变量里。
*   `rightBoder = getRightBorder(nums, target)`: 调用 `getRightBorder` 函数，找到右边界（下一个）的索引，并将结果存到 `rightBoder` 变量里。

接下来就是根据这两个边界来判断 `target` 的出现范围了：

*   `# 情况一 if leftBoder == -2 or rightBoder == -2: return [-1, -1]`: 这是检查 `target` 是否不存在的情况。如果 `leftBoder` 或 `rightBoder` 仍然是初始的 `-2`，说明在整个搜索过程中都没有找到任何符合条件的边界，因此 `target` 不存在，返回 `[-1, -1]`。
*   `# 情况三 if rightBoder -leftBoder >1: return [leftBoder + 1, rightBoder - 1]`: 这是检查 `target` 存在的情况。前面我们提到，`leftBoder` 是左边界的 **前一个** 索引，`rightBoder` 是右边界的 **下一个** 索引。如果 `rightBoder - leftBoder > 1`，说明在 `leftBoder` 和 `rightBoder` 之间至少有一个位置，这个位置就是 `target` 的实际出现范围。`leftBoder + 1` 就是第一个 `target` 的索引，`rightBoder - 1` 就是最后一个 `target` 的索引。
*   `# 情况二 return [-1, -1]`: 如果代码执行到这里，说明既不是情况一（目标不存在），也不是情况三（目标存在且有范围）。这唯一剩下的可能性就是 `target` 虽然在列表中，但它只有一个，或者根本不存在，而上面的情况一已经排除了完全不存在的情况。那么为什么会出现这种情况呢？考虑 `target` 在列表中的边缘位置，或者 `target` 只出现一次。在这种情况下，`rightBoder - leftBoder` 可能等于1。例如，如果 `target` 只出现一次，那么 `leftBoder` 和 `rightBoder` 会紧挨着，`rightBoder - leftBoder` 就是 1。这时说明 `target` 没有一个“范围”，或者说，它根本没有出现。因此返回 `[-1, -1]`。

**总结和一些Python语法点**

*   **函数定义:** Python 中使用 `def` 关键字来定义函数。例如 `def searchRange(...)`。
*   **类型提示:** `nums: List[int]` 和 `-> int` 是类型提示，它们帮助我们理解函数的参数类型和返回值类型。这是一种很好的编程习惯，让代码更容易阅读和理解。
*   **列表索引:** 在 Python 中，可以使用方括号 `[]` 来访问列表中特定位置的元素。例如 `nums[middle]`。
*   **变量赋值:** 使用等号 `=` 来给变量赋值。例如 `left, right = 0, len(nums)-1`。
*   **循环:** `while` 循环会一直执行，直到后面的条件不再满足。
*   **条件判断:** `if`, `elif` (else if), `else` 用于根据条件执行不同的代码块。
*   **整数除法:** `//` 进行整数除法，结果总是整数。
*   **返回:** `return` 关键字用于从函数中返回值。

**为什么采用这种找边界而不是直接找目标的方式？**

这段代码没有直接去寻找任意一个 `target` 的位置，然后向两边扩散。而是分别用两个二分查找来寻找 **左边界** 和 **右边界**。这样做的好处是：

1.  **效率高:** 两个独立的二分查找的效率仍然是 O(log n)，比线性查找 O(n) 快很多。
2.  **逻辑清晰:** 分开寻找左右边界，每次二分查找都有明确的目标，避免了在找到一个 `target` 后还需要额外处理左右边界的情况。

**可以改进的地方吗？**

这段代码已经是一个有效的解决方案了。不过，在一些情况下，可以对边界的初始化和判断进行微调，使其逻辑更加紧凑。例如，可以将初始的 `rightBoder` 和 `leftBoder` 设置为和 `left` 或 `right` 相关的值，这样可能可以简化最后的判断逻辑。但对于初学者来说，使用 `-2` 这样的特殊标记可以更直观地理解“未找到”的状态。

## 42. 接雨水

给定 `n` 个非负整数表示每个宽度为 `1` 的柱子的高度图，计算按此排列的柱子，下雨之后能接多少雨水。


**示例 1：**

![img](https://assets.leetcode-cn.com/aliyun-lc-upload/uploads/2018/10/22/rainwatertrap.png)

```
输入：height = [0,1,0,2,1,0,1,3,2,1,2,1]
输出：6
解释：上面是由数组 [0,1,0,2,1,0,1,3,2,1,2,1] 表示的高度图，在这种情况下，可以接 6 个单位的雨水（蓝色部分表示雨水）。 
```

**示例 2：**

```
输入：height = [4,2,0,3,2,5]
输出：9
```

````python
class Solution:
    def trap(self, height: List[int]) -> int:
        if not height:
            return 0
        
        left, right = 0, len(height) - 1
        max_left, max_right = height[left], height[right]
        water = 0
        
        while left <= right:
            if max_left <= max_right:
                max_left = max(max_left, height[left])
                water += max_left - height[left]
                left += 1
            else:
                max_right = max(max_right, height[right])
                water += max_right - height[right]
                right -= 1
                
        return water
````

### 🌧 **问题理解**
想象这是一排高低不一的水坝，下雨后雨水会积在低洼处。我们需要计算总的积水量。关键规律：**每个位置能积多少水，取决于它左边最高墙和右边最高墙中更矮的那个**。

举个例子：
```
高度 [3, 1, 2]
索引  0  1  2
```
- 中间的1号位置，左边最高是3，右边最高是2，较矮的是2
- 它能积的水量是 (较短边的高度 - 当前高度) = 2 - 1 = 1


### 🛠 **解题思路**
#### 双指针法（最优解）
1. **准备两个指针**：`left`在最左端，`right`在最右端
2. **记录左右最大值**：`max_left`记录左边最高墙，`max_right`记录右边最高墙
3. **移动指针的规则**：
   - 如果左边较矮，说明左边限制水量，处理左指针位置的水量
   - 如果右边较矮，处理右指针的水量
4. **水量累加**：根据左右最大值中的较小值计算当前列的积水量


### 📝 **代码解释（逐行分析）**
```python
class Solution:
    def trap(self, height: List[int]) -> int:
        if not height:  # 如果高度列表为空，直接返回0
            return 0
      
        # 初始化左右指针和最大值
        left, right = 0, len(height) - 1
        max_left, max_right = height[left], height[right]
        water = 0
      
        while left <= right:  # 当左右指针未相遇时继续
            if max_left <= max_right:
                # 左边较矮，处理左指针位置
                max_left = max(max_left, height[left])  # 更新左边最大值
                water += max_left - height[left]  # 计算积水量并累加
                left += 1  # 左指针右移
            else:
                # 右边较矮，处理右指针位置
                max_right = max(max_right, height[right])
                water += max_right - height[right]
                right -= 1  # 右指针左移
              
        return water
```


### 🧠 **关键Python语法知识**

#### 1. `if not height:`
- **作用**：检查列表是否为空
- **解释**：如果`height`是空列表（如`height=[]`），`not height`返回True，直接返回0

#### 2. `len(height)`
- **作用**：获取列表长度
- **示例**：如果`height = [0,1,0]`，`len(height)`返回3

#### 3. `while left <= right:`
- **作用**：循环直到左右指针相遇
- **流程图**：
  ```
  初始状态：left在最左，right在最右 → 移动更矮的一侧指针
              ↓              
  循环直到 left > right
  ```

#### 4. `max_left = max(max_left, height[left])`
- **作用**：动态更新左侧最大值
- **示例**：原`max_left`是3，当前`height[left]`是5 → `max_left`变为5

#### 5. `water += max_left - height[left]`
- **计算水量**：当前列的积水量 = 最小限制高度 - 当前高度
- **示例**：
  - 最小限制高度是左边的4，当前高度是1 → 水量为3
  - 最小限制高度是右边的3，当前高度是0 → 水量为3

## 49. 字母异位词分组

给你一个字符串数组，请你将 **字母异位词** 组合在一起。可以按任意顺序返回结果列表。

**字母异位词** 是由重新排列源单词的所有字母得到的一个新单词。

 

**示例 1:**

```
输入: strs = ["eat", "tea", "tan", "ate", "nat", "bat"]
输出: [["bat"],["nat","tan"],["ate","eat","tea"]]
```

**示例 2:**

```
输入: strs = [""]
输出: [[""]]
```

**示例 3:**

```
输入: strs = ["a"]
输出: [["a"]]
```

````python
class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        ans = {}
        for s in strs:
            key = ''.join(sorted(s))
            if key in ans:
                ans[key].append(s)
            else:
                ans[key] = [s]  # 初始化为包含当前字符串的列表
        return list(ans.values())
````

### **1. 代码结构与整体思路**
这段代码的目标是将字符串数组中所有字母异位词分组。其核心思想与使用 `defaultdict` 的版本完全一致，只是改用**普通字典**代替 `defaultdict` 来实现：

1. **字母异位词的定义**：两个字符串是字母异位词当且仅当它们的字符排序后结果相同。例如：
   - `"eat"`、`"tea"` 排序后均为 `"aet"`。
2. **分组逻辑**：
   - **键（Key）**：排序后的字符串（如 `"aet"`）作为字典的键。
   - **值（Value）**：所有对应原始字符串的列表（如 `["eat", "tea", ...]`）。

**与`defaultdict`版的区别**：
- 普通字典需要显式检查键是否存在（`if key in ans`）。
- `defaultdict`版本则通过 `defaultdict(list)` 自动创建新键对应的列表。

### **2. 代码逐行解释**
```python

```

#### **3.3 生成排序后的键**

- **`sorted(s)`**：
  - 字符串 `s` 是不可变类型，无法直接排序，需转为列表。例如：
    - 若 `s = "eat"`，则 `sorted(s)` 返回 `['a', 'e', 't']`。
- **`''.join(...)`**：
  - 将排序后的字符列表合并为字符串。例如：

### **4. 为什么这段代码可能比 `defaultdict` 版更快？**
虽然 `defaultdict` 设计初衷是为了简化代码逻辑，但在某些情况下，普通字典的显式判断可能更快。以下是可能的原因：

#### **4.1 减少函数调用开销**
- **`defaultdict` 的 `__missing__` 方法**：
  - 当访问一个不存在的键时，`defaultdict` 会调用 `__missing__` 方法来自动创建新键。例如：
    ```python
    # defaultdict的底层逻辑
    class defaultdict(dict):
        def __missing__(self, key):
            # 自动生成默认值并为此键创建 entry
            self[key] = val = self.default_factory()
            return val
    ```
  - 每次访问不存在的键时，都要调用 `__missing__` 方法，涉及额外的函数调用和逻辑判断。
- **普通字典的显式判断**：
  - 通过 `if key in ans` 直接判断键是否存在，避免了 `defaultdict` 的方法调用开销，可能在大规模数据时更高效。

#### **4.2 减少对象创建的额外开销**
- **`defaultdict` 的初始化**：
  - `defaultdict(list)` 需要维护一个 `default_factory` 属性，而普通字典无需此属性。
- **键的重复访问效率**：
  - 当键已存在时（常见于大量异位词的场景），直接通过 `ans[key].append(s)` 访问，无需再触发 `__missing__` 方法，效率更高。

#### **4.3 基准测试示例（假设数据量较大）**
假设输入 `strs` 是一个包含大量重复异位词的列表，例如：
```python

```
- **普通字典版本**：在大部分键已存在时，`if key in ans` 的判断可能比 `defaultdict` 的 `__missing__` 更快。
- **`defaultdict` 版本**：需在每个首次访问新键时调用方法，可能带来额外的开销。

### **5. 代码优化总结**
- **纯字典的显式条件判断**:
  - 在键已存在的情况下，直接操作字典更高效。
- **无需依赖第三方库**：
  - `defaultdict` 需要导入 `collections`，而普通字典无需额外导入。

**提示：**

- `listA` 中节点数目为 `m`
- `listB` 中节点数目为 `n`
- `1 <= m, n <= 3 * 104`
- `1 <= Node.val <= 105`
- `0 <= skipA <= m`
- `0 <= skipB <= n`
- 如果 `listA` 和 `listB` 没有交点，`intersectVal` 为 `0`
- 如果 `listA` 和 `listB` 有交点，`intersectVal == listA[skipA] == listB[skipB]`

````python
# Definition for singly-linked list.
from typing import Optional

class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

class Solution:
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> Optional[ListNode]:
        # 初始化两个指针
        ptrA, ptrB = headA, headB
        
        # 遍历链表
        while ptrA != ptrB:
            # 如果 ptrA 到达链表 A 的末尾，则将其移动到链表 B 的头节点
            ptrA = ptrA.next if ptrA else headB
            # 如果 ptrB 到达链表 B 的末尾，则将其移动到链表 A 的头节点
            ptrB = ptrB.next if ptrB else headA
        
        # 返回相交节点（如果存在），否则返回 None
        return ptrA
````

以下是这段代码的 **详细分步解释**，结合算法思路和 Python 语法说明：

### 一、**问题目标**
找到两个单链表的相交节点（即两个链表从某个节点开始，后续全部节点完全重合），若不相交返回 `None`。

### 二、**算法思路（双指针法）**
核心思想是 **消除两个链表的长度差**，让两个指针走相同的总路程来确保相遇。

1. **流程**：
   1. 指针 `ptrA` 从链表 A 头部开始遍历。
   2. 指针 `ptrB` 从链表 B 头部开始遍历。
   3. 当 `ptrA` 走完链表 A，它跳到链表 B 头部继续走。
   4. 当 `ptrB` 走完链表 B，它跳到链表 A 头部继续走。
   5. 当两个指针走过相同长度（A长度 + B长度）后：
      - **若相交**：会在交点处相遇（因为此时走过的节点数相同）。
      - **若不相交**：最终会同时走到 `None`。

### 三、**代码逐行解释**

#### 1. **定义链表节点（已提供）**
```python
class ListNode:
    def __init__(self, x):
        self.val = x    # 节点存储的值
        self.next = None  # 指向下一个节点的指针
```
- `ListNode` 是链表的节点类，每个节点包含值 `val` 和指向下一个节点的 `next`。

#### 2. **方法定义与参数**
```python
def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> Optional[ListNode]:
```
- `self`: 类方法的第一个参数，表示实例自身（`Solution` 类的实例）。
- **类型注解**（Python3特性）：
  - `headA: ListNode`: 表示 `headA` 参数应为 `ListNode` 类型的实例。
  - `-> Optional[ListNode]`: 表示返回值可能是一个 `ListNode` 或 `None`。
  - `Optional` 需要从 `typing` 模块导入（代码中未显式导入，可能是 LeetCode 环境自动处理）。

#### 3. **初始化指针**
```python
ptrA, ptrB = headA, headB
```
- 初始化两个指针 `ptrA` 和 `ptrB`，分别指向两个链表的头部。

#### 4. **双指针遍历**
```python
while ptrA != ptrB:
    ptrA = ptrA.next if ptrA else headB
    ptrB = ptrB.next if ptrB else headA
```

- **循环条件**：只要两指针不相等，就继续遍历。
<br>

- **`ptrA` 的移动逻辑**：
  - 如果 `ptrA` 未到末尾（`ptrA` 不为 `None`），则 `ptrA = ptrA.next`。
  - 如果 `ptrA` 已到末尾（`ptrA` 为 `None`），则跳到链表 B 的头部：`ptrA = headB`。
<br>

- **`ptrB` 的移动逻辑**：
  - 同理，`ptrB` 走完链表 B 后跳到链表 A 头部。

#### 5. **结束循环并返回**
```python
return ptrA
```
- **终止条件**：
  - **找到了交点**：`ptrA` 和 `ptrB` 在交点处相等（`ptrA == ptrB`）。
  - **无交点**：最终两指针会同时走到 `None`（此时 `ptrA == ptrB = None`），返回 `None`。

### 四、**关键 Python 语法解释**

#### 1. **条件表达式**
```python
ptrA = ptrA.next if ptrA else headB
```
- 等价于：`ptrA = (ptrA.next) if (ptrA is not None) else (headB)`。
- **逻辑**：若 `ptrA` 未到链表末尾，则继续移动；否则跳转到另一链表头部。

#### 2. **类型注解（Python3.5+）**
- **`Optional[ListNode]`**:
  - 相当于 `Union[ListNode, None]`，表示返回值可能是 `ListNode` 或 `None`。
  - 需导入：`from typing import Optional`（代码中可能省略了导入语句）。

### 五、**实例验证**
假设链表 A 长度为 `a`，链表 B 长度为 `b`，公共部分长度为 `c`。

- **总步数**：
  - `ptrA` 走过的节点数：`a + (b - c)`（从 A 开始，遍历 B 的非公共部分）。
  - `ptrB` 走过的节点数：`b + (a - c)`（从 B 开始，遍历 A 的非公共部分）。
  - **a + (b - c) = b + (a - c)**，所以在交点处相遇。

### 六、**边界情况**
- **无交点**：最终两指针均为 `None`，返回 `None`。
- **完全相同的链表**：直接返回头节点。
- **一条链表为空**：返回 `None`。

### 七、**复杂度**
- 时间复杂度：$O(m + n)$（$m$、$n$ 为两链表长度）。
- 空间复杂度：$O(1)$（仅用两个指针）。


总结：这段代码通过**巧妙的双指针遍历逻辑**，在时间和空间复杂度上做到最优，且代码简洁。

## 283. 移动零

给定一个数组 `nums`，编写一个函数将所有 `0` 移动到数组的末尾，同时保持非零元素的相对顺序。

**请注意** ，必须在不复制数组的情况下原地对数组进行操作。

 

**示例 1:**

```
输入: nums = [0,1,0,3,12]
输出: [1,3,12,0,0]
```

**示例 2:**

```
输入: nums = [0]
输出: [0]
```

````python
class Solution:
    def moveZeroes(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        slow = 0  # 慢指针指向下一个应该放置非零元素的位置
        for fast in range(len(nums)):
            if nums[fast] != 0:
                nums[slow] = nums[fast]
                slow += 1
        # 将slow之后的所有元素置为0
        for i in range(slow, len(nums)):
            nums[i] = 0
````

### 整体思路分析
这个算法使用了**双指针技巧**中的**快慢指针法**，主要分为两个阶段：
1. **收集非零元素**：用快指针扫描所有元素，遇到非零元素就放到慢指针的位置，然后慢指针前移
2. **填充零元素**：当所有非零元素都移动完毕后，将剩余位置全部填零

### 代码逐行解释
```python
from typing import List  # 导入类型标注模块

class Solution:
    def moveZeroes(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        slow = 0  # 慢指针：指向下一个应该放置非零元素的位置
      
        # 第一阶段：收集所有非零元素
        for fast in range(len(nums)):  # 快指针：扫描全部元素
            if nums[fast] != 0:       # 发现非零元素
                nums[slow] = nums[fast]  # 将非零元素放到慢指针位置
                slow += 1             # 慢指针前移
      
        # 第二阶段：填充零元素
        for i in range(slow, len(nums)):
            nums[i] = 0
```

### 关键点解释
#### 1. 快慢指针工作原理
- **快指针（fast）**：像扫描仪一样遍历每个元素
- **慢指针（slow）**：记录下一个非零元素应该放置的位置

以示例 `nums = [0,1,0,3,12]` 为例：

|步骤 | fast位置 | nums[fast] | slow位置 | 数组状态|
-----|----------|-----------|-----------|-------|
|初始 |         |           | 0       | [0,1,0,3,12]|
|1   | 0       | 0（跳过）   | 0       | 无变化|
|2   | 1       | 1          | 0→1     | [1,1,0,3,12]|
|3   | 2       | 0（跳过）   | 1       | 无变化|
|4   | 3       | 3          | 1→2     | [1,3,0,3,12]|
|5   | 4       | 12         | 2→3     | [1,3,12,3,12]|



#### 2. 为什么要分两个阶段？
第一阶段结束后：
- 所有非零元素已按顺序排列在数组前端
- 但原来在这些位置的元素可能被覆盖，比如最后的3和12仍然存在

第二阶段将 `slow` 之后的位置全部置零：
```python
# 原数组变为 [1,3,12,3,12]
# slow此时是3，从索引3开始置零
for i in range(3, 5):
    nums[i] = 0
# 最终得到 [1,3,12,0,0]
```

### Python语法细节
1. **类型标注**：
   - `List[int]` 表示整数列表
   - `-> None` 表示函数没有返回值
   - 这些标注不影响程序运行，但能帮助IDE检查类型

2. **原地修改**：
   - 直接通过索引修改列表元素（如 `nums[i] = 0`）
   - 不创建新列表，符合题目要求

3. **range()函数**：
   - `range(len(nums))` 生成从0到n-1的整数序列
   - `range(slow, len(nums))` 生成从slow到n-1的序列

### 复杂度分析
- 时间复杂度：O(n) → 遍历两次数组（2n次操作）
- 空间复杂度：O(1) → 只使用固定数量的变量

### 常见疑问解答
**Q：为什么不需要考虑元素被覆盖的问题？**
A：快指针总是先于/同步于慢指针移动，当慢指针位置被覆盖时，快指针已经处理过该位置的元素了

**Q：如果全是非零元素会怎样？**
A：慢指针会等于快指针，所有元素原地不动，最后的置零循环也不会执行（因为range(slow, len(nums))的起始等于结束）

**Q：如果最后一个元素是非零怎么办？**
A：快指针会将其复制到慢指针位置，之后置零循环不会修改它，例如输入[1,2,3]会保持原样

通过这种双指针方法，我们既保持了非零元素的顺序，又高效地完成了零元素的移动，是一个非常巧妙的解法！
