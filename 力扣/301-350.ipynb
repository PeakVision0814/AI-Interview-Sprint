## [344. 反转字符串](https://leetcode.cn/problems/reverse-string/)

编写一个函数，其作用是将输入的字符串反转过来。输入字符串以字符数组 `s` 的形式给出。

不要给另外的数组分配额外的空间，你必须**[原地](https://baike.baidu.com/item/原地算法)修改输入数组**、使用 O(1) 的额外空间解决这一问题。

 

**示例 1：**

```
输入：s = ["h","e","l","l","o"]
输出：["o","l","l","e","h"]
```

**示例 2：**

```
输入：s = ["H","a","n","n","a","h"]
输出：["h","a","n","n","a","H"]
```

````python
from typing import List

class Solution:
    def reverseString(self, s: List[str]) -> None:
        """
        Do not return anything, modify s in-place instead.
        """
        i = 0
        j = len(s) - 1
        while i < j:
            temp = s[i]
            s[i] = s[j]
            s[j] = temp
            i = i + 1
            j = j - 1

solution = Solution()
user_input = input("请输入一个字符列表")
s = list(user_input)
solution.reverseString(s)
print("反转后列表：", s)
````

### 题目大意

这道题的要求是，给定一个字符列表 `s`，你需要将它 **原地** 反转。什么是“原地”呢？就是你不能创建一个新的列表来存放反转后的结果，而必须直接修改原有的列表 `s`。此外，你不需要返回任何值（题目中 `-> None` 就表示函数不返回任何东西），只需要确保调用完你的函数后，原列表 `s` 已经被反转了。

举个例子：

如果输入的 `s` 是 `["h", "e", "l", "l", "o"]`，那么函数执行完后，`s` 应该变成 `["o", "l", "l", "e", "h"]`。

### 算法思路

这道题最直观的思路就是使用 **双指针** 的方法。

想象一下，我们要反转一个字符串，其实就是把第一个字符和最后一个字符交换，第二个字符和倒数第二个字符交换，以此类推，直到我们处理完一半的字符。

我们可以设置两个指针：

1.  **左指针 `i`：** 初始化指向列表的第一个元素（索引为 0）。
2.  **右指针 `j`：** 初始化指向列表的最后一个元素（索引为 `len(s) - 1`）。

然后，我们就可以开始循环了：

*   在每次循环中，我们交换左指针 `i` 所指向的字符和右指针 `j` 所指向的字符。
*   交换完成后，我们将左指针 `i` 向右移动一位（`i = i + 1`）。
*   同时，我们将右指针 `j` 向左移动一位（`j = j - 1`）。

这个过程一直持续到左指针 `i` 追上或超过右指针 `j` 为止。当 `i >= j` 的时候，说明我们已经完成了所有需要交换的字符对，整个列表也就被反转了。

## [347. 前 K 个高频元素](https://leetcode.cn/problems/top-k-frequent-elements/)

中等  相关标签  ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业



给你一个整数数组 `nums` 和一个整数 `k` ，请你返回其中出现频率前 `k` 高的元素。你可以按 **任意顺序** 返回答案。

 

**示例 1:**

```
输入: nums = [1,1,1,2,2,3], k = 2
输出: [1,2]
```

**示例 2:**

```
输入: nums = [1], k = 1
输出: [1]
```

````python
from typing import List

class Solution:
    def topKFrequent(self, nums:List ,k: int) -> List[int]:
        freq_map = {}
        for num in nums:
            if num in freq_map:
                freq_map[num] += 1
            else:
                freq_map[num] = 1
            
        sorted_freq = sorted(freq_map.items(),key=lambda x:x[1], reverse=True)

        result = []
        for i in range(k):
            result.append(sorted_freq[i][0])

        return result

solution = Solution()
nums = input("请输入nums：")
nums = list(nums)
k = input("请输入k：")
k = int(k)
result = solution.topKFrequent(nums, k)
print(result)
````

**算法思路：**

本代码采用的是“统计频率 + 排序”的思路，具体步骤如下：

1. **统计元素频率：** 使用一个字典 `freq_map` 来存储每个数字出现的频率。遍历输入的数组 `nums`，对于数组中的每个数字，如果它已经在 `freq_map` 中，就将其对应的频率加一；如果不在，则将其添加到 `freq_map` 中，并将频率初始化为一。
2. **根据频率排序：** 将 `freq_map` 中的键值对（数字和频率）转换为一个列表，然后根据每个键值对的频率（即值）进行降序排序。这样，频率最高的元素及其频率就会排在列表的前面。
3. **取出前 K 个高频元素：** 从排序后的列表中取出前 `k` 个元素，这些元素就是频率最高的前 `k` 个数字。将这些数字收集到一个新的列表中作为最终结果返回。

    * `freq_map.items()`: 提供要排序的可迭代对象（包含 `(数字, 频率)` 元组）。
    * `key=lambda x:x[1]`: 指定排序的依据。`lambda x: x[1]` 是一个匿名函数，它接收一个参数 `x`（在这里 `x` 是一个 `(数字, 频率)` 元组），并返回 `x[1]`（即频率）。因此，`sorted` 函数会根据元组的第二个元素（频率）进行排序。

    * 根据频率排序：将字典的 items 转换为列表的时间复杂度为 O(M)，其中 M 是不同元素的个数 (M <= N)。对这个列表进行排序的时间复杂度为 O(M log M)。在最坏情况下，所有元素都不同，M = N，排序的时间复杂度为 O(N log N)。

    * **总的时间复杂度：O(N + M log M + K)。** 由于 M <= N 且 K <= N，总时间复杂度可以简化为 **O(N + N log N) = O(N log N)**。

## [349. 两个数组的交集](https://leetcode.cn/problems/intersection-of-two-arrays/)

简单  相关标签  ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业



给定两个数组 `nums1` 和 `nums2` ，返回 *它们的 交集* 。输出结果中的每个元素一定是 **唯一** 的。我们可以 **不考虑输出结果的顺序** 。

 

**示例 1：**

```
输入：nums1 = [1,2,2,1], nums2 = [2,2]
输出：[2]
```

**示例 2：**

```
输入：nums1 = [4,9,5], nums2 = [9,4,9,8,4]
输出：[9,4]
解释：[4,9] 也是可通过的
```

````python
from typing import List

class Solution:
    def intersection(self, nums1: List[int], nums2: List[int]) -> List[int]:
        nums1_present = {}
        for number in nums1:
            nums1_present[number] = True 

        intersection_list = []

        added_to_result = set()

        for number in nums2:
            if number in nums1_present:
                if number not in added_to_result:
                    intersection_list.append(number)
                    added_to_result.add(number) 

        return intersection_list
````

```python
from typing import List

class Solution:
    def intersection(self, nums1: List[int], nums2: List[int]) -> List[int]:
        """
        找到两个数组的交集 (使用哈希表，更容易理解的去重方式)。

        Args:
            nums1: 第一个整数数组。
            nums2: 第二个整数数组。

        Returns:
            一个列表，包含两个数组的交集，每个元素唯一。
        """
        # 第一步：把 nums1 里面的元素都记录下来，方便快速查找
        nums1_present = {} # 创建一个空字典，用来记录 nums1 里面的数字
        for number in nums1:
            nums1_present[number] = True # 把 nums1 里面的每个数字作为一个“标记”存进字典里

        # 第二步：开始找交集
        intersection_list = [] # 创建一个空列表，用来存放找到的交集元素

        # 为了确保交集元素是唯一的，我们需要记住哪些元素已经添加到结果列表了
        added_to_result = set() # 创建一个空集合，用来记住已经添加到结果的元素

        # 遍历 nums2 里面的每一个数字
        for number in nums2:
            # 检查这个数字是否在 nums1 里出现过 (也就是在 nums1_present 字典里)
            if number in nums1_present:
                # 如果出现了，那它就是交集的一部分
                # 但是我们需要确保这个数字还没有被添加到 intersection_list 里
                if number not in added_to_result:
                    # 如果还没有添加，就把它添加到结果列表和已添加集合中
                    intersection_list.append(number)
                    added_to_result.add(number) # 记住这个数字已经添加过了

        # 第三步：返回结果列表
        return intersection_list
```

**算法解释 (解决问题的思路)**

想象一下，你有两堆玩具（`nums1` 和 `nums2`），你想找出这两堆玩具里都有哪些 **种类** 的玩具。而且，每种玩具你只需要记下一次，不用重复记。

这段代码的算法思路就是：

1.  **“清点”第一堆玩具 (nums1):** 我们需要一种快速的方法来知道第一堆玩具里有哪些种类的玩具。最笨的方法是拿个本子把所有玩具的名字都写下来。但这样太慢了，每次想知道某个玩具在不在，都要从头翻到尾。
    *   更好的方法是做一个“清单”或者“索引”，就像图书馆的目录一样。我们创建一个叫做 `nums1_present` 的东西，它就像一个字典。字典的“词条”就是玩具的名字（数字），旁边随便写个标记（比如 `True`），表示这个玩具在第一堆里。这样，以后想知道某个玩具在不在第一堆，只需要查字典就行，非常快！

2.  **检查第二堆玩具 (nums2) 并找出共同的种类:** 现在我们有了第一堆玩具的清单 (`nums1_present`)。我们开始检查第二堆玩具 (`nums2`)。
    *   我们从第二堆里拿出第一个玩具，查一下它在不在第一堆的清单里 (`nums1_present`)。
    *   如果它在清单里，说明它是两堆都有的共同玩具种类。
    *   我们把这个共同玩具种类记下来，放到一个叫做 `intersection_list` 的新列表里。这个列表就是我们最终要找的“交集”。
    *   **重点：** 但是，题目要求我们只记录 **唯一** 的玩具种类。所以，每当我们找到一个共同的玩具种类并准备把它加到 `intersection_list` 时，我们还要检查一下它是不是已经加过了。
    *   为了方便检查，我们再维护一个叫做 `added_to_result` 的东西，它就像一个“已经记录过的共同种类”的清单。每当我们把一个玩具种类加到 `intersection_list` 里，同时也在 `added_to_result` 里记一下。下次遇到同一个玩具种类时，如果发现它已经在 `added_to_result` 里了，就说明它之前已经加过了，这次就不用再加了。我们用一个“集合”（`set`）来做这个清单，因为集合天生就不会有重复的元素。

3.  **完成并给出结果:** 遍历完第二堆所有的玩具后，`intersection_list` 里就存放了所有两堆都有的、而且每种只记录了一次的玩具种类，这就是我们要找的交集。我们把这个列表返回出去。
