## [367. 有效的完全平方数](https://leetcode.cn/problems/valid-perfect-square/)

简单  相关标签   ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业



给你一个正整数 `num` 。如果 `num` 是一个完全平方数，则返回 `true` ，否则返回 `false` 。

**完全平方数** 是一个可以写成某个整数的平方的整数。换句话说，它可以写成某个整数和自身的乘积。

不能使用任何内置的库函数，如 `sqrt` 。



**示例 1：**

```
输入：num = 16
输出：true
解释：返回 true ，因为 4 * 4 = 16 且 4 是一个整数。
```

**示例 2：**

```
输入：num = 14
输出：false
解释：返回 false ，因为 3.742 * 3.742 = 14 但 3.742 不是一个整数。
```


````python
class Solution:
    def isPerfectSquare(self, num: int) -> bool:
        if num < 0:
            return False
        if num == 0 or num == 1:
            return True
        
        left, right = 1, num
        while left <= right:
            mid = (left + right) // 2
            square = mid * mid
            
            if square == num:
                return True
            elif square < num:
                left = mid + 1
            else:
                right = mid - 1
        
        return False

# Example usage:
solution = Solution()

try:
    input_str = input("请输入一个非负整数: ")

    x = int(input_str)

    if x < 0:
        raise ValueError("输入必须是非负整数。")
    else:
        result = solution.isPerfectSquare(x)
        # 在 print 中使用 f-string
        print(f"{result}")
    
except ValueError as e:
    print(f"错误: {e}")
except Exception as e:
    print(f"发生了一个意外错误: {e}")
````

**3. 算法解析 (二分查找)**

这段代码的核心算法是**二分查找 (Binary Search)**。二分查找是一种高效的搜索算法，用于在已排序的数组或列表中查找特定元素。虽然这里我们不是在一个数组中查找，但我们可以将问题转化为：在从 1 到 `num` 的整数范围内，是否存在一个整数 `x` 使得 `x * x == num`。

二分查找的基本思想是：

1.  **确定搜索范围：** 我们知道，如果 `num` 是一个完全平方数，它的平方根一定在 1 到 `num` 之间（实际上，如果 `num > 1`，平方根会在 1 到 `num // 2 + 1` 之间，但 1 到 `num` 也包含了所有可能的整数平方根）。我们将搜索范围的左边界设为 `left = 1`，右边界设为 `right = num`。
2.  **计算中间值：** 在当前的搜索范围内，计算中间位置的整数 `mid = (left + right) // 2`。
3.  **比较中间值的平方与目标值：**
    *   如果 `mid * mid == num`，说明 `mid` 就是 `num` 的平方根，`num` 是完全平方数，搜索成功，返回 `True`。
    *   如果 `mid * mid < num`，说明真正的平方根比 `mid` 大，我们需要在 `mid` 的右侧继续搜索。所以，将搜索范围的左边界更新为 `left = mid + 1`。
    *   如果 `mid * mid > num`，说明真正的平方根比 `mid` 小，我们需要在 `mid` 的左侧继续搜索。所以，将搜索范围的右边界更新为 `right = mid - 1`。
4.  **重复：** 重复步骤 2 和 3，直到找到目标值或者搜索范围为空 (`left > right`)。
5.  **结果：** 如果搜索范围为空仍然没有找到，说明 `num` 不是完全平方数，返回 `False`。

**为什么二分查找高效？**

每次比较后，二分查找都会将搜索范围缩小一半。这种效率非常高。对于一个大小为 N 的搜索范围，二分查找大约只需要 `log2(N)` 次比较就能找到结果或确定不存在。这比线性搜索（逐个检查）要快得多。

**在这个代码中的应用：**

*   `left = 1, right = num`: 初始化搜索范围。
*   `while left <= right:`: 循环继续，只要搜索范围还有效。
*   `mid = (left + right) // 2`: 找到中间值。
*   `mid_squared = mid * mid`: 计算中间值的平方。
*   根据 `mid_squared` 与 `num` 的比较，更新 `left` 或 `right`，缩小搜索范围。
*   如果找到 `mid_squared == num`，返回 `True`。
*   如果循环结束仍未找到，返回 `False`。

**举个例子：判断 `num = 16` 是否是完全平方数**

1.  `left = 1`, `right = 16`
2.  `mid = (1 + 16) // 2 = 8`. `mid_squared = 8 * 8 = 64`. `64 > 16`. `right = 8 - 1 = 7`.
3.  `left = 1`, `right = 7`
4.  `mid = (1 + 7) // 2 = 4`. `mid_squared = 4 * 4 = 16`. `16 == 16`. 返回 `True`.

**举个例子：判断 `num = 14` 是否是完全平方数**

1.  `left = 1`, `right = 14`
2.  `mid = (1 + 14) // 2 = 7`. `mid_squared = 7 * 7 = 49`. `49 > 14`. `right = 7 - 1 = 6`.
3.  `left = 1`, `right = 6`
4.  `mid = (1 + 6) // 2 = 3`. `mid_squared = 3 * 3 = 9`. `9 < 14`. `left = 3 + 1 = 4`.
5.  `left = 4`, `right = 6`
6.  `mid = (4 + 6) // 2 = 5`. `mid_squared = 5 * 5 = 25`. `25 > 14`. `right = 5 - 1 = 4`.
7.  `left = 4`, `right = 4`
8.  `mid = (4 + 4) // 2 = 4`. `mid_squared = 4 * 4 = 16`. `16 > 14`. `right = 4 - 1 = 3`.
9.  `left = 4`, `right = 3`. 现在 `left > right`。循环结束。返回 `False`.

**总结**

这段代码利用了二分查找算法的原理，在可能的平方根范围内高效地搜索。通过不断地计算中间值的平方并与目标数进行比较，逐步缩小搜索范围，直到找到平方根或者确定不存在。结合 Python 的基本语法（条件判断、循环、变量赋值），实现了判断一个正整数是否为完全平方数的功能。

## [383. 赎金信](https://leetcode.cn/problems/ransom-note/)

简单   相关标签   ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业



给你两个字符串：`ransomNote` 和 `magazine` ，判断 `ransomNote` 能不能由 `magazine` 里面的字符构成。

如果可以，返回 `true` ；否则返回 `false` 。

`magazine` 中的每个字符只能在 `ransomNote` 中使用一次。

 

**示例 1：**

```
输入：ransomNote = "a", magazine = "b"
输出：false
```

**示例 2：**

```
输入：ransomNote = "aa", magazine = "ab"
输出：false
```

**示例 3：**

```
输入：ransomNote = "aa", magazine = "aab"
输出：true
```

````python
class Solution:
    def canConstruct(self, ransomNote: str, magazine: str) -> bool:
        counts = {}
        for i in magazine:
            counts[i] = counts.get(i, 0) + 1
        for i in ransomNote:
            if i not in counts or counts[i] == 0:
                return False
            counts[i] -=1
        return True
````

        counts = {}

1.  首先，我们需要知道 `magazine` 里有哪些字母，以及每个字母有多少个。就像清点你的字母库存一样。
2.  然后，我们遍历 `ransomNote` 中的每个字母。
3.  对于 `ransomNote` 中的每个字母，我们去检查你的 "字母库存" (`counts` 字典) 是否有这个字母，并且数量是否够用（数量是否大于 0）。
4.  如果在检查过程中，发现 `ransomNote` 需要的某个字母在 `magazine` 的库存里没有，或者数量已经用完了，那么我们就知道无法构成 `ransomNote` 了，直接返回 `False`。
5.  如果 `ransomNote` 中的所有字母都在 `magazine` 的库存里找到了足够的数量，那么恭喜你，可以构成，返回 `True`。

2.  按顺序检查 `ransomNote` 需要的每个字母。
3.  每次检查，看看库存是否还有这个字母，并且数量够不够。
4.  如果够，就从库存里拿走一个，继续检查下一个字母。
5.  如果不够，或者库存里根本没有这个字母，那就宣告失败，立即结束。
6.  如果 `ransomNote` 所有的字母都检查完了，并且都能在库存里找到并拿走，那就宣告成功。
