## [202. 快乐数](https://leetcode.cn/problems/happy-number/)

简单  相关标签 ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业



编写一个算法来判断一个数 `n` 是不是快乐数。

**「快乐数」** 定义为：

- 对于一个正整数，每一次将该数替换为它每个位置上的数字的平方和。
- 然后重复这个过程直到这个数变为 1，也可能是 **无限循环** 但始终变不到 1。
- 如果这个过程 **结果为** 1，那么这个数就是快乐数。

如果 `n` 是 *快乐数* 就返回 `true` ；不是，则返回 `false` 。

 

**示例 1：**

```
输入：n = 19
输出：true
解释：
12 + 92 = 82
82 + 22 = 68
62 + 82 = 100
12 + 02 + 02 = 1
```

**示例 2：**

```
输入：n = 2
输出：false
```

````python
class Solution:
    def isHappy(self, n: int) -> bool:
        seen = set()
        while n != 1 and n not in seen:
            seen.add(n)
            sum_of_squares = 0
            for digit in str(n):
                sum_of_squares += int(digit) ** 2
            n = sum_of_squares

        return n == 1
````

**题目回顾：什么是快乐数？**

想象一下，你有一个正整数。你做以下操作：

1.  把你当前的数字的每一位数字拿出来。
2.  把每一位数字平方（自己乘以自己）。
3.  把所有平方后的结果加起来，得到一个新的数字。
4.  用这个新的数字重复上面的步骤。

如果你最终得到了数字 1，那么你最初的数字就是一个“快乐数”。如果你在这个过程中，数字进入了一个循环，并且永远变不到 1，那么它就不是快乐数。

**代码的思路：如何判断是否会进入循环？**

代码的核心思路是：如果一个数不是快乐数，它在重复计算平方和的过程中，**最终一定会进入一个循环**。如果它进入了循环，并且循环中不包含 1，那么它就永远不会变成 1。

所以，我们的任务就是检测是否出现了循环。

**代码逐行解释：**

```python
class Solution:
    def isHappy(self, n: int) -> bool:
        """
        判断一个数是否是快乐数。

        Args:
            n: 输入的整数。

        Returns:
            如果 n 是快乐数，返回 True；否则返回 False。
        """
        # 这一行定义了一个名为 isHappy 的函数，它是 Solution 类的一部分。
        # 它接收一个整数 n 作为输入，并期望返回一个布尔值（True 或 False）。

        seen = set()  # 用于记录出现过的数字，检测循环
        # 这一行创建了一个叫做 'seen' 的集合（set）。
        # 集合是一种无序的数据结构，它的主要特点是里面的元素都是唯一的，而且查找元素非常快。
        # 我们用这个集合来记住我们在计算过程中已经遇到的数字。
        # 如果我们再次遇到了一个已经在 'seen' 集合里的数字，就说明我们进入了一个循环。

        while n != 1 and n not in seen:
        # 这是一个 'while' 循环。它会一直重复执行里面的代码，直到后面的条件不再满足。
        # 条件有两个部分，用 'and' 连接，意味着两个条件都必须是真的，循环才会继续：
        # 1. `n != 1`: 当前的数字 `n` 不是 1。如果 `n` 已经是 1 了，说明我们找到了快乐数，循环应该停止。
        # 2. `n not in seen`: 当前的数字 `n` 还没有出现在我们之前记录过的数字集合 `seen` 里。
        #    如果 `n` 已经在 `seen` 里了，说明我们进入了循环，并且这个循环不包含 1（因为第一个条件排除了 n=1），循环应该停止。

            seen.add(n)
            # 如果上面的 while 循环条件满足（n 不是 1，且 n 还没见过），我们就把当前的数字 `n` 添加到 `seen` 集合里。
            # 这样，下次如果再次遇到这个数字，我们就知道进入循环了。

            sum_of_squares = 0
            # 这一行初始化一个变量 `sum_of_squares` 为 0。
            # 我们会用这个变量来存储当前数字各个位数的平方和。

            for digit in str(n):
            # 这是一个 'for' 循环。它会遍历某个序列中的每一个元素。
            # `str(n)` 将当前的数字 `n` 转换成一个字符串。
            # 例如，如果 n 是 19，str(n) 就是 "19"。
            # 这个 for 循环会把字符串中的每一个字符（也就是数字的每一位）取出来，分别赋值给变量 `digit`。
            # 在上面的例子中，第一次循环 `digit` 是 "1"，第二次循环 `digit` 是 "9"。

                sum_of_squares += int(digit) ** 2
                # 在 for 循环内部，对每一个数字位进行操作：
                # `int(digit)` 将当前的字符 `digit` 转换回整数。例如，将 "1" 转换成 1，将 "9" 转换成 9。
                # `** 2` 表示平方，也就是自己乘以自己。例如，1**2 = 1，9**2 = 81。
                # `sum_of_squares += ...` 将平方后的结果加到 `sum_of_squares` 变量里。
                # 这行代码计算了当前数字的每一位的平方，并将它们累加到 `sum_of_squares` 中。

            n = sum_of_squares
            # for 循环结束后，`sum_of_squares` 存储了当前数字所有位数的平方和。
            # 这一行把 `n` 更新为这个新的平方和。
            # 然后，while 循环会再次检查条件，使用这个新的 `n` 值。

        return n == 1
        # 当 while 循环结束时，说明循环条件不再满足。这有两种可能：
        # 1. `n` 变成了 1。这意味着我们找到了快乐数。
        # 2. `n` 已经出现在 `seen` 集合里了（因为 `n not in seen` 条件为 False）。这意味着我们进入了一个不包含 1 的循环。
        # 这一行返回 `n == 1` 的结果。
        # 如果 `n` 最终是 1，`n == 1` 就是 True，函数返回 True。
        # 如果 `n` 最终不是 1（而是进入了循环），`n == 1` 就是 False，函数返回 False。

```

**总结算法流程：**

1.  创建一个空的集合 `seen` 来记住我们已经见过的数字。
2.  开始一个循环，只要当前的数字 `n` 不等于 1，并且 `n` 还没有在 `seen` 集合里出现过，就继续循环。
3.  在循环内部：
    *   把当前的 `n` 加入到 `seen` 集合里。
    *   计算 `n` 的每一位数字的平方和，把结果存储在 `sum_of_squares` 中。
    *   把 `n` 更新为 `sum_of_squares`。
4.  当循环结束时，检查 `n` 是否等于 1。如果等于 1，返回 `True`（快乐数）。如果不等于 1，返回 `False`（非快乐数，进入了循环）。

**为什么这个算法有效？**

这个算法的有效性基于一个事实：对于任何非快乐数，在重复计算平方和的过程中，数字会越来越大，但不会无限增大。最终它们一定会进入一个循环。通过记录已经见过的数字，我们可以有效地检测到这个循环，从而判断这个数是不是快乐数。如果数字最终达到了 1，那它就是快乐数，并且不会进入循环。

## [203. 移除链表元素](https://leetcode.cn/problems/remove-linked-list-elements/)

简单   相关标签   ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业



给你一个链表的头节点 `head` 和一个整数 `val` ，请你删除链表中所有满足 `Node.val == val` 的节点，并返回 **新的头节点** 。

 

**示例 1：**

![img](https://assets.leetcode.com/uploads/2021/03/06/removelinked-list.jpg)

```
输入：head = [1,2,6,3,4,5,6], val = 6
输出：[1,2,3,4,5]
```

**示例 2：**

```
输入：head = [], val = 1
输出：[]
```

**示例 3：**

```
输入：head = [7,7,7,7], val = 7
输出：[]
```

````python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def removeElements(self, head: Optional[ListNode], val: int) -> Optional[ListNode]:
        dummy = ListNode(0)
        dummy.next = head

        current = dummy
        while current.next:
            if current.next.val == val:
                current.next = current.next.next
            else:
                current = current.next
        return dummy.next
````

**代码的结构：**

这段代码是Python语言写的，放在一个叫做 `Solution` 的“类”里面，有一个叫做 `removeElements` 的“方法”来完成我们的任务。

* `head: Optional[ListNode]`: 这个 `head` 就是我们刚开始拿到的那串珠子的第一颗珠子。它可能是一个珠子（一个 `ListNode` 对象），也可能一个都没有（`Optional` 表示可能是 `None`）。
* `val: int`: 这个 `val` 就是我们要拿掉的珠子上的那个特定数字。
* `-> Optional[ListNode]`: 这个表示这个方法最后会返回什么。它会返回新串好的珠子的第一颗珠子，也可能一个都没有（如果所有珠子都被拿掉了）。

**算法思路（怎么做）：**

为了更容易地拿掉珠子，特别是当第一颗珠子就是我们要拿掉的数字时，我们采用了一个小技巧：

1. **“假”的第一颗珠子 (虚拟头节点):** 我们先自己额外做一个“假”的珠子，把它放在我们原来那串珠子的最前面。这个“假”珠子上不放任何数字。它的作用就像一个“占位符”，让我们更容易处理原来链表头部的珠子。
2. **从“假”珠子开始检查:** 我们从这个“假”珠子开始，一个一个地往后看。
3. **看下一个珠子:** 每一步，我们都看当前珠子的**下一个**珠子。
4. **要拿掉吗？** 如果下一个珠子上的数字是我们指定的 `val`，我们就把它从链子里“跳过去”。具体做法是，把当前珠子的 `next` (指向下一个珠子的链接) 直接指向下一个珠子的**再下一个**珠子。这样，下一个珠子就“脱离”了链子。
5. **不用拿掉？** 如果下一个珠子上的数字不是我们要拿掉的 `val`，那它就是有用的珠子，我们就继续往前走一步，把“当前”珠子变成刚才看过的那个“下一个”珠子。
6. **一直走到最后:** 我们重复上面的步骤，直到走到链子的尽头，没有下一个珠子可以看了。
7. **返回新的第一颗珠子:** 最后，我们把那个“假”珠子去掉，它后面的那个珠子就是我们新串好的链子的第一颗珠子，我们就把这个珠子返回。

**语法解释（代码怎么写）：**

现在我们把上面的步骤对应到代码上：

```python
class Solution:
    def removeElements(self, head: Optional[ListNode], val: int) -> Optional[ListNode]:
        # 步骤 1: 创建“假”的第一颗珠子
        # ListNode() 创建一个珠子对象
        # next = head 表示这个新珠子的下一个就是原来的链表的头
        head = ListNode(next = head)
```
* `ListNode(next = head)`: 这里我们创建了一个新的 `ListNode` 对象。`next = head` 意思是这个新珠子的下一个指向原来链表的头。
* `head = ListNode(...)`: 我们把原来叫做 `head` 的变量，现在指向了这个新创建的“假”珠子。这样，我们的操作就都在这个“假”珠子开头的链上了。

```python
        # 步骤 2: 从“假”珠子开始检查
        i = head
```
* `i = head`: 我们创建一个新的变量 `i` (你可以想象它是一个手指，用来指向当前正在检查的珠子)。一开始，我们的手指指向的是那个“假”珠子。

```python
        # 步骤 6: 一直走到最后
        while i.next:
```
* `while i.next:`: 这是一个循环。它的意思是：“只要 `i` 指向的珠子的**下一个**不是空的（不是 `None`），就一直重复下面的事情。” 这确保我们能遍历到链表的倒数第二个节点，因为我们需要检查它的下一个节点。

```python
            # 步骤 3 & 4: 看下一个珠子，要拿掉吗？
            if i.next.val == val:
                i.next = i.next.next
```
* `if i.next.val == val:`: 如果 `i` 指向的珠子的**下一个**珠子的 `val` (上面的数字) 等于我们要删除的 `val`...
* `i.next = i.next.next`: ...那么，把 `i` 指向的珠子的 `next` (原来指向要删除的那个珠子) 直接改成指向要删除珠子的**下一个**珠子。这样，要删除的那个珠子就被跳过了。

```python
            # 步骤 5: 不用拿掉？
            else:
                i = i.next
```
* `else:`: 如果下一个珠子上的数字**不**等于 `val`...
* `i = i.next`: ...那么，我们就把我们的手指 `i` 移到下一个珠子上，继续往后检查。

```python
        # 步骤 7: 返回新的第一颗珠子
        return head.next
```
* `return head.next`: 循环结束了，我们已经处理完了所有的珠子。还记得我们最开始创建的那个“假”珠子吗？它现在还是在 `head` 指向的位置。它的 `next` 现在指向的就是经过删除操作后，真正链表的第一个珠子。所以，我们返回 `head.next`。

**为什么这样写比直接从原始 `head` 开始好？**

试想一下，如果原始链表的头节点 (`head`) 就是我们要删除的数字。如果我们直接从 `head` 开始，当我们删除 `head` 时，需要更新 `head` 指向新的节点。这会有点麻烦，需要额外的判断。

但是有了这个“假”珠子 (`head = ListNode(next = head)`)，我们要删除的头节点就变成了“假”珠子的**下一个**节点。我们只需要检查 `i.next` (也就是原来的 `head`)，如果它的值是 `val`，我们就执行 `i.next = i.next.next`。因为 `i` 指向的是“假”珠子，`i.next` 就指向了原来的头节点。通过 `i.next = i.next.next`，我们实际上是更新了“假”珠子的 `next` 指针，让它跳过原来的头节点，指向了新的头节点。这样，处理删除头节点的情况就和其他删除节点的情况一样了，代码更简洁。

最后，返回 `head.next` 自然就拿到了新的链表的头节点，因为 `head` 指向的是那个永远不会被删除的“假”珠子。

## [206. 反转链表](https://leetcode.cn/problems/reverse-linked-list/)

简单  相关标签  ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业

给你单链表的头节点 `head` ，请你反转链表，并返回反转后的链表。

 

**示例 1：**

![img](https://assets.leetcode.com/uploads/2021/02/19/rev1ex1.jpg)

```
输入：head = [1,2,3,4,5]
输出：[5,4,3,2,1]
```

**示例 2：**

![img](https://assets.leetcode.com/uploads/2021/02/19/rev1ex2.jpg)

```
输入：head = [1,2]
输出：[2,1]
```

**示例 3：**

```
输入：head = []
输出：[]
```

````python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        pre = None
        cur = head

        while cur:
            temp = cur.next
            cur.next = pre
            pre = cur
            cur = temp
        return pre
````

想象一下，你手里有一串珠子，每个珠子上都写了一个数字，而且每个珠子都用一根线连着它后面的珠子。这就是一个“链表”。

* **`head`**: 这就是这串珠子的第一个珠子。

现在，你的任务是把这串珠子“反过来”，也就是原来排在第一个的珠子现在排在最后一个，原来排在最后一个的珠子现在排在第一个。

咱们的这段代码，就像是一个帮你完成这个任务的“魔法师”，它一步一步地把珠子反过来。

**魔法师的工具和步骤：**

魔法师有几个工具（叫做“变量”），可以记住一些东西：

1. **`pre`**: 这是一个记事本，用来记住“我刚刚处理过的珠子是哪个”。最开始的时候，魔法师还没处理任何珠子，所以这个记事本里是空的，用 `None` 表示“空”。

2. **`cur`**: 这是魔法师现在正在盯着看的“当前”珠子。最开始的时候，魔法师从第一个珠子开始看，所以 `cur` 就指向第一个珠子 (`head`)。

现在，魔法师开始他的“反转魔法”（也就是 `while cur:` 循环）：

**循环的意思是：“只要魔法师还有正在看的珠子 (`cur` 不是空的)，就一直这样做下去！”**

在每一次循环里，魔法师会做以下几件事：

3. **`temp = cur.next`**: 魔法师在准备处理当前这个珠子之前，先看一眼它后面的珠子是哪个，并用另一个记事本 `temp` 暂时记下来。**这是为了等下把当前珠子反过来后，还能找到原来的下一个珠子继续处理。** 就像你在翻书之前，先记住下一页是哪页一样。

4. **`cur.next = pre`**: 这是最神奇的一步！魔法师把当前这个珠子 (`cur`) 上原来指向后面珠子的那根线，断开！然后把它重新连到他刚刚处理过的那个珠子 (`pre`) 上。**这就把当前珠子“反过来”指向前面了！**

5. **`pre = cur`**: 现在，当前这个珠子 (`cur`) 已经处理完了，并且它的线已经指向前面了。在下一轮处理下一个珠子时，这个处理好的珠子 (`cur`) 就会变成下一个珠子前面那个“刚刚处理过的珠子”。所以，魔法师把 `pre` 记事本里的内容更新，记下当前这个珠子 (`cur`)。

6. **`cur = temp`**: 魔法师处理完当前这个珠子后，他要去看下一个珠子了。他从之前那个 `temp` 记事本里找到刚刚记下来的那个“原来的下一个珠子”，然后把 `cur` 指向那个珠子。

**循环就这样一步一步进行：**

* 最开始：
    * `pre` 是空的
    * `cur` 指向第一个珠子

* 第一次循环：
    * `temp` 记住第二个珠子。
    * 第一个珠子原来指向第二个珠子的线，现在指向空的 (`pre`)。
    * `pre` 记住第一个珠子。
    * `cur` 去看第二个珠子。

* 第二次循环：
    * `temp` 记住第三个珠子。
    * 第二个珠子原来指向第三个珠子的线，现在指向第一个珠子 (`pre`)。
    * `pre` 记住第二个珠子。
    * `cur` 去看第三个珠子。

... 这样一直做下去，直到 `cur` 看完了所有的珠子，变成空的 (`None`)。

**循环结束了！**

7. **`return pre`**: 当魔法师把所有的珠子都处理完后，`cur` 变成空的了。这个时候，那个 `pre` 记事本里记住的是哪个珠子呢？它记住的是最后处理的那个珠子，也就是原来链表的最后一个珠子。现在，这个最后一个珠子在反转后的链表中，变成了第一个珠子！所以，魔法师把这个珠子 (`pre`) 作为反转后链表的开头，交给你。

## [209. 长度最小的子数组](https://leetcode.cn/problems/minimum-size-subarray-sum/)

中等  相关标签  ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业

给定一个含有 `n` 个正整数的数组和一个正整数 `target` **。**

找出该数组中满足其总和大于等于 `target` 的长度最小的 **子数组** `[numsl, numsl+1, ..., numsr-1, numsr]` ，并返回其长度**。**如果不存在符合条件的子数组，返回 `0` 。

 

**示例 1：**

```
输入：target = 7, nums = [2,3,1,2,4,3]
输出：2
解释：子数组 [4,3] 是该条件下的长度最小的子数组。
```

**示例 2：**

```
输入：target = 4, nums = [1,4,4]
输出：1
```

**示例 3：**

```
输入：target = 11, nums = [1,1,1,1,1,1,1,1]
输出：0
```

````python
from typing import List
class Solution:
    def minSubArrayLen(self, target: int, nums: List[int]) -> int:
        l = len(nums)
        left = 0
        right = 0
        min_len = float('inf')
        cur_sum = 0 #当前的累加值
        
        while right < l:
            cur_sum += nums[right]
            
            while cur_sum >= target: # 当前累加值大于目标值
                min_len = min(min_len, right - left + 1)
                cur_sum -= nums[left]
                left += 1
            
            right += 1
        
        return min_len if min_len != float('inf') else 0
````

想象一下你手里有一长串数字，比如 `[2, 3, 1, 2, 4, 3]`，你的目标是找到连续的一小段数字（我们称为“子数组”），让这段数字加起来的和大于等于某个特定的目标值，比如 `7`。而且，你希望找到所有符合条件的子数组中，最短的那一个。

这段代码就是用来解决这个问题的。

**从算法的角度来看**

这个算法的核心思想叫做“滑动窗口”。你可以想象有两个指针，一个叫做 `left` (左指针)，一个叫做 `right` (右指针)。这两个指针圈定了一个“窗口”，也就是当前我们关注的那段连续的数字。

1. **窗口扩大：** 最开始，窗口非常小，可能只包含第一个数字。我们用 `right` 指针不断向右移动，把更多的数字拉进窗口里。同时，我们用一个变量 `cur_sum` 记录当前窗口里所有数字的和。

2. **窗口缩小（当满足条件时）：** 每当 `cur_sum` 达到或超过我们的目标值 `target` 时，说明我们找到了一个符合条件的子数组。此时，我们就可以尝试让窗口变小一点，看看是否还能满足条件。我们通过移动 `left` 指针向右来缩小窗口，同时从 `cur_sum` 中减去 `left` 指针指向的那个数字。这样做是为了寻找更短的满足条件的子数组。

3. **记录最短长度：** 每当 `cur_sum >= target` 成立时，我们就比较当前窗口的长度 (`right - left + 1`) 和之前找到的最短长度 `min_len`，取两者中较小的值更新 `min_len`。这样，`min_len` 就始终记录着目前为止找到的最短符合条件的子数组的长度。

4. **重复以上步骤：** 我们一直重复扩大和缩小窗口的过程，直到 `right` 指针到达数字串的末尾。

**从 Python 语法的角度来看**

* `class Solution:` 这表示我们定义了一个类叫做 `Solution`。在编程中，类就像一个模板，用来创建对象。虽然在这个特定的问题中我们不需要创建 `Solution` 的对象来解决问题，但按照 LeetCode 等在线编程平台的习惯，通常会将解决问题的代码放在一个类里面。

* `def minSubArrayLen(self, target: int, nums: List[int]) -> int:` 这定义了一个叫做 `minSubArrayLen` 的函数，它是 `Solution` 类的一个方法（所以前面有 `self`）。
    * `self`: 代表类本身的实例，也就是通过这个类创建出来的对象。在这个问题中，我们不太会用到 `self` 的实际作用。
    * `target: int`: 表示这个函数接受一个参数叫做 `target`，并且我们期望它是一个整数 (`int`)。
    * `nums: List[int]`: 表示这个函数接受另一个参数叫做 `nums`，并且我们期望它是一个整数列表 (`List[int]`)。
    * `-> int`: 表示这个函数最后会返回一个整数。

* `l = len(nums)`: 这一行获取 `nums` 列表的长度，并将其赋值给变量 `l`。

* `left = 0` 和 `right = 0`: 初始化左指针 `left` 和右指针 `right`，都指向列表的第一个元素（索引为 0）。

* `min_len = float('inf')`: 初始化 `min_len` 为一个非常大的值 (`float('inf')`)。这样做是为了确保我们找到的第一个符合条件的子数组的长度一定会比 `min_len` 小，从而正确地更新 `min_len`。如果一开始设置为 0，那么即使找到一个符合条件的子数组，它的长度也可能大于 0，导致 `min_len` 保持 0 不变，这是不正确的。

* `cur_sum = 0`: 初始化 `cur_sum` 为 0，用来存储当前窗口内数字的和。

* `while right < l:`: 这是一个 `while` 循环。只要 `right` 指针还没有到达列表的末尾 (`l`)，循环就会一直执行。

* `cur_sum += nums[right]`: 将 `right` 指针当前指向的数字加到 `cur_sum` 中。这是扩大窗口的操作。

* `while cur_sum >= target:`: 这是一个嵌套的 `while` 循环。只要当前窗口的和 `cur_sum` 大于等于目标值 `target`，这个循环就会一直执行。这是窗口缩小的过程。
    * `min_len = min(min_len, right - left + 1)`: 计算当前窗口的长度 (`right - left + 1`)，并与当前的 `min_len` 进行比较，将较小的值赋值给 `min_len`。
    * `cur_sum -= nums[left]`: 从 `cur_sum` 中减去 `left` 指针指向的数字。这是因为我们要将 `left` 指针向右移动，这个数字就不再属于当前窗口了。
    * `left += 1`: 将 `left` 指针向右移动一位。

* `right += 1`: 在内层 `while` 循环结束后，将 `right` 指针向右移动一位，继续扩大窗口。

* `return min_len if min_len != float('inf') else 0`: 这是函数的返回值。
    * `min_len if min_len != float('inf')`: 如果 `min_len` 仍然是最初的那个非常大的值 (`float('inf')`)，说明在整个过程中都没有找到任何一个符合条件的子数组。
    * `else 0`: 如果 `min_len` 还是那个非常大的值，就返回 0，表示没有找到符合条件的子数组。
    * 否则，就返回 `min_len`，它就是找到的最短符合条件的子数组的长度。

请你仅使用两个队列实现一个后入先出（LIFO）的栈，并支持普通栈的全部四种操作（`push`、`top`、`pop` 和 `empty`）。

实现 `MyStack` 类：

- `void push(int x)` 将元素 x 压入栈顶。
- `int pop()` 移除并返回栈顶元素。
- `int top()` 返回栈顶元素。
- `boolean empty()` 如果栈是空的，返回 `true` ；否则，返回 `false` 。

 

**注意：**

- 你只能使用队列的标准操作 —— 也就是 `push to back`、`peek/pop from front`、`size` 和 `is empty` 这些操作。
- 你所使用的语言也许不支持队列。 你可以使用 list （列表）或者 deque（双端队列）来模拟一个队列 , 只要是标准的队列操作即可。

 

**示例：**

```
输入：
["MyStack", "push", "push", "top", "pop", "empty"]
[[], [1], [2], [], [], []]
输出：

myStack.pop(); // 返回 2

```

````python
from collections import deque

class MyStack:

    def __init__(self):
        """
        Python普通的Queue或SimpleQueue没有类似于peek的功能
        也无法用索引访问，在实现top的时候较为困难。

        用list可以，但是在使用pop(0)的时候时间复杂度为O(n)
        因此这里使用双向队列，我们保证只执行popleft()和append()，因为deque可以用索引访问，可以实现和peek相似的功能

        in - 存所有数据
        out - 仅在pop的时候会用到
        """
        self.queue_in = deque()
        self.queue_out = deque()

    def push(self, x: int) -> None:
        """
        直接append即可
        """
        self.queue_in.append(x)


    def pop(self) -> int:
        """
        1. 首先确认不空
        2. 因为队列的特殊性，FIFO，所以我们只有在pop()的时候才会使用queue_out
        3. 先把queue_in中的所有元素（除了最后一个），依次出列放进queue_out
        4. 交换in和out，此时out里只有一个元素
        5. 把out中的pop出来，即是原队列的最后一个
        
        tip：这不能像栈实现队列一样，因为另一个queue也是FIFO，如果执行pop()它不能像
        stack一样从另一个pop()，所以干脆in只用来存数据，pop()的时候两个进行交换
        """
        if self.empty():
            return None

        for i in range(len(self.queue_in) - 1):
            self.queue_out.append(self.queue_in.popleft())
        
        self.queue_in, self.queue_out = self.queue_out, self.queue_in    # 交换in和out，这也是为啥in只用来存
        return self.queue_out.popleft()

    def top(self) -> int:
        """
        写法一：
        1. 首先确认不空
        2. 我们仅有in会存放数据，所以返回第一个即可（这里实际上用到了栈）
        写法二：
        1. 首先确认不空
        2. 因为队列的特殊性，FIFO，所以我们只有在pop()的时候才会使用queue_out
        3. 先把queue_in中的所有元素（除了最后一个），依次出列放进queue_out
        4. 交换in和out，此时out里只有一个元素
        5. 把out中的pop出来，即是原队列的最后一个，并使用temp变量暂存
        6. 把temp追加到queue_in的末尾
        """
        # 写法一：
        # if self.empty():
        #     return None
        
        # return self.queue_in[-1]    # 这里实际上用到了栈，因为直接获取了queue_in的末尾元素

        # 写法二：
        if self.empty():
            return None

        for i in range(len(self.queue_in) - 1):
            self.queue_out.append(self.queue_in.popleft())
        
        self.queue_in, self.queue_out = self.queue_out, self.queue_in 
        temp = self.queue_out.popleft()   
        self.queue_in.append(temp)
        return temp


    def empty(self) -> bool:
        """
        因为只有in存了数据，只要判断in是不是有数即可
        """
        return len(self.queue_in) == 0
````

## [232. 用栈实现队列](https://leetcode.cn/problems/implement-queue-using-stacks/)

简单  相关标签  ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业



请你仅使用两个栈实现先入先出队列。队列应当支持一般队列支持的所有操作（`push`、`pop`、`peek`、`empty`）：

实现 `MyQueue` 类：

- `void push(int x)` 将元素 x 推到队列的末尾
- `int pop()` 从队列的开头移除并返回元素
- `int peek()` 返回队列开头的元素
- `boolean empty()` 如果队列为空，返回 `true` ；否则，返回 `false`

**说明：**

- 你 **只能** 使用标准的栈操作 —— 也就是只有 `push to top`, `peek/pop from top`, `size`, 和 `is empty` 操作是合法的。
- 你所使用的语言也许不支持栈。你可以使用 list 或者 deque（双端队列）来模拟一个栈，只要是标准的栈操作即可。

 

**示例 1：**

```
输入：
["MyQueue", "push", "push", "peek", "pop", "empty"]
[[], [1], [2], [], [], []]
输出：
[null, null, null, 1, 1, false]

解释：
MyQueue myQueue = new MyQueue();
myQueue.push(1); // queue is: [1]
myQueue.push(2); // queue is: [1, 2] (leftmost is front of the queue)
myQueue.peek(); // return 1
myQueue.pop(); // return 1, queue is [2]
myQueue.empty(); // return false
```

````python
class MyQueue:
    def __init__(self):
        """
        in主要负责push，out主要负责pop
        """
        self.stack_in = []
        self.stack_out = []

    def push(self, x: int) -> None:
        """
        有新元素进来，就往in里面push
        """
        self.stack_in.append(x)

    def pop(self) -> int:
        if self.empty():
            return None # 沿用你原代码的返回None
        if self.stack_out:
            return self.stack_out.pop()
        else:
            # 转移元素到 stack_out
            while self.stack_in: # 使用 while loop 更清晰
                self.stack_out.append(self.stack_in.pop())
            return self.stack_out.pop()

    def peek(self) -> int:
        """
        Get the front element without removing it.
        """
        # 确保 stack_out 有元素或者进行转移
        if not self.stack_out:
             # 如果 stack_in 也为空，说明队列为空
            if self.empty():
                return None # 沿用你原代码的返回None

            # 转移元素到 stack_out
            while self.stack_in:
                self.stack_out.append(self.stack_in.pop())

        # 现在 stack_out 肯定有元素了，返回最后一个元素（也就是队列头部元素）
        return self.stack_out[-1]


    def empty(self) -> bool:
        """
        只要in或者out有元素，说明队列不为空
        """
        return not (self.stack_in or self.stack_out)

````

**题目背景**

队列是一种先进先出（FIFO）的数据结构，而栈是一种后进先出（LIFO）的数据结构。这道题要求我们使用栈来实现队列的功能。

**算法思路**

核心思想是利用两个栈：一个用于处理入队操作（记为 `stack_in`），另一个用于处理出队和查看队首元素操作（记为 `stack_out`）。

1.  **入队 (push)**：
    *   入队操作非常简单，直接将元素添加到 `stack_in` 栈的顶部即可。这模拟了新元素从队尾进入队列。

2.  **出队 (pop) 和 查看队首元素 (peek)**：
    *   这是最关键的部分。队列的先进先出要求我们从队头取出元素。然而，`stack_in` 的顶部是最后入队的元素（队尾元素）。
    *   为了获取队头元素，我们需要将 `stack_in` 中的元素“倒过来”。
    *   我们将 `stack_in` 中的所有元素依次弹出，并压入 `stack_out` 栈。
    *   请注意，由于栈的后进先出特性，`stack_in` 中最先进入的元素（队头元素）会最后被弹出。当这些元素被压入 `stack_out` 后，这个最初的队头元素就变成了 `stack_out` 的底部元素。而 `stack_in` 中最后进入的元素（队尾元素）会最先被弹出并压入 `stack_out`，成为 `stack_out` 的顶部元素。
    *   但是，我们出队或查看的是队头元素，它现在位于 `stack_out` 的**底部**（从 `stack_in` 弹出的顺序来看）。
    *   **更准确地说：** `stack_in` 中的元素顺序是：队头 -> ... -> 队尾（顶部）。当把 `stack_in` 元素依次弹出压入 `stack_out` 后，`stack_out` 中的元素顺序是：队尾 -> ... -> 队头（顶部）。所以，队头元素实际上位于 `stack_out` 的顶部。
    *   因此，当需要出队或查看队首时：
        *   首先检查 `stack_out` 是否为空。如果 `stack_out` 不为空，说明之前已经进行过转移，并且队头元素已经在 `stack_out` 的顶部，直接从 `stack_out` 弹出或查看顶部元素即可。
        *   如果 `stack_out` 为空，则需要将 `stack_in` 中的所有元素弹出并压入 `stack_out`。转移完成后，`stack_in` 变为空，`stack_out` 的顶部元素就是当前的队头元素。然后从 `stack_out` 弹出（对于 `pop`）或查看顶部元素（对于 `peek`）。

3.  **判断队列是否为空 (empty)**：
    *   如果 `stack_in` 和 `stack_out` 都为空，那么队列就为空。只要其中一个栈还有元素，队列就不为空。

**Python 代码实现**

```python
class MyQueue:
    def __init__(self):
        # stack_in 用于接收新入队的元素
        self.stack_in = []
        # stack_out 用于准备出队和查看队首元素
        self.stack_out = []

    def push(self, x: int) -> None:
        """
        入队操作：将元素添加到 stack_in
        """
        self.stack_in.append(x)
        # append 方法是 Python 列表的一个方法，用于在列表末尾添加元素。
        # 在这里，我们用它来模拟栈的 push 操作（将元素压入栈顶）。

    def pop(self) -> int:
        """
        出队操作：从队列头部移除元素并返回
        """
        # 语法：if self.empty(): return None
        # 检查队列是否为空。如果为空，根据题意返回 None。
        # 另一种更 Pythonic 的做法是抛出异常，如 raise IndexError("pop from empty queue")

        if self.empty():
             return None

        # 语法：if self.stack_out:
        # 在 Python 中，非空的列表被视为 True，空的列表被视为 False。
        # 所以这个条件判断 stack_out 是否不为空。

        if self.stack_out:
            # 如果 stack_out 不为空，直接从 stack_out 弹出顶部元素（队头）
            return self.stack_out.pop()
            # pop() 方法是 Python 列表的一个方法，用于移除并返回列表的最后一个元素。
            # 在这里，我们用它来模拟栈的 pop 操作（弹出栈顶元素）。

        else:
            # 如果 stack_out 为空，需要将 stack_in 中的元素转移过来
            # 语法：while self.stack_in:
            # 同样，利用列表的 True/False 特性，当 stack_in 不为空时循环。
            # 也可以写成 for i in range(len(self.stack_in)): 但 while 循环在这里更简洁直观。

            while self.stack_in:
                # 将 stack_in 的顶部元素弹出，并添加到 stack_out 的末尾
                self.stack_out.append(self.stack_in.pop())
                # stack_in.pop() 弹出 stack_in 的顶部元素
                # stack_out.append(...) 将弹出的元素添加到 stack_out 底部（逻辑上）

            # 转移完成后，stack_out 的顶部元素就是队头元素，弹出并返回
            return self.stack_out.pop()


    def peek(self) -> int:
        """
        查看队首元素：返回队列头部元素，但不移除
        """
        # 和 pop 类似，首先确保 stack_out 有元素或者进行转移
        if not self.stack_out:
            # 语法：if self.empty():
            # 检查队列是否为空

            if self.empty():
                 return None # 同样，空队列返回 None 或抛异常

            # 如果 stack_out 为空但 stack_in 不为空，则进行转移
            while self.stack_in:
                self.stack_out.append(self.stack_in.pop())

        # 现在 stack_out 肯定有元素了，返回 stack_out 的顶部元素（队头）
        # 语法：return self.stack_out[-1]
        # [-1] 是 Python 列表的索引方式，表示列表的最后一个元素。
        # 在这里，stack_out 的最后一个元素就是转移过来的最先入队的元素，即队头元素。

        return self.stack_out[-1]


    def empty(self) -> bool:
        """
        判断队列是否为空：当且仅当 stack_in 和 stack_out 都为空时，队列为空
        """
        # 语法：return not (self.stack_in or self.stack_out)
        # self.stack_in or self.stack_out：如果 stack_in 或 stack_out 中有任何一个非空，
        # 表达式结果为 True。只有当两者都为空时，结果才为 False。
        # not (...)：对前面的结果取反。所以，当两个栈都为空时 (False)，取反后返回 True (队列为空)。
        # 当至少一个栈不为空时 (True)，取反后返回 False (队列不为空)。

        return not (self.stack_in or self.stack_out)

```

**算法复杂度分析**

*   **`push(x)`**: 将元素添加到 `stack_in`，时间复杂度为 O(1)，因为列表的 `append` 操作通常是常数时间。
*   **`pop()`**:
    *   如果 `stack_out` 不为空，弹出元素的时间复杂度为 O(1)。
    *   如果 `stack_out` 为空，需要将 `stack_in` 中的所有元素转移到 `stack_out`。假设 `stack_in` 中有 `n` 个元素，转移需要 O(n) 的时间。转移完成后，弹出元素是 O(1)。
    *   **摊还分析 (Amortized Analysis)**：虽然单次 `pop` 操作最坏情况下是 O(n)，但考虑一系列操作，每个元素从 `stack_in` 转移到 `stack_out` 只会发生一次。因此，在平均意义上，每个元素的出队操作的时间复杂度是 O(1)。所以，`pop` 的摊还时间复杂度是 O(1)。
*   **`peek()`**:
    *   和 `pop` 类似，如果 `stack_out` 不为空，查看元素的时间复杂度为 O(1)。
    *   如果 `stack_out` 为空，需要进行 O(n) 的转移。转移完成后，查看顶部元素是 O(1)。
    *   同样，`peek` 的摊还时间复杂度是 O(1)。
*   **`empty()`**: 判断两个列表是否为空，时间复杂度为 O(1)。

空间复杂度为 O(n)，其中 `n` 是队列中元素的数量，因为最坏情况下，所有元素会存储在 `stack_in` 或 `stack_out` 中。

**总结**

这段代码成功地利用了两个栈的特性，巧妙地模拟了队列的先进先出行为。通过将入队和出队操作分摊到两个栈上，并在需要时进行元素转移，实现了队列的核心功能。同时，代码中也运用了一些基本的 Python 列表操作(`append`, `pop`, 索引 `[-1]`) 和布尔逻辑判断。

## [239. 滑动窗口最大值](https://leetcode.cn/problems/sliding-window-maximum/)

困难   相关标签   ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业

给你一个整数数组 `nums`，有一个大小为 `k` 的滑动窗口从数组的最左侧移动到数组的最右侧。你只可以看到在滑动窗口内的 `k` 个数字。滑动窗口每次只向右移动一位。

返回 *滑动窗口中的最大值* 。

 

**示例 1：**

```
输入：nums = [1,3,-1,-3,5,3,6,7], k = 3
输出：[3,3,5,5,6,7]
解释：
滑动窗口的位置                最大值
---------------               -----
[1  3  -1] -3  5  3  6  7       3
 1 [3  -1  -3] 5  3  6  7       3
 1  3 [-1  -3  5] 3  6  7       5
 1  3  -1 [-3  5  3] 6  7       5
 1  3  -1  -3 [5  3  6] 7       6
 1  3  -1  -3  5 [3  6  7]      7
```

**示例 2：**

```
输入：nums = [1], k = 1
输出：[1]
```

````python
from typing import List
from collections import deque

class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        
        window = deque()
        result = []

        for i in range(len(nums)):
            if window and window[0] < i - k + 1:
                window.popleft()
            while window and nums[window[-1]] <= nums[i]:
                window.pop()

            window.append(i)

            if i >= k - 1:
                result.append(nums[window[0]])

        return result
    
solition = Solution()
input_nums = input("请输入nums：")
input_nums = eval(input_nums)
input_k = input("请输入k：")
input_k = int(input_k)
output_result = solition.maxSlidingWindow(input_nums, input_k)
print(output_result)
````

**算法思想：单调队列（Monotonic Queue）**

解决“滑动窗口最大值”这个问题的核心算法是**单调队列（Monotonic Queue）**。它的核心思想是维护一个队列，使得队列中的元素（或元素的索引）是按照某个特定的顺序（在这里是递减）排列的。

为什么要用单调队列呢？

传统的暴力方法是每次滑动窗口移动一步，就遍历窗口内的所有元素找到最大值。如果数组长度为 `n`，窗口大小为 `k`，那么每次遍历窗口需要 O(k) 的时间，总共需要 O(n * k) 的时间。当 `n` 和 `k` 都很大时，效率就很低。

单调队列的优点在于，它可以在 O(1) 的时间复杂度内找到当前窗口的最大值，从而将总时间复杂度降低到 O(n)。

单调队列具体是如何工作的呢？

* **队列中存储的是元素的索引**：我们存储的是元素在原始数组中的索引，而不是元素本身。这样做是因为我们需要知道元素是否还在当前的滑动窗口内。
* **队列中的元素对应的值是单调递减的**：这是关键。队列的头部（左边）始终是当前窗口内最大的元素的索引。
* **维护单调性**：
    * **移除过期元素**：当滑动窗口向右移动时，最左边的元素可能会超出窗口范围。我们需要将队列中这些超出范围的元素的索引移除。
    * **移除较小元素**：当一个新的元素进入窗口时，如果队列末尾的元素小于或等于这个新元素，那么这些队列末尾的元素永远不会成为当前窗口或未来窗口的最大值，因为新元素更大且更靠近窗口的右侧。所以，我们需要将这些较小的元素从队列末尾移除，以保持队列的递减性质。
* **获取最大值**：由于队列头部始终是当前窗口内最大元素的索引，我们只需要取出队列头部的索引，然后根据这个索引找到对应的元素值即可。

**代码解析（逐行讲解）**

现在我们结合代码来看单调队列是如何实现的：

```python
class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        """
        给定一个整数数组 nums，有一个大小为 k 的滑动窗口从数组的最左侧移动到数组的最右侧。
        你只可以看到在滑动窗口内的 k 个数字。滑动窗口每次只向右移动一位。

        返回 滑动窗口中的最大值。

        Args:
            nums: 整数数组。
            k: 滑动窗口的大小。

        Returns:
            滑动窗口中的最大值数组。
        """
        from collections import deque # 导入双端队列模块

        # 双端队列，用于存储窗口内可能成为最大值的元素的索引
        # 保证队列中的元素索引对应的值是单调递减的
        window = deque()
        result = [] # 用于存储每个窗口的最大值

        for i in range(len(nums)): # 遍历数组中的每一个元素，i 表示当前元素的索引
            # 移除窗口左侧已经不再窗口内的元素的索引
            # window[0] 存储的是当前窗口最左侧元素的索引
            # i 是当前元素的索引
            # k 是窗口大小
            # i - k + 1 是当前窗口最左侧元素的索引
            # 如果 window[0] < i - k + 1，说明 window[0] 对应的元素已经不在当前窗口内了
            if window and window[0] < i - k + 1:
                window.popleft() # 从队列左侧移除元素

            # 移除窗口右侧所有小于当前元素的元素的索引
            # 保证队列中的元素对应的值是单调递减的
            # while window and nums[window[-1]] <= nums[i]:
            # window[-1] 是队列末尾元素的索引
            # nums[window[-1]] 是队列末尾元素的值
            # nums[i] 是当前元素的值
            # 如果队列末尾元素的值小于等于当前元素的值，说明队列末尾元素不会是当前或未来窗口的最大值
            # 因为当前元素更大且在更靠右的位置
            while window and nums[window[-1]] <= nums[i]:
                window.pop() # 从队列右侧移除元素

            # 将当前元素的索引添加到队列末尾
            window.append(i)

            # 当窗口大小达到 k 时，窗口最左侧的元素（即 window[0]）就是当前窗口的最大值
            # 当 i >= k - 1 时，说明当前处理的元素是第 k 个元素或之后，此时窗口已经完整
            if i >= k - 1:
                result.append(nums[window[0]]) # 将当前窗口的最大值（队列头部索引对应的元素值）添加到结果列表

        return result # 返回存储最大值的列表
```

**Python 语法讲解**

* **`from collections import deque`**: 导入 Python 内置模块 `collections` 中的 `deque` 类。`deque` 是双端队列（double-ended queue）的缩写，它支持在队列的两端快速添加和删除元素，非常适合实现单调队列。
* **`window = deque()`**: 创建一个空的双端队列对象 `window`。
* **`result = []`**: 创建一个空的列表 `result`，用于存储每个滑动窗口的最大值。
* **`for i in range(len(nums)):`**: 这是一个标准的 Python 循环，遍历数组 `nums` 的每一个元素的索引。`len(nums)` 返回数组的长度，`range(len(nums))` 生成从 0 到 `len(nums) - 1` 的整数序列。
* **`if window and window[0] < i - k + 1:`**:
    * `window`: 检查队列是否为空。如果队列为空，则不执行后面的条件判断。
    * `window[0]`: 获取队列最左侧的元素的索引。
    * `i - k + 1`: 计算当前滑动窗口最左侧元素的索引。例如，当 `i = k - 1` 时，当前窗口是 `nums[0]` 到 `nums[k-1]`，最左侧元素的索引是 0，即 `(k - 1) - k + 1 = 0`。当 `i = k` 时，当前窗口是 `nums[1]` 到 `nums[k]`，最左侧元素的索引是 1，即 `k - k + 1 = 1`。
    * `window[0] < i - k + 1`: 判断队列最左侧元素的索引是否小于当前窗口最左侧元素的索引。如果小于，说明该元素已经移出了当前窗口。
* **`window.popleft()`**: 从双端队列的左侧（头部）移除一个元素。
* **`while window and nums[window[-1]] <= nums[i]:`**:
    * `window`: 检查队列是否为空。
    * `window[-1]`: 获取队列最右侧（尾部）元素的索引。
    * `nums[window[-1]]`: 根据队列尾部元素的索引，获取该元素在原始数组中的值。
    * `nums[i]`: 获取当前遍历到的元素的值。
    * `nums[window[-1]] <= nums[i]`: 比较队列尾部元素的值和当前元素的值。
    * `while`: 只要条件成立，就一直循环执行 `window.pop()`。
* **`window.pop()`**: 从双端队列的右侧（尾部）移除一个元素。
* **`window.append(i)`**: 将当前元素的索引 `i` 添加到双端队列的右侧（尾部）。
* **`if i >= k - 1:`**: 判断当前遍历到的元素索引是否已经使得滑动窗口形成完整的大小 `k`。当 `i` 等于 `k-1` 时，第一次形成完整窗口。
* **`result.append(nums[window[0]])`**:
    * `window[0]`: 获取队列头部元素的索引，这个索引对应的元素就是当前窗口的最大值。
    * `nums[window[0]]`: 根据头部索引获取最大值。
    * `result.append(...)`: 将当前窗口的最大值添加到 `result` 列表中。
* **`return result`**: 返回存储了所有滑动窗口最大值的列表。

**举例说明**

假设 `nums = [1, 3, -1, -3, 5, 3, 6, 7]`，`k = 3`。

1. **i = 0, nums[0] = 1:**
   - `window` 是空的。
   - 队列为空，不执行 `popleft()`。
   - 队列为空，不执行 `while` 循环。
   - `window.append(0)`，`window = deque([0])`。
   - `i = 0 < k - 1 = 2`，不添加到结果。

2. **i = 1, nums[1] = 3:**
   - `window = deque([0])`。`window[0] = 0`。`i - k + 1 = 1 - 3 + 1 = -1`。`0 < -1` 不成立，不执行 `popleft()`。
   - `window = deque([0])`。`nums[window[-1]] = nums[0] = 1`。`nums[i] = nums[1] = 3`。`1 <= 3` 成立。
   - `window.pop()`，`window = deque([])`。
   - `window.append(1)`，`window = deque([1])`。
   - `i = 1 < k - 1 = 2`，不添加到结果。

3. **i = 2, nums[2] = -1:**
   - `window = deque([1])`。`window[0] = 1`。`i - k + 1 = 2 - 3 + 1 = 0`。`1 < 0` 不成立，不执行 `popleft()`。
   - `window = deque([1])`。`nums[window[-1]] = nums[1] = 3`。`nums[i] = nums[2] = -1`。`3 <= -1` 不成立，不执行 `while` 循环。
   - `window.append(2)`，`window = deque([1, 2])`。
   - `i = 2 >= k - 1 = 2`。当前窗口是 `[1, 3, -1]`。`window[0] = 1`，`nums[1] = 3`。
   - `result.append(nums[window[0]])`，`result = [3]`。

4. **i = 3, nums[3] = -3:**
   - `window = deque([1, 2])`。`window[0] = 1`。`i - k + 1 = 3 - 3 + 1 = 1`。`1 < 1` 不成立，不执行 `popleft()`。
   - `window = deque([1, 2])`。`nums[window[-1]] = nums[2] = -1`。`nums[i] = nums[3] = -3`。`-1 <= -3` 不成立，不执行 `while` 循环。
   - `window.append(3)`，`window = deque([1, 2, 3])`。
   - `i = 3 >= k - 1 = 2`。当前窗口是 `[3, -1, -3]`。`window[0] = 1`，`nums[1] = 3`。
   - `result.append(nums[window[0]])`，`result = [3, 3]`。

5. **i = 4, nums[4] = 5:**
   - `window = deque([1, 2, 3])`。`window[0] = 1`。`i - k + 1 = 4 - 3 + 1 = 2`。`1 < 2` 成立。
   - `window.popleft()`，`window = deque([2, 3])`。
   - `window = deque([2, 3])`。`nums[window[-1]] = nums[3] = -3`。`nums[i] = nums[4] = 5`。`-3 <= 5` 成立。`window.pop()`，`window = deque([2])`。
   - `window = deque([2])`。`nums[window[-1]] = nums[2] = -1`。`nums[i] = nums[4] = 5`。`-1 <= 5` 成立。`window.pop()`，`window = deque([])`。
   - `window.append(4)`，`window = deque([4])`。
   - `i = 4 >= k - 1 = 2`。当前窗口是 `[-1, -3, 5]`。`window[0] = 4`，`nums[4] = 5`。
   - `result.append(nums[window[0]])`，`result = [3, 3, 5]`。

... 以此类推，直到遍历完整个数组。

最终 `result` 将会是 `[3, 3, 5, 5, 6, 7]`。

## [242. 有效的字母异位词](https://leetcode.cn/problems/valid-anagram/)

简单  相关标签  ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业



给定两个字符串 `s` 和 `t` ，编写一个函数来判断 `t` 是否是 `s` 的 字母异位词。

 

**示例 1:**

```
输入: s = "anagram", t = "nagaram"
输出: true
```

**示例 2:**

```
输入: s = "rat", t = "car"
输出: false
```

````python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        record = [0] * 26
        for i in s:
            record[ord(i) - ord("a")] += 1
        for i in t:
            record[ord(i) - ord("a")] -= 1
        for i in range(26):
            if record[i] != 0:
                return False
        return True
````

好的，我来详细解释一下这段判断两个字符串是否是字母异位词（Anagram）的Python代码。我会尽量用通俗易懂的语言，即使是编程小白也能理解。

**什么是字母异位词 (Anagram)?**

字母异位词是指两个字符串包含的字母种类和数量完全相同，但排列顺序可能不同。

*   例如："anagram" 和 "nagaram" 是字母异位词。
*   例如："rat" 和 "car" 不是字母异位词。

**代码的整体思路**

这段代码判断两个字符串 `s` 和 `t` 是否是字母异位词的核心思想是：

1.  统计字符串 `s` 中每个字母出现的次数。
2.  统计字符串 `t` 中每个字母出现的次数。
3.  比较两个字符串中每个字母的出现次数是否完全一致。

如果次数完全一致，那么它们就是字母异位词；否则就不是。

**代码分步解释**

```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        # 初始化一个长度为26的列表，用来记录每个字母的出现次数
        record = [0] * 26

        # 遍历字符串 s
        for i in s:
            # 计算当前字母在字母表中的位置 (0-25)
            # ord(i) 获取字母 i 的ASCII值
            # ord("a") 获取字母 "a" 的ASCII值
            # ord(i) - ord("a") 得到字母 i 相对于 "a" 的偏移量
            # 例如：ord("b") - ord("a") = 98 - 97 = 1 (代表字母 'b' 在字母表中的第1个位置，从0开始计数)
            index = ord(i) - ord("a")

            # 对应位置的计数器加一
            record[index] += 1

        # 遍历字符串 t
        for i in t:
            # 计算当前字母在字母表中的位置
            index = ord(i) - ord("a")

            # 对应位置的计数器减一
            record[index] -= 1

        # 检查 record 列表中所有的计数器
        for i in range(26):
            # 如果任何一个计数器不为0，说明两个字符串中对应字母的数量不一致
            if record[i] != 0:
                # 返回 False，表示不是字母异位词
                return False

        # 如果所有计数器都为0，说明两个字符串中所有字母的数量都一致
        # 返回 True，表示是字母异位词
        return True
```

*   `return False`: 如果条件满足，立即结束函数并返回 `False`。
*   `return True`: 如果前面的循环正常结束（即没有发现任何一个 `record[i]` 不等于 0），则执行这句，返回 `True`。

**算法解释**

这段代码使用的算法是基于**哈希表（或者说，这里使用了列表作为简化的哈希表）**的计数法。

1.  **初始化计数器:** 我们创建一个长度为26的列表 `record`，初始化所有元素为0。这个列表相当于一个哈希表，键（key）是字母（通过索引 `ord(i) - ord('a')` 得到），值（value）是该字母出现的次数。
2.  **统计第一个字符串:** 遍历第一个字符串 `s`。每遇到一个字母，就找到它在字母表中的对应位置（索引），并将 `record` 中该位置的计数器加一。这样，`record` 就记录了字符串 `s` 中每个字母出现的频率。
3.  **抵消第二个字符串:** 遍历第二个字符串 `t`。每遇到一个字母，就找到它在字母表中的对应位置，并将 `record` 中该位置的计数器减一。
4.  **检查结果:** 如果 `s` 和 `t` 是字母异位词，那么它们包含的每个字母的数量应该完全相同。在处理完 `s` 后，`record` 中存放的是 `s` 中各字母的计数。在处理完 `t` 后，`record` 中存放的是 `s` 中各字母的计数减去 `t` 中各字母的计数。如果它们是字母异位词，那么每个字母的计数差都应该是0。因此，我们检查 `record` 列表中的所有元素是否都为0。
5.  **判断:** 如果 `record` 中有任何一个元素不为0，说明有某个字母在 `s` 和 `t` 中出现的次数不同，它们就不是字母异位词，返回 `False`。如果所有元素都为0，说明所有字母出现的次数都相同，它们就是字母异位词，返回 `True`。

**为什么这个算法有效？**

这个算法巧妙地利用了计数器。通过对 `s` 中的字母进行加一操作，对 `t` 中的字母进行减一操作，如果两个字符串是字母异位词，那么最终每个字母的净变化（加的数量减去减的数量）一定是0。如果最终任何一个字母的计数不为0，就说明两个字符串在那个字母的数量上存在差异，因此它们不是字母异位词。

**优点**

*   **效率高:** 遍历字符串的时间复杂度是线性的（O(n)，n为字符串长度），最后检查计数器的时间复杂度是常数（O(26)，因为字母数量固定）。所以总的时间复杂度是 O(n)。
*   **空间复杂度低:** 只使用了固定大小的列表（长度为26），所以空间复杂度是 O(1)。
