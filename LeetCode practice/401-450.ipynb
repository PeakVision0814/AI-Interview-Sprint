## [415. 字符串相加](https://leetcode.cn/problems/add-strings/)

简单  相关标签  ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业



给定两个字符串形式的非负整数 `num1` 和`num2` ，计算它们的和并同样以字符串形式返回。

你不能使用任何內建的用于处理大整数的库（比如 `BigInteger`）， 也不能直接将输入的字符串转换为整数形式。

 

**示例 1：**

```
输入：num1 = "11", num2 = "123"
输出："134"
```

**示例 2：**

```
输入：num1 = "456", num2 = "77"
输出："533"
```

**示例 3：**

```
输入：num1 = "0", num2 = "0"
输出："0"
```

````python
class Solution:
    def addStrings(self, num1: str, num2: str) -> str:
        i, j = len(num1) - 1, len(num2) - 1
        carry = 0
        res = []
        
        while i >= 0 or j >= 0 or carry:
            # 字符转数字: ord('5') - ord('0') = 5
            n1 = int(num1[i]) if i >= 0 else 0
            n2 = int(num2[j]) if j >= 0 else 0
            
            temp_sum = n1 + n2 + carry
            carry = temp_sum // 10
            res.append(str(temp_sum % 10))
            
            i -= 1
            j -= 1
            
        # 结果是反的，要倒回来
        return "".join(res[::-1])

# --- 交互式测试 ---
solution = Solution()
print("\n=== 测试字符串相加 ===")
n1 = input("请输入第一个数字 (字符串): ").strip()
n2 = input("请输入第二个数字 (字符串): ").strip()

result = solution.addStrings(n1, n2)
print(f"计算结果: {result}")
````

**🧠 算法思想（整体思路）**

我们要像小学时做竖式加法那样：
- 从两个数字的**最后一位**开始相加；
- 每次加上进位（carry）；
- 得到当前位的结果和新的进位；
- 最后把结果反转（因为我们是从低位往高位算的，但存储是从左到右）。

例如：
```
  123
+ 456
-----
  579
```

如果位数不同、有进位（比如 999 + 1 = 1000），也要能处理。

**🔍 代码逐行解析**

```python
class Solution:
    def addStrings(self, num1: str, num2: str) -> str:
```
- 定义一个类 `Solution`，其中有一个方法 `addStrings`。
- 输入是两个**字符串形式的非负整数** `num1` 和 `num2`。
- 返回它们相加后的**字符串结果**。

```python
        i, j = len(num1) - 1, len(num2) - 1
```
- `i` 和 `j` 是两个指针，分别指向 `num1` 和 `num2` 的**最后一个字符**（即最低位）。
- 因为字符串索引从 0 开始，所以末尾是 `len - 1`。

```python
        carry = 0
        res = []
```
- `carry`：记录进位，初始为 0。
- `res`：用来收集每一位的计算结果（注意：我们会**先存低位**，所以最后要反转）。

```python
        while i >= 0 or j >= 0 or carry:
```
- 循环条件：只要还有数字没处理完（`i >= 0` 或 `j >= 0`），或者还有进位（`carry != 0`），就继续。
- 特别注意：即使两个数都处理完了，但如果还有进位（比如 999 + 1 最后多出一个 1），也要再循环一次。

```python
            n1 = int(num1[i]) if i >= 0 else 0
            n2 = int(num2[j]) if j >= 0 else 0
```
- 如果 `i >= 0`，说明 `num1` 还有数字，取当前位转成整数；
- 否则当作 0（相当于短的数字前面补 0）。
- `int('5')` 是合法的，Python 允许这样转单个数字字符。

> 💡 注释里提到 `ord('5') - ord('0')` 是另一种方式，但这里直接用 `int()` 更简洁，也是允许的（题目只禁止整体转成大整数，逐位转没问题）。

```python
            temp_sum = n1 + n2 + carry
```
- 当前位的总和 = num1 当前位 + num2 当前位 + 上一次的进位。

```python
            carry = temp_sum // 10
            res.append(str(temp_sum % 10))
```
- `temp_sum // 10`：得到新的进位（比如 15 // 10 = 1）。
- `temp_sum % 10`：得到当前位应写的数字（比如 15 % 10 = 5）。
- 把这个数字转成字符串，加入 `res` 列表。

> ⚠️ 注意：此时 `res` 是**低位在前**，比如计算 123 + 456，`res` 会先存 `'9'`，再 `'7'`，再 `'5'` → `['9','7','5']`。

```python
            i -= 1
            j -= 1
```
- 指针左移，处理更高一位（向字符串开头移动）。

```python
        return "".join(res[::-1])
```
- `res[::-1]`：将列表反转，变成高位在前。
- `"".join(...)`：拼成最终字符串。

例如：`['9','7','5']` → 反转为 `['5','7','9']` → `"579"`。

**✅ 举个例子走一遍**

输入：`num1 = "456"`, `num2 = "77"`

步骤：
- i=2 ('6'), j=1 ('7') → 6+7+0=13 → carry=1, append '3'
- i=1 ('5'), j=0 ('7') → 5+7+1=13 → carry=1, append '3'
- i=0 ('4'), j=-1 → 4+0+1=5 → carry=0, append '5'
- i=-1, j=-1, carry=0 → 循环结束
- res = ['3','3','5'] → 反转 → "533"

正确！

**📌 总结关键点**

| 要点 | 说明 |
|------|------|
| **双指针** | 从右往左遍历两个字符串 |
| **补零处理** | 短的数前面视为 0 |
| **进位处理** | 每次加完更新 carry |
| **结果反转** | 因为先算低位，要反转才是正确顺序 |
| **不转整体整数** | 符合题目限制（逐位转是可以的） |

## [450. 删除二叉搜索树中的节点](https://leetcode.cn/problems/delete-node-in-a-bst/)

中等  相关标签  ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业



给定一个二叉搜索树的根节点 **root** 和一个值 **key**，删除二叉搜索树中的 **key** 对应的节点，并保证二叉搜索树的性质不变。返回二叉搜索树（有可能被更新）的根节点的引用。

一般来说，删除节点可分为两个步骤：

1. 首先找到需要删除的节点；
2. 如果找到了，删除它。

 

**示例 1:**

![img](https://assets.leetcode.com/uploads/2020/09/04/del_node_1.jpg)

```
输入：root = [5,3,6,2,4,null,7], key = 3
输出：[5,4,6,2,null,null,7]
解释：给定需要删除的节点值是 3，所以我们首先找到 3 这个节点，然后删除它。
一个正确的答案是 [5,4,6,2,null,null,7], 如下图所示。
另一个正确答案是 [5,2,6,null,4,null,7]。
```

**示例 2:**

```
输入: root = [5,3,6,2,4,null,7], key = 0
输出: [5,3,6,2,4,null,7]
解释: 二叉树不包含值为 0 的节点
```

**示例 3:**

```
输入: root = [], key = 0
输出: []
```

````python
from typing import Optional

# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    def deleteNode(self, root: Optional[TreeNode], key: int) -> Optional[TreeNode]:
        
        if not root:
            return None
        
        if key < root.val:
            root.left = self.deleteNode(root.left, key)
        elif key > root.val:
            root.right = self.deleteNode(root.right, key)
        else:
            if not root.left:
                return root.right
            if not root.right:
                return root.left
            
            succesor = self.get_min(root.right)

            root.val = succesor.val

            root.right = self.deleteNode(root.right, succesor.val)

        return root
    def get_min(self, node):
        while node.left:
            node = node.left
            return node
````

**1. 核心直觉 (First Principles)**

在 BST 中删除节点比插入节点难，因为直接删除可能会导致树“断裂”。我们需要做的是**“断肢重连”**。

整个过程分为两个阶段：

1. **查找 (Search)**: 利用 BST 左小右大的特性，递归找到目标节点。
2. **处理 (Handle)**: 找到后，根据目标节点的“家庭状况”（子节点数量），分三种情况处理。


**2. 三种情况详解 (Case Analysis)**

当 `key == root.val` 时，我们要删除当前节点 `root`：

情况 1：无子节点 (叶子)

* **策略**: 直接删除。
* **操作**: 返回 `None`。
* **效果**: 父节点原本指向它的指针，现在指向了 `None`。

情况 2：只有一个子节点 (单亲)

* **策略**: **“子承父业”**。
* **操作**:
* 如果只有左孩子，返回 `root.left`。
* 如果只有右孩子，返回 `root.right`。


* **效果**: 父节点跳过当前节点，直接连到了孙子节点上（链表删除风格）。

情况 3：有两个子节点 (双全) —— **重难点**

* **问题**: 不能直接删，否则树会断成三截。
* **策略**: **“寻找替身” (Successor)**。
* **原理**: 在**右子树**中找到**最小**的节点（Successor）。
* 为什么要找它？因为它**大于**左子树所有节点，且**小于**右子树其他节点。它是唯一能完美顶替 `root` 位置的人。


* **操作步骤**:
1. **查找**: 在右子树中一路向左，找到最小值 `successor`。
2. **夺舍**: 将 `successor.val` 赋值给当前 `root.val` (只换值，不动结构)。
3. **灭口**: 既然值已经拿来了，原来的 `successor` 节点就没用了。递归调用 `deleteNode`，在右子树中把那个 `successor` 删掉。


* *注：因为 successor 是右子树最小节点，它一定没有左孩子，所以删除它只会触发 Case 1 或 Case 2，非常简单。*



**3. 代码实现 (Python Standard Template)**

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right

class Solution:
    def deleteNode(self, root: Optional[TreeNode], key: int) -> Optional[TreeNode]:
        # Base Case: 找不到节点，直接返回 None
        if not root:
            return None
        
        # --- 第一阶段：查找 ---
        if key < root.val:
            # 去左边删，删完后的新左子树接回来
            root.left = self.deleteNode(root.left, key)
        elif key > root.val:
            # 去右边删，删完后的新右子树接回来
            root.right = self.deleteNode(root.right, key)
            
        else:
            # --- 第二阶段：找到了，开始删除 ---
            
            # Case 1 & 2: 没有孩子 或 只有一个孩子
            if not root.left:
                return root.right # 如果左空，返回右（哪怕右也是空，也没关系）
            if not root.right:
                return root.left  # 如果右空，返回左
            
            # Case 3: 两个孩子都有
            # 1. 找替身 (右子树的最小值)
            successor = self.get_min(root.right)
            
            # 2. 夺舍 (值替换)
            root.val = successor.val
            
            # 3. 灭口 (去右子树把那个替身删掉)
            # 注意：这里递归删的是 successor.val
            root.right = self.deleteNode(root.right, successor.val)
            
        return root

    def get_min(self, node):
        """辅助函数：找最左边的节点"""
        while node.left:
            node = node.left
        return node

```

**5. 面试官追问 (Q&A)**

**Q: 为什么要找右子树的最小节点？找左子树的最大节点可以吗？**
**A:** 可以。

* **右子树最小 (Successor)**: 后继节点。
* **左子树最大 (Predecessor)**: 前驱节点。
两者都可以维持 BST 的性质。代码里选哪一个都行，标准库通常选 Successor。

**Q: 为什么 Case 3 中删除 Successor 不需要再考虑 Case 3？**
**A:** 因为 Successor 是“右子树中最小的节点”，这意味着它**一定没有左孩子**（如果有左孩子，它就不是最小的了）。
所以，删除 Successor 时，只可能遇到 Case 1 (它是叶子) 或 Case 2 (它只有右孩子)。这保证了递归一定能终止，不会无限循环。
