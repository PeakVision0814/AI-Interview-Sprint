## [101. 对称二叉树](https://leetcode.cn/problems/symmetric-tree/)

简单  相关标签  ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业



给你一个二叉树的根节点 `root` ， 检查它是否轴对称。

 

**示例 1：**

![img](https://pic.leetcode.cn/1698026966-JDYPDU-image.png)

```
输入：root = [1,2,2,3,4,4,3]
输出：true
```

**示例 2：**

![img](https://pic.leetcode.cn/1698027008-nPFLbM-image.png)

```
输入：root = [1,2,2,null,3,null,3]
输出：false
```

 

````python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    def isSymmetric(self, root: Optional[TreeNode]) -> bool:
        if not root:
            return True
        return self.isMirror(root.left, root.right)

    def isMirror(self, left: Optional[TreeNode], right: Optional[TreeNode]) -> bool:
        if not left and not right:
            return True
        if not left or not right:
            return False
        if left.val != right.val:
            return False
        return self.isMirror(left.left, right.right) and self.isMirror(left.right, right.left)
````

**1. 核心思路：递归 (Recursion)**
判断一棵树是否轴对称，本质上是判断**根节点的左子树**和**右子树**是否互为**镜像**。

这需要我们同时从两边进行遍历。我们定义一个辅助函数 `isMirror(p, q)`，逻辑如下：
* **当前层比较**：`p` 和 `q` 的值必须相等。
* **下一层比较（镜像法则）**：
    1.  **外侧对内侧**：`p` 的左孩子 (`p.left`) 必须等于 `q` 的右孩子 (`q.right`)。
    2.  **内侧对内侧**：`p` 的右孩子 (`p.right`) 必须等于 `q` 的左孩子 (`q.left`)。

**2. 代码实现**
```python
class Solution:
    def isSymmetric(self, root: Optional[TreeNode]) -> bool:
        if not root:
            return True
        # 开启镜像检查模式：左子树 vs 右子树
        return self.isMirror(root.left, root.right)

    def isMirror(self, left: Optional[TreeNode], right: Optional[TreeNode]) -> bool:
        # 1. 终止条件 (Terminator)
        if not left and not right:
            return True  # 两个都为空，是对称的
        if not left or not right:
            return False # 一个空一个不空，不对称
        if left.val != right.val:
            return False # 值不一样，不对称

        # 2. 递归下探 (Drill Down)
        # 必须同时满足：外侧相等 AND 内侧相等
        outside = self.isMirror(left.left, right.right) # 左的左 vs 右的右
        inside = self.isMirror(left.right, right.left)  # 左的右 vs 右的左
        
        return outside and inside

```

**3. 复杂度分析**

* **时间复杂度**: 。每个节点都被访问一次。
* **空间复杂度**: 。递归栈的深度等于树的高度。

**4. 记忆口诀**

**“左手的左手找右手的右手（外侧），左手的右手找右手的左手（内侧）。”**

## [102. 二叉树的层序遍历](https://leetcode.cn/problems/binary-tree-level-order-traversal/)

中等  相关标签   ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业



提示



给你二叉树的根节点 `root` ，返回其节点值的 **层序遍历** 。 （即逐层地，从左到右访问所有节点）。

 

**示例 1：**

![img](https://assets.leetcode.com/uploads/2021/02/19/tree1.jpg)

```
输入：root = [3,9,20,null,null,15,7]
输出：[[3],[9,20],[15,7]]
```

**示例 2：**

```
输入：root = [1]
输出：[[1]]
```

**示例 3：**

```
输入：root = []
输出：[]
```

````python
import collections
from typing import Optional, List
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    def levelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
        if not root:
            return[]
        result = []
        queue = collections.deque([root])

        while queue:
            level_size = len(queue)
            current_level_valuse = []
            for _ in range(level_size):
                node = queue.popleft()
                current_level_valuse.append(node.val)
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
            result.append(current_level_valuse)
        return result
````

**代码解释：**

1. **`import collections`**: 导入 `collections` 模块，为了使用 `deque`。
2. **`class TreeNode:`**: 这是 LeetCode 题目提供的 `TreeNode` 类的定义，用于表示树的节点。
3. **`class Solution:`**: 按照 LeetCode 的惯例，我们的解题代码会放在一个类里面。
4. **`def levelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:`**: 这是我们要实现的函数，接收一个 `TreeNode` 类型的 `root` 节点作为输入，返回一个 `List[List[int]]` 类型的结果。`Optional` 表示 `root` 可以是 `None`，`List[List[int]]` 表示一个列表，其中每个元素也是一个整数列表。
5. **`if not root: return []`**: 处理空树的情况。
6. **`result = []`**: 初始化一个空列表，用来存储每一层的节点值列表。
7. **`queue = collections.deque([root])`**: 创建一个双端队列 `deque`，并将根节点 `root` 加入队列作为初始元素。`deque` 相比普通的 list 在队列操作（头部添加和删除）上更高效。
8. **`while queue:`**: 只要队列不为空，就继续循环。这意味着还有未访问的节点。
9. **`level_size = len(queue)`**: 在开始处理当前层之前，获取队列的长度。这个长度就是当前层节点的数量。**这是 BFS 层序遍历的关键之处，它帮助我们区分不同的层。**
10. **`current_level_values = []`**: 初始化一个空列表，用来存储当前这一层的所有节点的值。
11. **`for _ in range(level_size):`**: 循环 `level_size` 次，处理当前层的每一个节点。`_` 是一个习惯用法，表示我们不关心循环的计数变量本身。
12. **`node = queue.popleft()`**: 从队列的左端取出（并移除）一个节点。这是队列的“出队”操作。
13. **`current_level_values.append(node.val)`**: 将当前节点的值添加到 `current_level_values` 列表中。
14. **`if node.left: queue.append(node.left)`**: 如果当前节点有左孩子，将左孩子添加到队列的右端（入队）。
15. **`if node.right: queue.append(node.right)`**: 如果当前节点有右孩子，将右孩子添加到队列的右端（入队）。
16. **`result.append(current_level_values)`**: 当前层的所有节点处理完毕，将 `current_level_values` 列表（包含当前层所有节点的值）添加到 `result` 列表中。
17. **`return result`**: 循环结束（队列为空），所有节点都被访问，返回最终的 `result` 列表。

**Python 语法解释：**

- **`if not root:`**: 检查 `root` 是否为 `None`。在 Python 中，`None`、空字符串、空列表、空字典、数字 0 等在布尔上下文中都被视为 `False`。所以 `not root` 当 `root` 为 `None` 时为 `True`。
- **`[root]`**: 创建一个只包含 `root` 元素的列表。
- **`collections.deque([...])`**: 创建一个 `deque` 对象。
- **`queue.append(element)`**: 在 `deque` 的右端添加一个元素。
- **`queue.popleft()`**: 从 `deque` 的左端移除并返回一个元素。
- **`len(queue)`**: 获取 `deque` 的长度（元素数量）。
- **`for _ in range(level_size):`**: `range(level_size)` 会生成一个从 0 到 `level_size - 1` 的整数序列。这个 `for` 循环会执行 `level_size` 次。
- **`node.val`, `node.left`, `node.right`**: 访问 `TreeNode` 对象的属性。
- **`[]`**: 创建一个空列表。
- **`list.append(element)`**: 在列表的末尾添加一个元素。
- **`list.append(other_list)`**: 在列表的末尾添加另一个列表。在这里我们是将 `current_level_values` 这个列表添加到 `result` 列表中。

## [104. 二叉树的最大深度](https://leetcode.cn/problems/maximum-depth-of-binary-tree/)

简单 相关标签 ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业



给定一个二叉树 `root` ，返回其最大深度。

二叉树的 **最大深度** 是指从根节点到最远叶子节点的最长路径上的节点数。

 

**示例 1：**

![img](https://assets.leetcode.com/uploads/2020/11/26/tmp-tree.jpg)

 

```
输入：root = [3,9,20,null,null,15,7]
输出：3
```

**示例 2：**

```
输入：root = [1,null,2]
输出：2
```

 

````python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    def maxDepth(self, root: Optional[TreeNode]) -> int:
        if not root:
            return 0
        left_depth = self.maxDepth(root.left)
        right_depth = self.maxDepth(root.right)

        return max(left_depth, right_depth) + 1
````

**💡 解题思路：递归（深度优先搜索 DFS）**

要计算一棵二叉树的最大深度，我们可以采用 **分治思想**：

- 如果当前节点为空（`None`），说明没有节点，深度为 0。
- 否则，分别求出左子树和右子树的最大深度。
- 当前节点的最大深度 = `max(左子树深度, 右子树深度) + 1`（+1 表示当前这个节点本身）。

这本质上是一种 **后序遍历（post-order traversal）**：先处理左右子树，再处理当前节点。


**🧠 算法步骤**

1. **递归终止条件**：若 `root == None`，返回 0。
2. **递归调用**：
   - 计算左子树的最大深度：`left_depth = maxDepth(root.left)`
   - 计算右子树的最大深度：`right_depth = maxDepth(root.right)`
3. **合并结果**：返回 `max(left_depth, right_depth) + 1`

**🧪 代码实现（Python）**

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right

class Solution:
    def maxDepth(self, root: Optional[TreeNode]) -> int:
        # 1. 终止条件：如果节点为空，深度为 0
        if not root:
            return 0
        
        # 2. 递归调用：分别去问左孩子和右孩子
        left_depth = self.maxDepth(root.left)
        right_depth = self.maxDepth(root.right)
        
        # 3. 处理逻辑：取最大值 + 1（加的是当前这个节点本身）
        return max(left_depth, right_depth) + 1
```



**⏱️ 复杂度分析**

- **时间复杂度**：`O(n)`  
  每个节点都会被访问一次，其中 `n` 是二叉树中的节点总数。

- **空间复杂度**：`O(h)`  
  递归调用栈的深度等于树的高度 `h`。最坏情况下（树退化为链表），`h = n`；最好情况下（完全平衡树），`h = log n`。

🧩 扩展思考

- 如果要求 **最小深度**（即到最近叶子节点的路径长度），逻辑类似但需注意：**必须到达叶子节点**（左右子树都为空才算叶子）。
- 也可以使用 **层序遍历（BFS）** 实现，每遍历一层深度 +1，适合面试中展示多种解法。

## [105. 从前序与中序遍历序列构造二叉树](https://leetcode.cn/problems/construct-binary-tree-from-preorder-and-inorder-traversal/)

中等  相关标签  ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业



给定两个整数数组 `preorder` 和 `inorder` ，其中 `preorder` 是二叉树的**先序遍历**， `inorder` 是同一棵树的**中序遍历**，请构造二叉树并返回其根节点。

 

**示例 1:**

![img](https://assets.leetcode.com/uploads/2021/02/19/tree.jpg)

```
输入: preorder = [3,9,20,15,7], inorder = [9,3,15,20,7]
输出: [3,9,20,null,null,15,7]
```

**示例 2:**

```
输入: preorder = [-1], inorder = [-1]
输出: [-1]
```

````python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right

class Solution:
    def buildTree(self, preorder: List[int], inorder: List[int]) -> Optional[TreeNode]:
        # 1. Base Case: 没牌了，返回空
        if not preorder or not inorder:
            return None
        
        # 2. 只有一张牌，直接生成节点返回（可选的优化，其实不写也能跑通）
        if len(preorder) == 1:
            return TreeNode(preorder[0])

        # 3. 核心逻辑
        root_val = preorder[0]
        root = TreeNode(root_val)
        
        # 在 inorder 中找到切分点
        # ⚠️ 注意：index() 操作是 O(N) 的，这会导致整体复杂度上升
        mid_index = inorder.index(root_val)
        
        # 4. 递归构建 (Slicing)
        # 左子树：
        # Preorder: 跳过根，取 mid_index 长度
        # Inorder: 从头取到 mid_index
        root.left = self.buildTree(preorder[1 : 1+mid_index], inorder[0 : mid_index])
        
        # 右子树：
        # Preorder: 跳过根+左子树，取剩下所有
        # Inorder: 从 mid_index+1 取到最后
        root.right = self.buildTree(preorder[1+mid_index : ], inorder[mid_index+1 : ])
        
        return root
````

给你二叉树的根节点 `root` ，返回其节点值 **自底向上的层序遍历** 。 （即按从叶子节点所在层到根节点所在的层，逐层从左向右遍历）

 

**示例 1：**

![img](https://assets.leetcode.com/uploads/2021/02/19/tree1.jpg)

```
输入：root = [3,9,20,null,null,15,7]
输出：[[15,7],[9,20],[3]]
```

```
输入：root = [1]
输出：[[1]]
```

**示例 3：**

```

输出：[]
```

````python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def levelOrderBottom(self, root: Optional[TreeNode]) -> List[List[int]]:
        if not root:
            return []
        queue = collections.deque([root])
        result = []
        while queue:
            level = []
            for _ in range(len(queue)):
                cur = queue.popleft()
                level.append(cur.val)
                if cur.left:
                    queue.append(cur.left)
                if cur.right:
                    queue.append(cur.right)
            result.append(level)
        return result[::-1]
````

## [112. 路径总和](https://leetcode.cn/problems/path-sum/)

简单  相关标签  ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业



给你二叉树的根节点 `root` 和一个表示目标和的整数 `targetSum` 。判断该树中是否存在 **根节点到叶子节点** 的路径，这条路径上所有节点值相加等于目标和 `targetSum` 。如果存在，返回 `true` ；否则，返回 `false` 。

**叶子节点** 是指没有子节点的节点。

 

**示例 1：**

![img](https://assets.leetcode.com/uploads/2021/01/18/pathsum1.jpg)

```
输入：root = [5,4,8,11,null,13,4,7,2,null,null,null,1], targetSum = 22
输出：true
解释：等于目标和的根节点到叶节点路径如上图所示。
```

**示例 2：**

![img](https://assets.leetcode.com/uploads/2021/01/18/pathsum2.jpg)

```
输入：root = [1,2,3], targetSum = 5
输出：false
解释：树中存在两条根节点到叶子节点的路径：
(1 --> 2): 和为 3
(1 --> 3): 和为 4
不存在 sum = 5 的根节点到叶子节点的路径。
```

**示例 3：**

```
输入：root = [], targetSum = 0
输出：false
解释：由于树是空的，所以不存在根节点到叶子节点的路径。
```

````python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    def hasPathSum(self, root, targetSum):
        # 第一步：如果树是空的，直接返回 False
        if root is None:
            return False

        # 第二步：看看当前节点是不是叶子节点
        # 叶子节点就是：左边没有孩子，右边也没有孩子
        is_leaf = (root.left is None) and (root.right is None)

        # 如果是叶子节点
        if is_leaf:
            # 那我们就检查：当前节点的值是不是正好等于 targetSum？
            if root.val == targetSum:
                return True
            else:
                return False

        # 第三步：如果不是叶子节点，说明还能继续往下走
        # 那我们就先把当前节点的值从目标中减掉
        remaining = targetSum - root.val

        # 然后分别去左边和右边找
        left_has_path = self.hasPathSum(root.left, remaining)
        right_has_path = self.hasPathSum(root.right, remaining)

        # 只要左边或者右边有一条路满足条件，就返回 True
        if left_has_path or right_has_path:
            return True
        else:
            return False
````

**1. 核心思路：减法递归 (Subtraction & Recursion)**
这道题要求判断是否存在一条从**根节点**到**叶子节点**的路径，其和等于 `targetSum`。

与其从上往下做“加法”（维护一个 `current_sum` 累加），不如使用**“减法思维”**：
1.  每经过一个节点，就用目标值 `targetSum` **减去** 当前节点的值。
2.  带着剩下的残差（`newTarget`）继续往下走。
3.  **终极判断**：当你走到**叶子节点**时，如果剩下的目标值**刚好等于**当前节点的值（即减完归零），说明找到了路径。

**2. 代码解析**

```python
class Solution:
    def hasPathSum(self, root: Optional[TreeNode], targetSum: int) -> bool:
        # 1. 兜底逻辑 (Base Case)
        # 如果走到了空节点（掉下悬崖），说明这条路肯定不通
        # 注意：即使 targetSum 为 0，空节点也不算路径
        if not root:
            return False
        
        # 2. 叶子节点审判 (Leaf Node Check)
        # 只有在叶子节点（无左且无右），我们才有资格判断路径是否结束
        # 此时检查：手里剩下的 targetSum 是否刚好够支付当前节点的过路费
        if not root.left and not root.right:
            return targetSum == root.val
        
        # 3. 递归甩锅 (Recursive Step)
        # 如果不是叶子节点，就把剩下的“债”算出来
        newTarget = targetSum - root.val
        
        # 只要左边能通，或者右边能通，整条路就算通 (OR 逻辑)
        return self.hasPathSum(root.left, newTarget) or self.hasPathSum(root.right, newTarget)

```

**3. 图解演示**

假设 `targetSum = 22`，路径为 `5 -> 4 -> 11 -> 2`：

1. **Root (5)**: 目标变为 `22 - 5 = 17`，传给孩子。
2. **Node (4)**: 目标变为 `17 - 4 = 13`，传给孩子。
3. **Node (11)**: 目标变为 `13 - 11 = 2`，传给孩子。
4. **Leaf (2)**: 它是叶子！检查 `targetSum (2) == val (2)`。
* **相等！** 返回 `True`。
* 信号逐层向上传递，最终根节点返回 `True`。



**4. 复杂度分析**

* **时间复杂度**: 。最坏情况下（例如树是线性的或者没有匹配路径），我们需要遍历每一个节点。
* **空间复杂度**: 。递归调用栈的深度等于树的高度。

**5. 面试避坑点**

* **不要漏掉 `if not root**`: 空树或者递归进入 `None` 节点时，必须返回 `False`。
* **必须是“到叶子节点”**: 路径不能停在半路。代码中 `if not root.left and not root.right` 就是在确保这一点。

## 128. 最长连续序列

给定一个未排序的整数数组 `nums` ，找出数字连续的最长序列（不要求序列元素在原数组中连续）的长度。

请你设计并实现时间复杂度为 `O(n)` 的算法解决此问题。

 

**示例 1：**

```
输入：nums = [100,4,200,1,3,2]
输出：4
解释：最长数字连续序列是 [1, 2, 3, 4]。它的长度为 4。
```

**示例 2：**

```
输入：nums = [0,3,7,2,5,8,4,6,0,1]
输出：9
```

**示例 3：**

```
输入：nums = [1,0,1,2]
输出：3
```

````python
class Solution:
    def longestConsecutive(self, nums: List[int]) -> int:
        num_set = set(nums)
        max_length = 0
        
        for num in num_set:
            if (num - 1) not in num_set:
                current_num = num
                current_length = 1
                
                while current_num + 1 in num_set:
                    current_num += 1
                    current_length += 1
                
                max_length = max(max_length, current_length)
        
        return max_length
````

**问题理解**
我们需要在一组无序的数字中，找到所有连续数字构成的序列（如2,3,4,5这样的序列），并找到其中最长的那个序列的长度。

例如：若输入是`[100,4,200,1,3,2]`，最长的连续序列是`[1,2,3,4]`，长度为4。

**核心思路**
1. **为什么要用集合（set）？**
   - 集合可以让我们在O(1)时间内判断一个元素是否存在。
   - 例如：快速查找某个数字的下一个数字是否存在。

2. **如何找到连续序列的起点？**
   - 如果当前数字`x`的前一个数字`x-1`不在集合中，说明`x`是一个连续序列的起点（因为没有更小的前驱）。
   - 例如：数字**1**的前驱**0**不在集合中，所以**1**是某段连续序列的起点。

3. **如何扩展序列？**
   - 从起点`x`开始，依次检查`x+1`、`x+2`...是否存在，直到中断为止。
   - 例如：从**1**出发，找到**2**、**3**、4，直到无法继续。


**代码逐行解释**

```python
class Solution:
    def longestConsecutive(self, nums: List[int]) -> int:
        # 创建一个集合，存储所有数字（自动去重）
        num_set = set(nums)
        max_length = 0  # 记录最长序列的长度
      
        for num in num_set:  # 遍历集合中的每个数字
            if (num - 1) not in num_set:  # 检查是否是连续序列的起点
                current_num = num
                current_length = 1  # 至少包括当前数字
              
                # 向后查找下一个连续数字
                while current_num + 1 in num_set:
                    current_num += 1
                    current_length += 1
              
                # 更新最大长度
                max_length = max(max_length, current_length)
      
        return max_length
```


 **详细分步解析**

**集合的作用**
```python
num_set = set(nums)
```
- 将输入的列表`nums`转换为集合`num_set`。
- 集合的特性：自动去重（重复元素不影响结果），并且查找元素的时间复杂度是O(1)。

**遍历集合中的每个元素**
```python
for num in num_set:
```
- 遍历集合`num_set`中的每个数字（不重复）。
- 为什么不遍历原列表`nums`？因为重复元素会影响效率，而集合去重了。

**检查是否是连续序列的起点**
```python
if (num - 1) not in num_set:
```
- 如果当前数字`num`的前一个数字`num-1`不在集合中，说明`num`是某个连续序列的“起点”。
- 例如：若数字是**1**，且`0`不在集合中，那么**1**会成为起点。

**从起点扩展连续序列**
```python
current_num = num
current_length = 1  # 至少包含当前的num

while current_num + 1 in num_set:
    current_num += 1
    current_length += 1
```
- 从起点`num`开始，向后查找`num+1`、`num+2`...是否在集合中。
- 每找到一个连续数字，`current_length`加1。
- 例如：从**1**出发，找到**2**、**3**、**4**，直到遇到不在集合中的数字（比如**5`结束，此时`current_length=4`）。

**更新最大长度**
```python
max_length = max(max_length, current_length)
```
- 每次找到一个连续序列后，比较当前序列长度和已记录的最大长度，更新最大值。


**示例模拟**
以输入`[100,4,200,1,3,2]`为例：
1. 转换为集合：`{1,2,3,4,100,200}`。
2. 遍历到**1**时：
   - `1-1=0`不在集合中 → 是起点。
   - 向后找到**2**、**3**、4 → `current_length=4`。
3. 其他元素（如4,200,100）会被跳过，因为它们不是起点（例如4-1=3在集合中）。
4. 最终最长长度是4。


**关键点**
- **时间复杂度O(n)**：每个元素最多被访问两次（一次作为起点，一次被其他序列包含）。
- **空间复杂度O(n)**：需要一个集合存储所有元素。

## [139. 单词拆分](https://leetcode.cn/problems/word-break/)

中等   相关标签   ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业


给你一个字符串 `s` 和一个字符串列表 `wordDict` 作为字典。如果可以利用字典中出现的一个或多个单词拼接出 `s` 则返回 `true`。

**注意：**不要求字典中出现的单词全部都使用，并且字典中的单词可以重复使用。
 

**示例 1：**

```
输入: s = "leetcode", wordDict = ["leet", "code"]
输出: true
解释: 返回 true 因为 "leetcode" 可以由 "leet" 和 "code" 拼接成。
```

**示例 2：**

```
输入: s = "applepenapple", wordDict = ["apple", "pen"]
输出: true
解释: 返回 true 因为 "applepenapple" 可以由 "apple" "pen" "apple" 拼接成。
     注意，你可以重复使用字典中的单词。
```

**示例 3：**

```
输入: s = "catsandog", wordDict = ["cats", "dog", "sand", "and", "cat"]
输出: false
```

````python
class Solution:
    def wordBreak(self, s: str, wordDict: List[str]) -> bool:
        word_set = set(wordDict)  # 转为 set 加快查找
        n = len(s)
        dp = [False] * (n + 1)
        dp[0] = True  # 空字符串可以被拆分
        
        for i in range(1, n + 1):
            for j in range(i):
                if dp[j] and s[j:i] in word_set:
                    dp[i] = True
                    break  # 找到一个合法分割就够了，提前退出
        return dp[n]
````

**核心逻辑：接力赛隐喻 (The Algorithm)**

想象字符串 `s` 是一条跑道，`wordDict` 是你能用的交通工具（字典里的词）。
我们要判断：**能不能从起点走到终点？**

  * **问题**: 字符串 `s` 能被拆分吗？
  * **转化**: 我能不能先走到中间某个点 `j`，然后从 `j` 搭乘一个“单词”坐车到终点 `i`？

**这里的 `dp` 数组是什么？**

`dp[i]` 是一个布尔值（True/False）。

  * **含义**: 字符串的前 `i` 个字符（即 `s[0...i-1]`），能否被完美拆分？
  * **目标**: 求 `dp[n]` （n 是字符串总长度，即整条跑道能不能跑完）。

**核心推导公式 (State Transition)**

要让 `dp[i]` 变 True（能跑到 `i` 点），必须满足两个条件：

1.  **前置条件**: 存在一个中间点 `j` (\< `i`)，使得 `dp[j]` 是 True。（意味着：有人能成功跑到 `j` 点接力给我）。
2.  **当前条件**: 从 `j` 到 `i` 的这段子串 `s[j:i]`，刚好在字典里。（意味着：我有合法的车能从 `j` 开到 `i`）。

**代码逐行深度解析 (Syntax & Logic)**

```python
class Solution:
    def wordBreak(self, s: str, wordDict: List[str]) -> bool:
        # [语法点 1] 为什么要转 set?
        # List 的查找复杂度是 O(N)，Set (哈希表) 是 O(1)。
        # 如果字典很大，这行代码能让速度快几百倍。面试必写！
        word_set = set(wordDict) 
        
        n = len(s)
        
        # [语法点 2] 初始化 DP 表
        # 长度为什么是 n + 1？
        # 因为我们要计算 "前0个字符", "前1个字符" ... "前n个字符"。
        # dp[i] 对应的是 s 的前 i 个字符。
        dp = [False] * (n + 1)
        
        # [算法点 1] Base Case (初始状态)
        # dp[0] 代表 "前0个字符"（空串）。
        # 逻辑上，空串不需要拆分也是合法的。它是接力赛的起点。
        dp[0] = True 
        
        # [循环结构] 
        # i 代表我们现在想验证的 "终点" 位置 (从第1个字符走到第n个字符)
        for i in range(1, n + 1):
            
            # j 代表 "切割点" (从 0 走到 i-1)
            # 我们在找：有没有一个 j，既能从起点跑到 j，又能从 j 跳到 i？
            for j in range(i):
                
                # [核心逻辑 Check]
                # 1. dp[j]: 前 j 个字符能不能拼出来？ (接力棒在 j 这里吗？)
                # 2. s[j:i]: 从 j 到 i 的这段字符串，是不是一个单词？
                # [语法点 3] 切片 s[j:i]
                # Python 切片是 "左闭右开" [j, i)，刚好取出从 index j 到 i-1 的字符。
                if dp[j] and s[j:i] in word_set:
                    
                    dp[i] = True
                    
                    # [算法点 2] 剪枝 (Pruning)
                    # 只要找到一种切法能到 i，那就说明 i 是可达的。
                    # 不需要再试别的 j 了，直接退出内层循环，去算 i+1。
                    # 这能极大节省时间。
                    break 
        
        # 返回我们要的最终结果：前 n 个字符（整句）能不能拆分
        return dp[n]
```

2.  **i = 5, 6, 7**:

      * 找不到合法的 `j` 和单词，`dp[5], dp[6], dp[7]` 都是 False。

      * 程序尝试 `j = 0` -\> `s[0:8]` "leetcode" 不在字典。
      * ...
      * 程序尝试 **`j = 4`**:
          * `dp[4]` 是 True (我们在步骤1算出来的！)。
          * `s[4:8]` 是 "code" -\> **在字典里**。
          * **Hit\!** 两个条件满足。
      * **结果**: `dp[8] = True`。

## [141. 环形链表](https://leetcode.cn/problems/linked-list-cycle/)

简单   相关标签   ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业

给你一个链表的头节点 `head` ，判断链表中是否有环。

如果链表中有某个节点，可以通过连续跟踪 `next` 指针再次到达，则链表中存在环。 为了表示给定链表中的环，评测系统内部使用整数 `pos` 来表示链表尾连接到链表中的位置（索引从 0 开始）。**注意：`pos` 不作为参数进行传递** 。仅仅是为了标识链表的实际情况。

*如果链表中存在环* ，则返回 `true` 。 否则，返回 `false` 。

 

**示例 1：**

![img](https://assets.leetcode-cn.com/aliyun-lc-upload/uploads/2018/12/07/circularlinkedlist.png)

```
输入：head = [3,2,0,-4], pos = 1
输出：true
解释：链表中有一个环，其尾部连接到第二个节点。
```

**示例 2：**

![img](https://assets.leetcode-cn.com/aliyun-lc-upload/uploads/2018/12/07/circularlinkedlist_test2.png)

```
输入：head = [1,2], pos = 0
输出：true
解释：链表中有一个环，其尾部连接到第一个节点。
```

**示例 3：**

![img](https://assets.leetcode-cn.com/aliyun-lc-upload/uploads/2018/12/07/circularlinkedlist_test3.png)

```
输入：head = [1], pos = -1
输出：false
解释：链表中没有环。
```

````python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

class Solution:
    def hasCycle(self, head: Optional[ListNode]) -> bool:
        if not head or not head.next:
            return False
        
        slow = head
        fast = head.next
        
        while slow != fast:
            # 如果 fast 跑到头了，说明没环
            if not fast or not fast.next:
                return False
            
            slow = slow.next      # 慢跑 1 步
            fast = fast.next.next # 快跑 2 步
            
        return True
````

**🎯 题目背景：判断链表是否有环**

首先，什么是“链表有环”？

- 正常链表：`1 → 2 → 3 → 4 → None`（最后指向空）
- 有环链表：`1 → 2 → 3 → 4 → 2 → 3 → 4 → 2...`（某个节点指回前面，形成“圈”）

题目要求：**给一个链表的头节点 `head`，判断它有没有环。**

---

**🧠 算法思想：快慢指针（Floyd 判圈算法）**

这是本题最经典的解法，也叫 **“龟兔赛跑”**：

- 想象两个人在跑道上跑步：
  - **乌龟（slow）**：每次走 1 步
  - **兔子（fast）**：每次走 2 步
- 如果跑道是**直的（无环）**，兔子会先跑到终点（`None`），比赛结束。
- 如果跑道是**环形的（有环）**，兔子最终会从后面追上乌龟！

> ✅ 关键结论：**如果有环，快慢指针一定会相遇；如果没有环，快指针会先到终点。**

**🔍 逐行解释你的代码**

```python
class Solution:
    def hasCycle(self, head: Optional[ListNode]) -> bool:
```

- 定义一个类 `Solution`，里面有个方法 `hasCycle`
- 参数 `head` 是链表的头节点，类型是 `Optional[ListNode]`（意思是：可能是 `ListNode`，也可能是 `None`）
- 返回值是 `bool`（`True` 或 `False`）

> 💡 `Optional[T]` = `Union[T, None]`，就是“可以是 T，也可以是 None”

---

```python
        if not head or not head.next:
            return False
```

- **边界情况处理**：
  - 如果链表是空的（`head is None`），肯定没环 → `False`
  - 如果只有一个节点（`head.next is None`），也构不成环 → `False`

✅ 这一步很关键，避免后面访问 `head.next.next` 出错。

---

```python
        slow = head
        fast = head.next
```

- 初始化两个指针：
  - `slow` 从头开始（第 1 个节点）
  - `fast` 从第二个节点开始（`head.next`）

> 🤔 为什么 `fast` 不也从 `head` 开始？
> 其实也可以！但这里让 `fast` 先走一步，是为了让 `while` 循环条件 `slow != fast` 在一开始就能进入（否则第一次就相等，直接返回 `True`，错误！）

---

```python
        while slow != fast:
```

- 只要快慢指针**还没相遇**，就继续跑
- 一旦相遇，说明有环，跳出循环，返回 `True`

---

```python
            if not fast or not fast.next:
                return False
```

- 在每一步移动前，先检查 `fast` 是否已经跑到尽头：
  - `fast is None` → 到头了
  - `fast.next is None` → 下一步就到头了
- 只要有一个成立，说明**没有环**，直接返回 `False`

> ⚠️ 注意顺序：必须先检查 `fast`，再检查 `fast.next`，否则如果 `fast` 是 `None`，访问 `fast.next` 会报错！


```python
            slow = slow.next      # 慢指针走 1 步
            fast = fast.next.next # 快指针走 2 步
```

- 移动指针：
  - `slow` 向前一步
  - `fast` 向前两步（先到 `next`，再到 `next.next`）

```python
        return True
```

- 如果退出了 `while` 循环，说明 `slow == fast`，即**相遇了**
- 相遇 ⇒ 有环 ⇒ 返回 `True`

---

**🧪 举个例子演示**

**无环链表：`1 → 2 → 3 → None`**

```
初始：slow=1, fast=2
第1轮：slow=2, fast=3.next=None → 检查发现 fast.next 不存在 → return False
```

**有环链表：`1 → 2 → 3 → 4 → 2（回到2）`**

```
初始：slow=1, fast=2
第1轮：slow=2, fast=3
第2轮：slow=3, fast=2
第3轮：slow=4, fast=4 ← 相遇！return True
```

---

**✅ 为什么这个算法是对的？**

| 情况 | 结果 |
|------|------|
| 有环 | 快指针在环里“绕圈”，速度比慢指针快，一定会追上 |
| 无环 | 快指针先到达 `None`，提前退出 |

而且：
- **时间复杂度**：O(n) —— 最多走一圈
- **空间复杂度**：O(1) —— 只用了两个指针，没开额外空间

---

**📌 小贴士（常见疑问）**

**Q1：为什么快指针走 2 步，不是 3 步或更多？**
- 走 2 步是最小的“能保证追上”的速度差。
- 走太多步反而可能跳过相遇点（虽然理论上也能检测，但没必要）。

**Q2：能不能都从 `head` 开始？**
可以！但要改写成 `do-while` 形式（Python 没有），所以常用技巧是：
```python
slow = fast = head
while fast and fast.next:
    slow = slow.next
    fast = fast.next.next
    if slow == fast:
        return True
return False
```
这也是常见写法，和你的一样正确！

````python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def detectCycle(self, head: Optional[ListNode]) -> Optional[ListNode]:
        fast, slow = head, head
        while fast and fast.next:
            fast = fast.next.next
            slow = slow.next
            if fast == slow:
                p = slow
                q = head
                while p != q:
                    p = p.next
                    q = q.next
                return p
        return None
````

**核心问题：检测链表中的环并找到环的入口**

想象一下一个链表，就像一串糖葫芦，每个糖葫芦（节点）都有一个指针指向下一个糖葫芦。通常，链表的最后一个糖葫芦会指向 None，表示链表结束。但是，如果最后一个糖葫芦或者中间的某个糖葫芦指向了链表前面已经存在的某个糖葫芦，那么这个链表就形成了一个“环”，就像一个闭合的圈。

这段代码的目的就是找出链表里有没有这样的环，如果有，找出环的第一个节点是哪个。

**算法：快慢指针法 (Floyd's Cycle-Finding Algorithm)**

这段代码使用了著名的快慢指针算法来解决这个问题。这个算法非常巧妙，它的核心思想是：

1.  **两个指针，一个快一个慢：** 我们创建两个指针，一个叫做 `fast` (快指针)，一个叫做 `slow` (慢指针)。
2.  **慢指针一步步走，快指针两步两步走：** 在链表中移动指针时，慢指针每次移动一步（指向下一个节点），快指针每次移动两步（指向下下个节点）。
3.  **相遇即有环：** 如果链表中存在环，那么快指针最终一定会追上慢指针，它们会在环内的某个节点相遇。这就像在操场上跑步，你跑一圈，我跑两圈，如果你不动，我肯定会追上你。在环里，快指针总会缩短与慢指针的距离，最终会相遇。
4.  **相遇后找到环入口：** 相遇后，我们知道链表里有环，但相遇的地方不一定是环的入口。为了找到环的入口，我们再使用两个指针：一个仍然留在相遇的地方（我们用 `p` 来表示，初始化为 `slow` 或 `fast`），另一个从链表的头部开始（我们用 `q` 来表示，初始化为 `head`）。然后，这两个指针每次都只移动一步，当它们再次相遇时，相遇的节点就是环的入口。

**为什么相遇后从头部开始走也能找到环入口？**

这是一个数学上的结论，可以证明：从链表头部到环入口的距离，等于从相遇点到环入口的距离。

假设：
*   链表头部到环入口的距离是 `L`。
*   环的长度是 `C`。
*   快慢指针第一次相遇时，慢指针在环内走了 `K` 步。

当快慢指针相遇时，慢指针走了 `L + K` 步，快指针走了 `L + K + n * C` 步 (其中 `n` 是快指针在环里比慢指针多跑的圈数)。
因为快指针的速度是慢指针的两倍，所以快指针走的步数是慢指针的两倍：
`L + K + n * C = 2 * (L + K)`
`L + K + n * C = 2L + 2K`
`n * C - K = L`

这个公式告诉我们，从链表头部到环入口的距离 `L`，等于从相遇点沿着环继续走 `n * C - K` 步。由于 `-K` 表示向后走 `K` 步，而 `n * C` 表示在环里绕了 `n` 圈，所以 `n * C - K` 实际上就是从相遇点再走 `L` 步就会回到环入口。

因此，当一个指针从头部开始，另一个指针从相遇点开始，它们都一步步走，当它们相遇时，相遇点就是环的入口。

**代码语法和逐行解释：**

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def detectCycle(self, head: Optional[ListNode]) -> Optional[ListNode]:
        # 'head' 是链表的第一个节点。Optional[ListNode] 表示 head 可以是 ListNode 对象，也可以是 None。
        # 如果链表为空，head 就是 None。

        fast, slow = head, head
        # 初始化快指针 'fast' 和慢指针 'slow'，它们都指向链表的头部。

        while fast and fast.next:
            # 这是一个循环，条件是 'fast' 指针不为 None，并且 'fast.next' 也不为 None。
            # 'fast' 指针需要移动两步，所以如果 fast 或者 fast.next 是 None，说明链表已经结束了，没有环。
            # 'and' 是逻辑运算符，表示两个条件都必须为 True。

            fast = fast.next.next
            # 快指针向前移动两步。首先移动到 fast.next，然后再移动到 (fast.next).next。

            slow = slow.next
            # 慢指针向前移动一步。移动到 slow.next。

            if fast == slow:
                # 检查快指针和慢指针是否指向同一个节点。
                # 如果它们相遇了，说明链表中存在环。

                p = slow
                # 初始化指针 'p'，指向相遇的节点。

                q = head
                # 初始化指针 'q'，指向链表的头部。

                while p != q:
                    # 这是一个新的循环，条件是 'p' 指针和 'q' 指针不指向同一个节点。

                    p = p.next
                    # 指针 'p' 向前移动一步。

                    q = q.next
                    # 指针 'q' 向前移动一步。

                return p
                # 当 'p' 和 'q' 相遇时，它们指向的节点就是环的入口。
                # 我们返回这个节点。

        return None
        # 如果循环结束了（也就是 while fast and fast.next 条件不再满足），
        # 并且快慢指针没有相遇，说明链表中没有环。
        # 此时，我们返回 None。
```

**代码中的类和注释：**

*   `# Definition for singly-linked list.` 和下面的 `class ListNode:` 是一段注释和类的定义。它定义了链表中节点的结构：每个节点有一个 `val` 属性存储值，一个 `next` 属性指向下一个节点。这是常见的链表节点定义方式，不是解决环问题的核心代码，只是为了表示链表的结构。
*   `class Solution:` 是 LeetCode 等在线编程平台常用的格式，表示一个解决方案类。
*   `def detectCycle(self, head: Optional[ListNode]) -> Optional[ListNode]:` 是类的成员方法。
    *   `self` 是 Python 类方法中必须的第一个参数，表示方法所属的对象。
    *   `head: Optional[ListNode]` 是方法的参数。`head` 是链表的头节点。`Optional[ListNode]` 是 Python 的类型提示，表示 `head` 可以是一个 `ListNode` 对象，也可以是 `None`。这提高了代码的可读性。
    *   `-> Optional[ListNode]` 是方法的返回值类型提示，表示这个方法返回一个 `ListNode` 对象（如果找到环入口）或者 `None`（如果链表中没有环）。

**总结：**

这段代码通过快慢指针的巧妙移动，首先判断链表中是否存在环。如果存在，快慢指针会在环内相遇。然后，利用从头部开始的指针和从相遇点开始的指针同步移动，最终找到环的入口节点。如果快指针到达链表末尾都没有与慢指针相遇，则说明链表中没有环。

这个算法的时间复杂度是 O(n)，其中 n 是链表的长度，因为每个节点最多会被访问几次。空间复杂度是 O(1)，因为它只使用了几个额外的指针变量，不依赖于链表的长度。

**前序遍历（Preorder Traversal）题解**

* **思路：** 前序遍历的核心思想是“先访问当前节点，再访问左右子树”。这自然地契合了递归的结构。
* **递归步骤：**
    1. **访问（处理）当前节点：** 将当前节点的值添加到结果列表中。
    2. **递归遍历左子树：** 对当前节点的左孩子调用前序遍历函数。
    3. **递归遍历右子树：** 对当前节点的右孩子调用前序遍历函数。
* **终止条件：** 当遇到空节点（`None`）时，停止递归。
* **代码结构（递归）：**

```python
class Solution:
    def preorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        res = [] # 存储结果的列表

        def dfs(node): # 递归辅助函数
            if node is None: # 终止条件
                return

            res.append(node.val) # 访问当前节点

            dfs(node.left) # 递归遍历左子树
            dfs(node.right) # 递归遍历右子树

        dfs(root) # 从根节点开始调用
        return res
```

* **特点：** 这种递归方式非常直观，代码简洁易懂。它利用了函数调用栈来隐式地处理节点的访问顺序。

**后序遍历（Postorder Traversal）题解**

* **思路：** 后序遍历的核心思想是“先访问左右子树，最后访问当前节点”。与前序遍历相比，只是访问当前节点的位置不同。
* **递归步骤：**
    1. **递归遍历左子树：** 对当前节点的左孩子调用后序遍历函数。
    2. **递归遍历右子树：** 对当前节点的右孩子调用后序遍历函数。
    3. **访问（处理）当前节点：** 将当前节点的值添加到结果列表中。
* **终止条件：** 当遇到空节点（`None`）时，停止递归。
* **代码结构（递归）：**

```python
class Solution:
    def postorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        res = [] # 存储结果的列表

        def dfs(node): # 递归辅助函数
            if node is None: # 终止条件
                return

            dfs(node.left) # 递归遍历左子树
            dfs(node.right) # 递归遍历右子树

            res.append(node.val) # 访问当前节点

        dfs(root) # 从根节点开始调用
        return res
```

* **特点：** 后序遍历在某些场景下非常有用，例如在删除二叉树时，需要先删除子节点才能删除父节点。

**前序遍历和后序遍历的对比：**

主要的区别在于**访问当前节点的位置**：

* **前序遍历：** **当前节点** -> 左子树 -> 右子树
* **后序遍历：** 左子树 -> 右子树 -> **当前节点**

这两种遍历方式都是二叉树的**深度优先搜索（DFS）** 的应用。它们都通过递归深入到树的最深处，然后再逐层返回。

你的代码完全体现了这两种遍历方式的递归实现。非常棒！

如果你想了解更多关于二叉树遍历的知识，比如：

* **中序遍历（Inorder Traversal）：** 左子树 -> 当前节点 -> 右子树
* **迭代（非递归）实现这三种遍历方式**
* **层序遍历（Level Order Traversal）：** 按层级从左到右访问节点

## [150. 逆波兰表达式求值](https://leetcode.cn/problems/evaluate-reverse-polish-notation/)

````python
from typing import List

class Solution:
    def evalRPN(self, tokens: List[str]) -> int:
        stack = []
        for token in tokens:
            if token not in ["+", "-", "*", "/"]:
                stack.append(int(token))
            else:
                num2 = stack.pop()
                num1 = stack.pop()

                if token == "+":
                    result = num1 + num2
                elif token == "-":
                    result = num1 - num2
                elif token == "*":
                    result = num1 * num2
                elif token == "/":
                    result = int(num1 / num2)

                stack.append(result)
        finres = stack.pop()
        return finres
solution = Solution()
input_string = input("请输入tokens：")
input_tokens = input_string.split()
output_no = solution.evalRPN(input_tokens)
print(f"该算式最后的结果为：{output_no}")
````

**算法思想：使用栈**

计算逆波兰表达式最直观和常用的方法是利用 **栈 (Stack)** 数据结构。栈具有“后进先出 (LIFO)”的特性，这非常适合处理逆波兰表达式的计算顺序。

算法步骤如下：

1.  **初始化一个空栈。** 这个栈用来临时存储遇到的数字和中间计算结果。
2.  **遍历输入的逆波兰表达式（字符串数组）中的每一个元素。**
3.  **判断当前元素是数字还是运算符：**
    *   **如果是数字：** 将该数字从字符串类型转换为整数类型，然后将其压入栈中。
    *   **如果是运算符 (+, -, \*, /)：** 从栈中依次弹出两个元素。**注意，先弹出的元素是右操作数 (num2)，后弹出的元素是左操作数 (num1)。** 根据运算符进行相应的数学运算（加、减、乘、除）。
    *   **进行运算：** 对于加法、减法、乘法，直接进行运算。对于除法，需要根据题目要求进行 **向零截断** 的整数除法。
    *   **将运算结果压回栈中。**
4.  **遍历完成后：** 当遍历完表达式中的所有元素后，栈中应该只剩下一个元素，这个元素就是整个表达式的计算结果。弹出栈中唯一的元素并返回。

**Python 代码解析**

下面我们逐行解析上面给出的 Python 代码：

```python
from typing import List

class Solution:
    def evalRPN(self, tokens: List[str]) -> int: # 定义函数，接收一个字符串列表，返回一个整数
        stack = [] # 初始化一个空列表作为栈

        for token in tokens: # 遍历输入的 tokens 列表中的每一个元素
            if token not in ["+", "-", "*", "/"]: # 判断当前元素是否为运算符
                # 如果不是运算符，则认为是数字
                stack.append(int(token)) # 将字符串数字转换为整数并压入栈中
            else:
                # 如果是运算符
                num2 = stack.pop() # 弹出栈顶元素，作为右操作数
                num1 = stack.pop() # 再次弹出栈顶元素，作为左操作数

                # 根据运算符进行相应的运算
                if token == "+":
                    result = num1 + num2
                elif token == "-":
                    result = num1 - num2
                elif token == "*":
                    result = num1 * num2
                elif token == "/":
                    # 处理除法：使用 int(num1 / num2) 实现向零截断
                    # Python 3 的 // 是向下取整，对于负数结果不同于向零截断
                    # 例如：-7 // 2 = -4，而 int(-7 / 2) = int(-3.5) = -3
                    result = int(num1 / num2)

                stack.append(result) # 将运算结果压回栈中

        # 遍历结束后，栈中剩下的最后一个元素就是最终结果
        finres = stack.pop()
        return finres

# def evalRPN(self, tokens: List[str]) -> int:: 定义类的方法，self 表示类实例本身
```

**语法要点总结：**

*   **列表作为栈：** 在 Python 中，列表 (`list`) 可以很方便地模拟栈的操作。`append()` 方法将元素添加到列表末尾（入栈），`pop()` 方法移除并返回列表末尾的元素（出栈）。
*   **类型转换：** 使用 `int()` 函数将字符串类型的数字转换为整数类型。
*   **条件判断：** 使用 `if/elif/else` 语句根据运算符执行不同的计算。
*   **列表元素的访问与判断：** 使用 `in` 和 `not in` 运算符判断元素是否在列表中。
*   **函数定义与调用：** 定义类方法 `evalRPN`，并通过创建类的实例来调用它。
*   **字符串的 `split()` 方法：** `split()` 方法是字符串对象的一个非常有用的方法，它可以根据指定的分隔符将字符串分割成子字符串列表。默认情况下，它会根据空格进行分割。
*   **`input()` 函数：** 从标准输入读取一行字符串。
*   **`print()` 函数与 f-string：** 用于向标准输出打印信息，f-string (格式化字符串字面量) 是一种简洁方便的字符串格式化方式。
