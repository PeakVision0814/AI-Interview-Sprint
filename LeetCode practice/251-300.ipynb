## [283. 移动零](https://leetcode.cn/problems/move-zeroes/)

提示



给定一个数组 `nums`，编写一个函数将所有 `0` 移动到数组的末尾，同时保持非零元素的相对顺序。

**请注意** ，必须在不复制数组的情况下原地对数组进行操作。

 

**示例 1:**

```
输入: nums = [0,1,0,3,12]
输出: [1,3,12,0,0]
```

**示例 2:**

```
输入: nums = [0]
输出: [0]
```

````python
from typing import List

class Solution:
    def moveZeroes(self, nums: List[int]) -> None:
        insert_pos = 0
        for i in range(len(nums)):
            if nums[i] != 0:
                nums[insert_pos] = nums[i]
                insert_pos += 1
        for i in range(insert_pos, len(nums)):
            nums[i] = 0
        # 不需要 return

# 使用方式
solution = Solution()
arr = [0, 1, 0, 3, 12]
solution.moveZeroes(arr)
print(f"Move Zeroes Result: {arr}") 
````

**🧩 一、整体目标**

这段代码要解决的问题是：  
> **把一个整数列表（数组）中的所有 0 移到末尾，同时保持非零数字的相对顺序不变。**

比如输入：
```python
[0, 1, 0, 3, 12]
```
输出应该是：
```python
[1, 3, 12, 0, 0]
```

而且题目要求 **不能返回新列表，必须直接修改原来的列表（in-place）**。

**📜 二、代码结构解析（语法角度）**

**1. 导入类型提示**
```python
from typing import List
```
这是 Python 的“类型提示”功能。`List[int]` 表示这个函数接收的是一个**整数列表**。它不会影响程序运行，但能让代码更清晰、更容易被 IDE 或其他程序员理解。

---

**2. 定义类和方法**
```python
class Solution:
    def moveZeroes(self, nums: List[int]) -> None:
```
- `class Solution`: 定义一个叫 `Solution` 的类（就像一个工具箱）。
- `def moveZeroes(...)`: 在这个类里定义一个叫 `moveZeroes` 的方法（函数）。
- `self`: 所有类里的方法第一个参数都要写 `self`，代表这个方法属于这个类的实例（你可以暂时理解为“自己”）。
- `nums: List[int]`: 参数 `nums` 是一个整数列表。
- `-> None`: 表示这个函数**不返回任何值**（但注意：后面代码写了 `return nums`，其实和题意冲突，我们后面会说）。

> ✅ 小贴士：题目说 “Do not return anything”，所以严格来说不应该 `return nums`，但因为 Python 中列表是“可变对象”，修改 `nums` 本身就会改变原始列表，所以即使不 return，外面也能看到变化。

---

**3. 核心逻辑：两步走策略**

**第一步：把所有非零数“紧凑地”放到前面**
```python
insert_pos = 0
for i in range(len(nums)):
    if nums[i] != 0:
        nums[insert_pos] = nums[i]
        insert_pos += 1
```

- `insert_pos`：记录下一个**应该放非零数的位置**，初始是 0（开头）。
- 遍历整个列表（`i` 从 0 到最后）：
  - 如果当前数字 `nums[i]` **不是 0**，就把它放到 `insert_pos` 的位置。
  - 然后 `insert_pos` 往后移一位（准备放下一个非零数）。

📌 举个例子（逐步演示）：

原始：`[0, 1, 0, 3, 12]`

| i | nums[i] | 是否非零？ | 操作                          | insert_pos 变化 |
|---|--------|-----------|-----------------------------|----------------|
| 0 | 0      | 否         | 跳过                         | 0              |
| 1 | 1      | 是         | nums[0] = 1 → `[1, 1, 0, 3, 12]` | 1              |
| 2 | 0      | 否         | 跳过                         | 1              |
| 3 | 3      | 是         | nums[1] = 3 → `[1, 3, 0, 3, 12]` | 2              |
| 4 | 12     | 是         | nums[2] = 12 → `[1, 3, 12, 3, 12]` | 3            |

此时 `insert_pos = 3`，说明前 3 个位置已经放好了所有非零数：`[1, 3, 12, ...]`

> 注意：后面两个位置还是旧数据（3 和 12），但我们马上会覆盖它们！

**第二步：把剩下的位置全部填成 0**
```python
for i in range(insert_pos, len(nums)):
    nums[i] = 0
```

- 从 `insert_pos` 开始，一直到列表末尾，全部设为 0。
- 上例中 `insert_pos = 3`，`len(nums) = 5`，所以 `i = 3, 4`
- 把 `nums[3] = 0`，`nums[4] = 0`

最终结果：`[1, 3, 12, 0, 0]` ✅ 完美！

**4. 返回值问题（小瑕疵）**
```python
return nums
```
虽然代码写了 `return nums`，但函数声明是 `-> None`，而且题目明确说 **不要返回，只修改原数组**。

✅ 正确做法是：**删掉 `return nums` 这一行**，或者至少不打印它的返回值。

但为什么你的代码还能工作？
- 因为 `nums` 是一个**可变对象（列表）**，你在函数内部直接修改了它，所以外面的 `arr` 也被改了。
- 即使你不 return，`print(arr)` 也会显示正确结果。

不过你写了：
```python
print(f"Move Zeroes Result: {solution.moveZeroes(arr)}")
```
这会打印 `None`（因为函数实际没返回东西，或者说返回了但类型不符），但其实 `arr` 已经被修改了！

👉 更好的写法是：
```python
solution.moveZeroes(arr)
print(f"Move Zeroes Result: {arr}")
```

**🔍 三、算法思想总结（核心！）**

这个算法叫做 **“双指针”** 或 **“快慢指针”** 思想：

- `i` 是**快指针**：遍历整个数组，找非零元素。
- `insert_pos` 是**慢指针**：只在遇到非零时才移动，指向下一个要写入的位置。

优点：
- **时间复杂度**：O(n) —— 只遍历两次（其实可以优化成一次，但两次也很高效）。
- **空间复杂度**：O(1) —— 没有使用额外数组，直接在原地修改。

**✅ 四、给小白的建议**

1. **记住这个模式**：遇到“移动元素、保留顺序、原地修改”类问题，优先考虑“双指针”。
2. **列表是可变的**：在函数里改列表，外面会看到变化。
3. **别被 `return` 迷惑**：题目说不返回，就别 return，专注修改输入。
