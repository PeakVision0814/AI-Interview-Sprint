## [867. 转置矩阵](https://leetcode.cn/problems/transpose-matrix/)

提示

给你一个二维整数数组 `matrix`， 返回 `matrix` 的 **转置矩阵** 。

矩阵的 **转置** 是指将矩阵的主对角线翻转，交换矩阵的行索引与列索引。

![img](https://assets.leetcode.com/uploads/2021/02/10/hint_transpose.png)

 

**示例 1：**

```
输入：matrix = [[1,2,3],[4,5,6],[7,8,9]]
输出：[[1,4,7],[2,5,8],[3,6,9]]
```

**示例 2：**

```
输入：matrix = [[1,2,3],[4,5,6]]
输出：[[1,4],[2,5],[3,6]]
```

````python
class Solution:
    def transpose(self, matrix: List[List[int]]) -> List[List[int]]:
        # 1. 获取原矩阵的行数 R 和列数 C
        R, C = len(matrix), len(matrix[0])

        # 2. 创建一个新的结果矩阵
        # 注意：新矩阵的维度是 C 行 R 列 (C x R)
        res = [[0] * R for _ in range(C)]

        # 3. 遍历原矩阵，将元素放入新矩阵的对应位置
        for r in range(R):
            for c in range(C):
                res[c][r] = matrix[r][c]

        return res

solution = Solution()
matrix_inpout = input("请输入矩阵（例如：[[1,2,3],[4,5,6],[7,8,9]]）：")
matrix = eval(matrix_inpout)
transposed_matrix = solution.transpose(matrix)
print("转置后的矩阵为：", transposed_matrix)
````

**第一部分：算法逻辑（核心思想）**

这个算法的核心目标是**根据原矩阵构建一个新的矩阵**，并改变数据的排列方式。

**1. 维度变换 (Shape Transformation)**

  * **原矩阵**：假设是一个长方形，有 $R$ 行，$C$ 列（例如 $2 \times 3$）。
  * **新矩阵**：转置后，形状必须颠倒，变成 $C$ 行，$R$ 列（例如 $3 \times 2$）。
  * **逻辑**：所以我们在第 6 行初始化 `res` 时，外层循环用 `range(C)`（新行数），内层列表长度为 `R`（新列数）。

**2. 坐标映射 (Coordinate Mapping)**

  * **原位置**：`matrix[r][c]` 表示原矩阵第 `r` 行第 `c` 列的元素。
  * **新位置**：转置的定义就是行列互换。所以这个元素应该跑到新矩阵的第 `c` 行第 `r` 列，即 `res[c][r]`。
  * **操作**：在双重循环中，把 `matrix[r][c]` 的值搬运到 `res[c][r]` 去。

**第二部分：Python 语法逐行解析**

这段代码里包含了好几个 Python 面试中常问的语法点：

**函数定义与类型提示 (Type Hinting)**

```python
def transpose(self, matrix: List[List[int]]) -> List[List[int]]:
```

  * **`matrix: List[List[int]]`**：这是 Python 3.5+ 引入的**类型提示**。它告诉阅读代码的人（以及 IDE 工具），参数 `matrix` 应该是一个“由整数列表组成的列表”（即二维数组）。
  * **`-> List[List[int]]`**：表示这个函数执行完，返回的也是一个二维整数列表。
  * **注意**：在实际运行中，Python 解释器会忽略这些提示，它们主要用于代码规范和文档说明，但在大厂规范（如 Google Style）中是必须写的。

**2\. 多变量赋值与解包 (Tuple Unpacking)**

```python
R, C = len(matrix), len(matrix[0])
```

  * **`len(matrix)`**：计算大列表里有多少个小列表，也就是**行数 ($R$)**。
  * **`len(matrix[0])`**：取出第一行，计算它有多长，也就是**列数 ($C$)**。
  * **`R, C = ...`**：这是一种快捷写法。等号右边其实隐式构成了一个元组 `(行数, 列数)`，然后自动解包分别赋值给 `R` 和 `C`。这比写两行代码更 Pythonic。

**列表推导式初始化矩阵 (List Comprehension) [重点]**

```python
res = [[0] * R for _ in range(C)]
```

这是面试中最容易被考察的一行代码，它做了三件事：

1.  **`[0] * R`**：利用列表乘法，快速生成一个包含 $R$ 个 0 的列表。例如 `R=2`，这就生成 `[0, 0]`。这代表**新矩阵的一行**。
2.  **`for _ in range(C)`**：循环 $C$ 次。
      * **`_` (下划线)**：这也是一个变量名，但在 Python 约定俗成中，表示\*\*“我不关心在这个循环里具体是第几次，我只需要循环这么多次”\*\*。这是一个占位符，比写 `i` 或 `j` 更专业。
3.  **组合**：这行代码的意思是：“生成 $C$ 个 `[0, 0, ...]` 这样的列表，并把它们放在一个大列表里”。

**为什么要这样写？**
如果你写成 `res = [[0] * R] * C`，这是一个经典的 **浅拷贝陷阱**！那样修改其中一行，所有行都会跟着变。使用 `for` 循环推导式是创建二维数组最安全、最标准的方法。

**双重循环与赋值**

```python
for r in range(R):
    for c in range(C):
        res[c][r] = matrix[r][c]
```

  * **`range(R)`**：生成从 0 到 $R-1$ 的整数序列。
  * **`res[c][r] = ...`**：这里直接通过索引访问并修改已存在的 `res` 列表。因为我们在第 3 步已经把 `res` 的“坑”都挖好了（填满了 0），所以这里可以直接赋值。如果 `res` 是空的，这里就会报错 `IndexError`。
