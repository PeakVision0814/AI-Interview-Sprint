## [541. 反转字符串 II](https://leetcode.cn/problems/reverse-string-ii/)

简单   相关标签  ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业

给定一个字符串 `s` 和一个整数 `k`，从字符串开头算起，每计数至 `2k` 个字符，就反转这 `2k` 字符中的前 `k` 个字符。

- 如果剩余字符少于 `k` 个，则将剩余字符全部反转。
- 如果剩余字符小于 `2k` 但大于或等于 `k` 个，则反转前 `k` 个字符，其余字符保持原样。

 

**示例 1：**

```
输入：s = "abcdefg", k = 2
输出："bacdfeg"
```

**示例 2：**

```
输入：s = "abcd", k = 2
输出："bacd"
```

````python
class Solution:
    def reverseStr(self, s:str, k:int) -> str:
        s_list = list(s)
        n = len(s_list)
        i = 0


        while i < n:
            start = i
            end = min(i + k - 1, n - 1)

            left = start
            right = end

            while left < right:
                temp = s_list[left]
                s_list[left] = s_list[right]
                s_list[right] = temp
                left += 1
                right -= 1
            i += 2 * k
        return "".join(s_list)


solution = Solution()
s_input = input("请输入字符串s：")
k_input_str = input("请输入k：")
k_input = int(k_input_str)
result = solution.reverseStr(s_input, k_input)
print(result)


````

**解题思路：**

这道题的关键在于如何根据规则在字符串中找到需要反转的子串，并进行反转。

1.  **分组处理：** 题目要求每隔 2k 个字符处理一次。我们可以把字符串想象成按 2k 的长度进行分组。
2.  **反转逻辑：** 在每个 2k 长度的分组中，我们只需要反转**前 k 个字符**。
3.  **边界情况：** 需要考虑字符串末尾不足 2k 个字符的情况。
    *   如果剩余字符少于 k 个：反转所有剩余字符。
    *   如果剩余字符大于或等于 k 个，但小于 2k 个：只反转前 k 个字符，剩余字符保持原样。
4.  **操作字符串：** 字符串在 Python 中是不可变的，这意味着你不能直接修改字符串的某个字符。为了方便操作，我们可以先将字符串转换为一个可变的列表（例如字符列表），在列表中进行反转操作，最后再将列表重新组合成字符串。

**算法步骤：**

1.  **将字符串转换为列表：** 将输入的字符串 `s` 转换为一个字符列表 `s_list`。这样我们可以方便地通过索引修改字符。
2.  **迭代处理：** 使用一个循环来遍历字符串（或列表）。每次循环跳跃 `2k` 个位置，以处理下一个 2k 长度的分组。我们用一个变量 `i` 来表示当前分组的起始索引。
3.  **确定反转子串的范围：** 在当前分组（从索引 `i` 开始）中，需要反转的子串的范围是 `[i, i + k - 1]`。但是需要考虑边界情况，所以实际的反转结束索引应该是 `min(i + k - 1, n - 1)`，其中 `n` 是字符串的长度。这样可以确保不会超出列表的边界。
4.  **反转子串：** 对于确定好的反转范围 `[start, end]`，使用双指针法（`left` 和 `right`）来反转子串。`left` 从 `start` 开始，`right` 从 `end` 开始，每次交换 `s_list[left]` 和 `s_list[right]`，然后 `left` 右移，`right` 左移，直到 `left >= right`。
5.  **更新循环索引：** 每次完成一个分组的处理后，将 `i` 增加 `2k`，跳到下一个分组的起始位置。
6.  **拼接回字符串：** 循环结束后，将修改后的 `s_list` 列表中的字符重新拼接成一个字符串，并返回。

## [560. 和为 K 的子数组](https://leetcode.cn/problems/subarray-sum-equals-k/)

中等  相关标签     ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业



给你一个整数数组 `nums` 和一个整数 `k` ，请你统计并返回 *该数组中和为 `k` 的子数组的个数* 。

子数组是数组中元素的连续非空序列。

 

**示例 1：**

```
输入：nums = [1,1,1], k = 2
输出：2
```

**示例 2：**

```
输入：nums = [1,2,3], k = 3
输出：2
```

 

**提示：**

- `1 <= nums.length <= 2 * 104`
- `-1000 <= nums[i] <= 1000`
- `-107 <= k <= 107`

````python
class Solution:
    def subarraySum(self, nums: List[int], k: int) -> int:
        # 字典用于存储前缀和出现的次数
        prefix_sum_count = {0: 1}
        current_sum = 0  # 当前前缀和
        count = 0  # 满足条件的子数组计数
        
        for num in nums:
            current_sum += num  # 更新当前前缀和
            
            # 检查是否存在前缀和为 current_sum - k，如果存在说明找到了一个满足条件的子数组
            if current_sum - k in prefix_sum_count:
                count += prefix_sum_count[current_sum - k]
            
            # 更新哈希表中 current_sum 的计数
            if current_sum in prefix_sum_count:
                prefix_sum_count[current_sum] += 1
            else:
                prefix_sum_count[current_sum] = 1
        
        return count

solution = Solution()
input_str_nums = input("请输入nums（例如: 1,2,3）：") # 提示用户输入不带方括号
# 去掉空格，然后按逗号分割，将每个部分转换为整数
nums_list_str = input_str_nums.split(',') # 分割成字符串列表，例如 ['1', '2', '3']
nums_list_int = [int(x.strip()) for x in nums_list_str] # 移除空白并转换为整数列表
input_k = input("请输入k：")
input_k = int(input_k)
output_count = solution.subarraySum(nums_list_int, input_k) # 将整数列表传递给函数
print(output_count)
````

**从算法角度解释**


这段代码使用了**前缀和 (Prefix Sum)** 和 **哈希表 (Hash Map 或 Dictionary)** 的思想来解决问题。

核心思想：

我们要找一个子数组 `nums[i:j]` (包含索引 `i` 到 `j` 的元素) 使得其和等于 `k`。
我们可以用前缀和来表示子数组的和：
`sum(nums[i:j]) = sum(nums[0:j]) - sum(nums[0:i-1])`

令 `current_sum = sum(nums[0:j])` (从开头到当前索引 `j` 的前缀和)
令 `previous_sum = sum(nums[0:i-1])` (从开头到索引 `i-1` 的前缀和)

那么，我们要找的就是是否存在一个 `previous_sum`，使得：
`current_sum - previous_sum = k`
或者
`previous_sum = current_sum - k`

算法步骤：

1.  **初始化**:
    *   创建一个字典 `prefix_sum_count` 来存储每个前缀和以及它出现的次数。
    *   将 `0` 作为初始前缀和加入字典，次数为 1。这是为了处理那些从数组开头开始的子数组，它们的 `previous_sum` 相当于 0。
    *   初始化 `current_sum` 为 0。
    *   初始化 `count` 为 0。

2.  **遍历数组**: 逐个遍历数组 `nums` 中的每个元素。

3.  **更新前缀和**: 在每次遍历中，将当前元素加到 `current_sum` 上，更新当前的累积和。

4.  **查找目标前缀和**: 计算目标前缀和 `target_prefix_sum = current_sum - k`。我们想知道之前是否有某个位置的前缀和等于 `target_prefix_sum`。

5.  **检查字典**: 在字典 `prefix_sum_count` 中查找 `target_prefix_sum` 是否存在。
    *   如果 `target_prefix_sum` 存在于字典中，说明之前出现过这个前缀和。每出现一次这样的前缀和，就意味着我们找到了一个以当前位置为结束、以该前缀和对应的位置+1为开始的子数组，其和为 `k`。因此，我们将 `count` 加上 `prefix_sum_count[target_prefix_sum]`（因为可能有多个位置的前缀和是 `target_prefix_sum`）。

6.  **更新字典**: 将当前的 `current_sum` 加入或更新到字典 `prefix_sum_count` 中。
    *   如果 `current_sum` 已经在字典中，就将它的计数加 1。
    *   如果 `current_sum` 不在字典中，就将它添加到字典中，并将计数设为 1。

7.  **返回结果**: 遍历完整个数组后，`count` 的值就是所有和为 `k` 的连续子数组的数量。

**举例说明:**

`nums = [1, 1, 1]`, `k = 2`

1.  **初始化**: `prefix_sum_count = {0: 1}`, `current_sum = 0`, `count = 0`

2.  **遍历第一个元素 `num = 1`**:
    *   `current_sum` 更新为 `0 + 1 = 1`。
    *   `target_prefix_sum = current_sum - k = 1 - 2 = -1`。
    *   `-1` 不在 `prefix_sum_count` 中。
    *   更新 `prefix_sum_count`: `current_sum = 1`，不在字典中，加入 `{0: 1, 1: 1}`。

3.  **遍历第二个元素 `num = 1`**:
    *   `current_sum` 更新为 `1 + 1 = 2`。
    *   `target_prefix_sum = current_sum - k = 2 - 2 = 0`。
    *   `0` 在 `prefix_sum_count` 中！`prefix_sum_count[0] = 1`。
    *   `count` 更新为 `0 + 1 = 1`。（找到了子数组 `[1, 1]`，从索引 0 到 1）
    *   更新 `prefix_sum_count`: `current_sum = 2`，不在字典中，加入 `{0: 1, 1: 1, 2: 1}`。

4.  **遍历第三个元素 `num = 1`**:
    *   `current_sum` 更新为 `2 + 1 = 3`。
    *   `target_prefix_sum = current_sum - k = 3 - 2 = 1`。
    *   `1` 在 `prefix_sum_count` 中！`prefix_sum_count[1] = 1`。
    *   `count` 更新为 `1 + 1 = 2`。（找到了子数组 `[1, 1]`，从索引 1 到 2）
    *   更新 `prefix_sum_count`: `current_sum = 3`，不在字典中，加入 `{0: 1, 1: 1, 2: 1, 3: 1}`。

5.  **遍历结束**: 返回 `count = 2`。

**为什么这种方法比暴力法要好？**

暴力法是检查所有可能的连续子数组，计算它们的和，然后看是否等于 `k`。如果有 `n` 个元素，子数组的数量大约是 `n^2` 个。对于每个子数组，计算和的时间复杂度是 `O(n)` (或者如果用前缀和优化，可以降到 O(1))。所以暴力法的总时间复杂度可能是 `O(n^3)` 或 `O(n^2)`。

而使用前缀和和哈希表的方法，我们只遍历数组一次（`O(n)`），在字典中的查找和更新操作平均时间复杂度是 `O(1)`。所以整个算法的时间复杂度是 `O(n)`。空间复杂度是 `O(n)`，因为字典最多可能存储 `n` 个不同的前缀和。

**总结**

这段代码利用了前缀和的概念，将查找子数组和的问题转化为查找特定前缀和的问题。通过使用哈希表来高效地存储和查询前缀和出现的次数，它能够在 O(n) 的时间复杂度内解决问题，这比暴力方法要高效得多。Python 语法在这里主要使用了变量、列表遍历、条件判断和字典操作。

## [566. 重塑矩阵](https://leetcode.cn/problems/reshape-the-matrix/)

简单   相关标签   提示

在 MATLAB 中，有一个非常有用的函数 `reshape` ，它可以将一个 `m x n` 矩阵重塑为另一个大小不同（`r x c`）的新矩阵，但保留其原始数据。

给你一个由二维数组 `mat` 表示的 `m x n` 矩阵，以及两个正整数 `r` 和 `c` ，分别表示想要的重构的矩阵的行数和列数。

重构后的矩阵需要将原始矩阵的所有元素以相同的 **行遍历顺序** 填充。

如果具有给定参数的 `reshape` 操作是可行且合理的，则输出新的重塑矩阵；否则，输出原始矩阵。

 

**示例 1：**

![img](https://assets.leetcode.com/uploads/2021/04/24/reshape1-grid.jpg)

```
输入：mat = [[1,2],[3,4]], r = 1, c = 4
输出：[[1,2,3,4]]
```

**示例 2：**

![img](https://assets.leetcode.com/uploads/2021/04/24/reshape2-grid.jpg)

```
输入：mat = [[1,2],[3,4]], r = 2, c = 4
输出：[[1,2],[3,4]]
```

````python
class Solution:
    def matrixReshape(self, mat: List[List[int]], r: int, c: int) -> List[List[int]]:
        # 1. 获取原矩阵的行数和列数
        m, n = len(mat), len(mat[0])

        # 2. 合法性判断：如果与那苏总数不一致，直接返回原矩阵
        if m * n != r * c:
            return mat
        
        # 3. 扁平化：将二维矩阵转换为一维列表
        # 这里的逻辑是：遍历每一行，再遍历行里的每一个元素
        flat = [num for row in mat for num in row]

        # 4. 构造新矩阵
        ans = []
        # 步长为c，意味着每次取c个元素构成新矩阵的一行
        for i in range(0, len(flat), c):
            # 利用切片flat[i:i+c]取出从i开始的c个元素，作为新矩阵的一行
            ans.append(flat[i:i+c])

        return ans
    
solution = Solution()
input_mat_str = input("请输入矩阵mat（例如: [[1,2],[3,4]]）：")
mat = eval(input_mat_str)  # 使用 eval 将字符串转换为列表
r = int(input("请输入行数 r："))
c = int(input("请输入列数 c："))
input_mat = solution.matrixReshape(mat, r, c)  # 使用 eval 将字符串转换为列表
print(input_mat)
````

**算法逻辑分析 (Algorithm)**

这段代码的核心思想是：**先破后立**（先降维，再升维）。

  * **步骤一：数据守恒检查 (Conservation of Elements)**

      * **逻辑**：如果你要把一块面团捏成另一个形状，面团的总量（体积）是不能变的。
      * **代码**：`if m * n != r * c: return mat`
      * **解释**：原矩阵有 $m \times n$ 个元素，新矩阵需要 $r \times c$ 个元素。如果两者不相等，说明无法重塑，直接根据题意返回原样。

  * **步骤二：降维打击 (Flattening)**

      * **逻辑**：因为计算机内存本质上是线性的，或者题目要求“行遍历顺序”，最简单的办法是先把二维矩阵“拉直”成一条长长的一维数组。
      * **代码**：`flat = [...]`
      * **效果**：`[[1, 2], [3, 4]]` $\rightarrow$ `[1, 2, 3, 4]`。

  * **步骤三：重新分组 (Reshaping / Slicing)**

      * **逻辑**：手里拿着长条形的数据，按照新矩阵的要求，每 `c` 个数据切一刀，作为新矩阵的一行。
      * **代码**：`ans.append(flat[i : i + c])`
      * **图解**：
        数据流： `1, 2, 3, 4, 5, 6`
        若 `c=2`： `[1, 2]` (切) `[3, 4]` (切) `[5, 6]` (切)


**Python 语法亮点 (Syntax)**

这段代码用到了很多 Python "Pythonic"（地道）的写法：

**A. 类型提示 (Type Hinting)**

```python
def matrixReshape(self, mat: List[List[int]], ... ) -> List[List[int]]:
```

  * **`List[List[int]]`**：这告诉阅读代码的人（以及 IDE），`mat` 应该是一个“列表的列表，且里面装的是整数”（即二维整数数组）。
  * **`->`**：箭头后面表示这个函数**返回**的数据类型。
  * *注意：这只是给开发者看的“备注”，Python 运行时并不会强制检查类型。*

**B. 元组解包 (Tuple Unpacking)**

```python
m, n = len(mat), len(mat[0])
```

  * Python 允许同时给多个变量赋值。这行代码等同于：
    ```python
    m = len(mat)    # 行数
    n = len(mat[0]) # 列数
    ```

**C. 双重列表推导式 (Nested List Comprehension)**

```python
flat = [num for row in mat for num in row]
```

  * 这是 Python 的“大杀器”。它把两层 `for` 循环压缩成了一行，专门用来做**扁平化**操作。

**D. `range` 的步长 (Step)**

```python
for i in range(0, len(flat), c):
```

  * **`range(start, stop, step)`**：
      * 从 `0` 开始。
      * 到 `len(flat)` 结束。
      * **`step=c`**：这是关键！循环变量 `i` 每次增加 `c`。
      * 比如 `c=2`，`i` 的值会依次是 `0, 2, 4, 6...`。这正好对应每一行的**起始索引**。

**E. 列表切片 (List Slicing)**

```python
flat[i : i+c]
```

  * **`[start : end]`**：截取列表的一部分。
  * **左闭右开**：包含 `start`，**不包含** `end`。
  * 刚好截取从 `i` 开始的 `c` 个元素，构成新的一行。

**F. `eval()` 函数 (测试部分)**

```python
mat = eval(input_mat_str)
```

  * **功能**：`eval` 是 "evaluate" 的缩写。它把**字符串**当作**有效的 Python 表达式**来执行。
  * **作用**：当你输入字符串 `[[1,2],[3,4]]` 时，`input()` 拿到的是文本。`eval()` 把它一看：“哟，这是个列表定义啊”，于是直接把它变成了真正的列表对象。
  * *安全警示*：在写算法题测试脚本时 `eval` 很好用，但在\*\*实际开发（Web后端等）\*\*中绝对禁止直接对用户输入使用 `eval`，因为用户可以输入恶意代码（如删除文件命令）来攻击服务器。
