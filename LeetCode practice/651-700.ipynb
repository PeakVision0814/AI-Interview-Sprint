## [661. 图片平滑器](https://leetcode.cn/problems/image-smoother/)

简单  相关标签    ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业



**图像平滑器** 是大小为 `3 x 3` 的过滤器，用于对图像的每个单元格平滑处理，平滑处理后单元格的值为该单元格的平均灰度。

每个单元格的 **平均灰度** 定义为：该单元格自身及其周围的 8 个单元格的平均值，结果需向下取整。（即，需要计算蓝色平滑器中 9 个单元格的平均值）。

如果一个单元格周围存在单元格缺失的情况，则计算平均灰度时不考虑缺失的单元格（即，需要计算红色平滑器中 4 个单元格的平均值）。

![img](https://assets.leetcode.com/uploads/2021/05/03/smoother-grid.jpg)

给你一个表示图像灰度的 `m x n` 整数矩阵 `img` ，返回对图像的每个单元格平滑处理后的图像 。

 

**示例 1:**

![img](https://assets.leetcode.com/uploads/2021/05/03/smooth-grid.jpg)

```
输入:img = [[1,1,1],[1,0,1],[1,1,1]]
输出:[[0, 0, 0],[0, 0, 0], [0, 0, 0]]
解释:
对于点 (0,0), (0,2), (2,0), (2,2): 平均(3/4) = 平均(0.75) = 0
对于点 (0,1), (1,0), (1,2), (2,1): 平均(5/6) = 平均(0.83333333) = 0
对于点 (1,1): 平均(8/9) = 平均(0.88888889) = 0
```

**示例 2:**

![img](https://assets.leetcode.com/uploads/2021/05/03/smooth2-grid.jpg)

```
输入: img = [[100,200,100],[200,50,200],[100,200,100]]
输出: [[137,141,137],[141,138,141],[137,141,137]]
解释:
对于点 (0,0), (0,2), (2,0), (2,2): floor((100+200+200+50)/4) = floor(137.5) = 137
对于点 (0,1), (1,0), (1,2), (2,1): floor((200+200+50+200+100+100)/6) = floor(141.666667) = 141
对于点 (1,1): floor((50+200+200+200+200+100+100+100+100)/9) = floor(138.888889) = 138
```

````python
class Solution:
    def imageSmoother(self, img: List[List[int]]) -> List[List[int]]:
        # 获取矩阵的行数 m 和列数 n
        m, n = len(img), len(img[0])
        
        # 创建一个同样大小的零矩阵用于存放结果
        # 注意：不能直接 copy，需要深拷贝或者重新创建，防止引用同一个对象
        res = [[0] * n for _ in range(m)]
        
        # 遍历每一个单元格 (i, j) 作为中心点
        for i in range(m):
            for j in range(n):
                total_sum = 0
                count = 0
                
                # 遍历以 (i, j) 为中心的 3x3 区域
                # x 的范围是 [i-1, i+1]，y 的范围是 [j-1, j+1]
                for x in range(i - 1, i + 2):
                    for y in range(j - 1, j + 2):
                        # 边界检查：确保 (x, y) 在矩阵范围内
                        if 0 <= x < m and 0 <= y < n:
                            total_sum += img[x][y]
                            count += 1
                
                # 计算平均值并向下取整
                res[i][j] = total_sum // count
                
        return res
    
solution = Solution()

img = input("请输入矩阵img（例如 [[1,2,3],[4,5,6],[7,8,9]]）：")
img = eval(img)
print(solution.imageSmoother(img))
````

**算法逻辑解释 (Algorithm)**

这个算法的核心思想是：**“创建一个新画板，对照着旧画板描绘”**。

1.  **获取尺寸 (`m, n`)**：
      * 首先知道图片有多大（几行几列），以便后续遍历。
2.  **预分配空间 (`res`)**：
      * **关键点**：我们需要创建一个全新的矩阵 `res` 来存结果。
      * **原因**：如果直接在原矩阵 `img` 上修改，当你计算 `(0,0)` 的平均值并修改了它，那么在计算旁边的 `(0,1)` 时，就会用到 `(0,0)` 修改后的**新值**，而不是**原始值**，这会导致结果错误。
3.  **四重循环 (核心逻辑)**：
      * **外层两重循环 (`i, j`)**：遍历矩阵中的每一个像素点，作为“中心点”。
      * **内层两重循环 (`x, y`)**：对于每一个中心点，遍历它周围 3x3 的区域（九宫格）。
4.  **边界检查 (`if 0 <= x < m ...`)**：
      * 这是矩阵题最容易出错的地方。因为九宫格在边缘（如角落）时，部分邻居坐标是负数或超出了矩阵范围。我们需要用 `if` 把这些无效坐标过滤掉，只累加有效的点。
5.  **求平均值**：
      * 将累加的 `total_sum` 除以有效点的个数 `count`，并向下取整，填入新矩阵 `res`。

**Python 语法重点 (Syntax)**

这段代码包含几个非常地道的 Python 用法：

**A. 元组解包 (Tuple Unpacking)**

```python
m, n = len(img), len(img[0])
```

  * **解释**：Python 允许一行代码同时给多个变量赋值。
  * **等价于**：
    ```python
    m = len(img)      # 行数
    n = len(img[0])   # 列数
    ```

**二维列表推导式 (List Comprehension)**

```python
res = [[0] * n for _ in range(m)]
```

  * **解释**：这是创建二维数组的标准写法。
  * **`[0] * n`**：创建一个包含 `n` 个 0 的列表（一行）。
  * **`for _ in range(m)`**：将上面那个列表重复创建 `m` 次。
  * **`_` (下划线)**：表示在这个循环中，我们不需要用到循环变量（如 `i`），只是单纯想循环 `m` 次，这是一种编程习惯。
  * **易错点**：
      * **千万不要写成** `res = [[0] * n] * m`。
      * 这种写法会导致所有行都**引用同一个对象**。修改第一行，所有行都会跟着变（浅拷贝陷阱）。上面的列表推导式写法保证了每一行都是独立的内存空间。

**range() 函数的“左闭右开”**

```python
for x in range(i - 1, i + 2):
```

  * **解释**：`range(start, stop)` 生成的序列包含 `start`，**但不包含** `stop`。
  * **逻辑**：我们需要遍历 `i-1`, `i`, `i+1` 这三个坐标。
  * **写法**：为了取到 `i+1`，`stop` 参数必须写成 `i+2`。

**连续比较运算符**

```python
if 0 <= x < m and 0 <= y < n:
```

  * **解释**：Python 允许链式比较。
  * **其他语言写法**：通常需要写成 `if (x >= 0 && x < m)`。
  * **Python写法**：`0 <= x < m` 直观且符合数学书写习惯，表示 `x` 在 `[0, m)` 区间内。

#**地板除 (Floor Division)**

```python
total_sum // count
```

  * **解释**：`//` 运算符表示整除，结果会**向下取整**（丢弃小数部分）。
  * **对比**：
      * `5 / 2` 结果是 `2.5` (浮点数)
      * `5 // 2` 结果是 `2` (整数)
      * 题目要求“向下取整”，所以这里必须用 `//`。
