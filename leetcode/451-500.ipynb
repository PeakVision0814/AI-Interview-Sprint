## [438. 找到字符串中所有字母异位词](https://leetcode.cn/problems/find-all-anagrams-in-a-string/)

中等



相关标签

![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业



给定两个字符串 `s` 和 `p`，找到 `s` 中所有 `p` 的 **异位词** 的子串，返回这些子串的起始索引。不考虑答案输出的顺序。

 

**示例 1:**

```
输入: s = "cbaebabacd", p = "abc"
输出: [0,6]
解释:
起始索引等于 0 的子串是 "cba", 它是 "abc" 的异位词。
起始索引等于 6 的子串是 "bac", 它是 "abc" 的异位词。
```

 **示例 2:**

```
输入: s = "abab", p = "ab"
输出: [0,1,2]
解释:
起始索引等于 0 的子串是 "ab", 它是 "ab" 的异位词。
起始索引等于 1 的子串是 "ba", 它是 "ab" 的异位词。
起始索引等于 2 的子串是 "ab", 它是 "ab" 的异位词。
```

 

**提示:**

- `1 <= s.length, p.length <= 3 * 104`
- `s` 和 `p` 仅包含小写字母

````python
from typing import List

class Solution:
    def findAnagrams(self, s: str, p: str) -> List[int]:
        result = []
        len_p = len(p)
        len_s = len(s)

        if len_p > len_s:
            return result

        count_p = [0] * 26
        current_count = [0] * 26

        for char in p:
            count_p[ord(char) - ord('a')] += 1

        # 初始化第一个窗口的字符计数
        for i in range(len_p):
            current_count[ord(s[i]) - ord('a')] += 1

        if current_count == count_p:
            result.append(0)

        # 滑动窗口
        for i in range(len_p, len_s):
            # 移除窗口最左侧的字符
            left_char = s[i - len_p]
            current_count[ord(left_char) - ord('a')] -= 1

            # 添加新字符到窗口最右侧
            new_char = s[i]
            current_count[ord(new_char) - ord('a')] += 1

            # 判断当前窗口是否满足条件
            if current_count == count_p:
                result.append(i - len_p + 1)

        return result
````

## [455. 分发饼干](https://leetcode.cn/problems/assign-cookies/)

简单   相关标签   ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业



假设你是一位很棒的家长，想要给你的孩子们一些小饼干。但是，每个孩子最多只能给一块饼干。

对每个孩子 `i`，都有一个胃口值 `g[i]`，这是能让孩子们满足胃口的饼干的最小尺寸；并且每块饼干 `j`，都有一个尺寸 `s[j]` 。如果 `s[j] >= g[i]`，我们可以将这个饼干 `j` 分配给孩子 `i` ，这个孩子会得到满足。你的目标是满足尽可能多的孩子，并输出这个最大数值。

**示例 1:**

```
输入: g = [1,2,3], s = [1,1]
输出: 1
解释: 
你有三个孩子和两块小饼干，3 个孩子的胃口值分别是：1,2,3。
虽然你有两块小饼干，由于他们的尺寸都是 1，你只能让胃口值是 1 的孩子满足。
所以你应该输出 1。
```

**示例 2:**

```
输入: g = [1,2], s = [1,2,3]
输出: 2
解释: 
你有两个孩子和三块小饼干，2 个孩子的胃口值分别是 1,2。
你拥有的饼干数量和尺寸都足以让所有孩子满足。
所以你应该输出 2。
```

 

**提示：**

- `1 <= g.length <= 3 * 104`
- `0 <= s.length <= 3 * 104`
- `1 <= g[i], s[j] <= 231 - 1`

 

````python
from typing import List
class Solution():
    def findContentChildren(self, g: List[int], s: List[int]) -> int:
        g.sort()
        s.sort()
        index = 0
        for i in range(len(s)):
            if index < len(g) and g[index] <= s[i]:
                index += 1
        return index
sol = Solution()
g = [1, 2]
s= [1, 1, 4]
result = sol.findContentChildren(g, s)
print(result)
````

**代码的目标就是找出最多能满足多少个小朋友。**

**现在，我们一步步来解析这段代码：**

1.  **`from typing import List`**: 这行代码是用来给代码添加“类型提示”的。`List`表示列表。就像我们写作业会标上语文、数学一样，这里的`List`是告诉Python这个变量是一个列表。这对于我们理解代码很有帮助，尤其是在大型项目中。即使没有这行，代码也能运行，但加上它会让代码更清晰易懂。

2.  **`class Solution():`**: 在Python中，我们经常使用类（`class`）来组织代码。`Solution`只是这个类的一个名字，你可以理解为一个“工具箱”，里面放着我们解决问题的方法。

3.  **`def findContentChildren(self, g: List[int], s: List[int]) -> int:`**:
    *   `def` 关键字表示我们要定义一个函数，函数可以理解为一段用来完成特定任务的代码块。
    *   `findContentChildren` 是这个函数的名字，它告诉我们这个函数的功能是找到能够满足的小朋友的数量。
    *   `self` 是类中的一个特殊参数，代表这个类的实例本身。在调用类中的函数时，`self`会自动被传递，我们暂时不用太关心它。
    *   `g: List[int]` 表示输入的第一个参数是`g`，它的类型是一个整数列表（`List[int]`）。`g`代表小朋友的胃口值列表。
    *   `s: List[int]` 表示输入的第二个参数是`s`，它的类型也是一个整数列表（`List[int]`）。`s`代表饼干的尺寸列表。
    *   `-> int` 表示这个函数会返回一个整数。这个整数就是我们找到的能满足的小朋友的数量。

4.  **`g.sort()`**: 这行代码对小朋友的胃口值列表`g`进行了排序。默认是升序排序，也就是说胃口小的排在前面。

5.  **`s.sort()`**: 这行代码对饼干的尺寸列表`s`进行了排序。默认也是升序排序，也就是说小尺寸的饼干排在前面。

    **为什么我们要排序呢？** 这是解决这个问题的关键思路！我们可以思考一下，如果我们想尽可能多地满足小朋友，是不是应该优先用最小的能满足小朋友的饼干去满足胃口最小的小朋友呢？这样就不会浪费大尺寸的饼干去满足小胃口的小朋友了。排序就是为了方便我们按照这个策略来分发饼干。

6.  **`index = 0`**: 我们初始化一个变量`index`，并将其设置为0。这个变量用来记录我们已经满足了多少个小朋友。一开始，一个小朋友也没满足，所以是0。

7.  **`for i in range(len(s)):`**: 这是一个`for`循环，用来遍历饼干列表`s`。`range(len(s))`会生成一个从0到`len(s) - 1`的数字序列。`i`会依次取到这个序列中的每一个数字，代表当前我们正在看的是饼干列表中的第`i`个饼干。

8.  **`if index < len(g) and g[index] <= s[i]:`**: 这是循环中的一个条件判断。
    *   `index < len(g)`: 这个条件判断我们是否还有小朋友没有被满足。如果`index`等于`len(g)`，说明我们已经满足了所有小朋友了。
    *   `g[index] <= s[i]`: 这个条件判断当前的饼干（`s[i]`）是否足够大，能够满足当前还没有被满足的胃口最小的小朋友（`g[index]`）。
    *   `and` 连接了这两个条件，表示只有当两个条件都满足时，后面的代码才会执行。

9.  **`index += 1`**: 如果上面的条件成立，也就是说我们还有小朋友没满足，并且当前的饼干可以满足当前胃口最小的小朋友，那么我们就成功满足了一个小朋友！所以我们将`index`加1，表示满足的小朋友数量增加了一个。

10. **`return index`**: 循环结束后，`index`的值就是我们成功满足的小朋友的总数，所以我们将它作为函数的返回值返回。

**最后面的代码是用来测试我们写的函数的：**

1.  **`sol = Solution()`**: 我们创建了一个`Solution`类的实例，并将其命名为`sol`。现在`sol`就是一个“工具箱”，我们可以使用里面的函数。
2.  **`g = [1, 2]`**: 定义了小朋友的胃口值列表。
3.  **`s= [1, 1, 4]`**: 定义了饼干的尺寸列表。
4.  **`result = sol.findContentChildren(g, s)`**: 调用`sol`工具箱里的`findContentChildren`函数，并将小朋友和饼干列表作为参数传递进去。函数执行后，返回的结果会被赋值给`result`变量。
5.  **`print(result)`**: 打印出`result`的值，也就是我们找到的能满足的小朋友的数量。

**总的算法思想：**

这个算法采用了**贪心策略**。我们首先将小朋友的胃口和饼干的尺寸都进行排序。然后，我们从小胃口的小朋友开始，尝试用能满足他的最小饼干来满足他。如果当前的饼干不够，我们就看下一块饼干。如果当前的饼干够了，我们就用这块饼干满足这个小朋友，然后看下一个小朋友，再尝试用下一块饼干来满足。这样一步步地进行，直到所有的饼干都分发完毕或者所有的小朋友都被满足。这种策略能够确保我们在每一步都做出最优的选择，从而最终得到最多的被满足的小朋友数量。

**这个算法的时间复杂度是O(n log n)，主要是由于排序的时间消耗。遍历饼干列表的时间是O(m)，其中n是小朋友的数量，m是饼干的数量。所以总的时间复杂度取决于排序中较大的那个。**

## [454. 四数相加 II](https://leetcode.cn/problems/4sum-ii/)

中等   相关标签  ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业



给你四个整数数组 `nums1`、`nums2`、`nums3` 和 `nums4` ，数组长度都是 `n` ，请你计算有多少个元组 `(i, j, k, l)` 能满足：

- `0 <= i, j, k, l < n`
- `nums1[i] + nums2[j] + nums3[k] + nums4[l] == 0`

 

**示例 1：**

```
输入：nums1 = [1,2], nums2 = [-2,-1], nums3 = [-1,2], nums4 = [0,2]
输出：2
解释：
两个元组如下：
1. (0, 0, 0, 1) -> nums1[0] + nums2[0] + nums3[0] + nums4[1] = 1 + (-2) + (-1) + 2 = 0
2. (1, 1, 0, 0) -> nums1[1] + nums2[1] + nums3[0] + nums4[0] = 2 + (-1) + (-1) + 0 = 0
```

**示例 2：**

```
输入：nums1 = [0], nums2 = [0], nums3 = [0], nums4 = [0]
输出：1
```

````python
from typing import List

class Solution:
    def fourSumCount(self, nums1: List[int], nums2: List[int], nums3: List[int], nums4: List[int]) -> int:
        # 使用哈希表存储 nums1[i] + nums2[j] 的和及其出现的次数
        sum_ab = {}
        for a in nums1:
            for b in nums2:
                current_sum = a + b
                sum_ab[current_sum] = sum_ab.get(current_sum, 0) + 1

        # 遍历 nums3 和 nums4，查找满足条件的情况
        count = 0
        for c in nums3:
            for d in nums4:
                target_sum = -(c + d)  # 需要在 sum_ab 中查找的目标和
                if target_sum in sum_ab:
                    count += sum_ab[target_sum]

        return count

solution = Solution()
input_str = input("请输入nums1 (用空格分隔): ")
nums1 = [int(x) for x in input_str.split()]

input_str = input("请输入nums2 (用空格分隔): ")
nums2 = [int(x) for x in input_str.split()]

input_str = input("请输入nums3 (用空格分隔): ")
nums3 = [int(x) for x in input_str.split()]

input_str = input("请输入nums4 (用空格分隔): ")
nums4 = [int(x) for x in input_str.split()]

output_count = solution.fourSumCount(nums1, nums2, nums3, nums4)

print(f"有{output_count}个元组能满足")
````

我们的目标是找到四个数组 `nums1`, `nums2`, `nums3`, `nums4` 中各取一个元素，使得它们的和等于零。也就是说，找到满足 `nums1[i] + nums2[j] + nums3[k] + nums4[l] == 0` 的 `(i, j, k, l)` 组合的数量。

**代码整体思路：**

直观地想，我们可以使用四个嵌套循环来遍历所有可能的 `(i, j, k, l)` 组合，然后判断它们的和是否为零。但是，如果每个数组的长度是 `n`，那么总共有 `n * n * n * n = n^4` 种组合。当 `n` 比较大时，这种方法会非常慢，效率很低。

为了提高效率，我们可以利用哈希表（字典）来“空间换时间”。核心思想是将问题分解：

1. **计算前两个数组的和：** 遍历 `nums1` 和 `nums2`，计算所有可能的 `nums1[i] + nums2[j]` 的和，并将每个和以及它出现的次数存储在一个哈希表中。
2. **查找后两个数组的匹配和：** 遍历 `nums3` 和 `nums4`，计算所有可能的 `nums3[k] + nums4[l]` 的和。对于每一个这样的和，如果 `- (nums3[k] + nums4[l])` （也就是 `-(c + d)`）存在于之前计算的哈希表中，说明我们找到了一个满足条件的组合。具体来说，如果 `-(c + d)` 在哈希表中出现的次数是 `x`，那么对于当前的 `c` 和 `d` 组合，它可以与哈希表中所有对应和为 `-(c + d)` 的 `(i, j)` 组合匹配，因此我们要将计数增加 `x`。

**代码逐行解释：**

```python
from typing import List

class Solution:
    def fourSumCount(self, nums1: List[int], nums2: List[int], nums3: List[int], nums4: List[int]) -> int:
        """
        计算满足 nums1[i] + nums2[j] + nums3[k] + nums4[l] == 0 的元组 (i, j, k, l) 的数量。

        Args:
            nums1: 第一个整数数组。
            nums2: 第二个整数数组。
            nums3: 第三个整数数组。
            nums4: 第四个整数数组。

        Returns:
            满足条件的元组数量。
        """
        # 这一部分是函数的定义和文档字符串，说明了函数的作用、参数和返回值。
        # type hinting (List[int]) 是为了让代码更清晰，说明参数是整数列表。
        # """...""" 是文档字符串，解释函数的功能，方便其他人理解。

        # 使用哈希表存储 nums1[i] + nums2[j] 的和及其出现的次数
        sum_ab = {} # 这一行创建了一个空的字典（哈希表），用来存储 nums1[i] + nums2[j] 的和以及它们出现的次数。
                    # 字典的 key 是和，value 是出现的次数。

        for a in nums1: # 外层循环，遍历 nums1 中的每一个元素，并将其赋值给变量 a。
            for b in nums2: # 内层循环，遍历 nums2 中的每一个元素，并将其赋值给变量 b。
                            # 这个双重循环会遍历 nums1 和 nums2 中所有可能的元素对 (a, b)。
                current_sum = a + b # 计算当前元素对的和，存储在变量 current_sum 中。
                # sum_ab[current_sum] = sum_ab.get(current_sum, 0) + 1
                # 这一行是字典操作的核心：
                # sum_ab.get(current_sum, 0): 尝试从字典 sum_ab 中获取以 current_sum 为 key 的值。
                # 如果字典中已经有了这个 key，就返回对应的 value (也就是这个和之前出现的次数)。
                # 如果字典中没有这个 key，就返回默认值 0。
                # + 1: 在获取到的次数上加 1，表示这个和又出现了一次。
                # sum_ab[current_sum] = ...: 将更新后的次数重新存入字典中，以 current_sum 为 key。
                # 简而言之，这行代码的功能是：如果 current_sum 已经在字典中，就把它对应的次数加 1；如果不在，就创建一个新的条目， key 是 current_sum， value 是 1。

        # 遍历 nums3 和 nums4，查找满足条件的情况
        count = 0 # 初始化一个计数器，用来记录满足 nums1[i] + nums2[j] + nums3[k] + nums4[l] == 0 的组合数量。

        for c in nums3: # 外层循环，遍历 nums3 中的每一个元素，并将其赋值给变量 c。
            for d in nums4: # 内层循环，遍历 nums4 中的每一个元素，并将其赋值给变量 d。
                            # 这个双重循环会遍历 nums3 和 nums4 中所有可能的元素对 (c, d)。

                target_sum = -(c + d)  # 计算当前元素对 (c, d) 的和的负值，存储在变量 target_sum 中。
                                      # 为什么是负值？因为我们希望 nums1[i] + nums2[j] + nums3[k] + nums4[l] == 0。
                                      # 这等价于 nums1[i] + nums2[j] == -(nums3[k] + nums4[l])。
                                      # 所以，如果我们当前的 nums3[k] + nums4[l] 的和是 (c + d)，
                                      # 我们需要在之前计算的 nums1[i] + nums2[j] 的和中找到值为 -(c + d) 的情况。
                                      # target_sum 就是我们要查找的目标和。

                if target_sum in sum_ab: # 检查 target_sum 是否在之前构建的字典 sum_ab 中作为 key 存在。
                                         # 如果存在，说明存在至少一个 (i, j) 组合，使得 nums1[i] + nums2[j] 等于 target_sum。
                                         # 也就是存在 (i, j) 组合，使得 nums1[i] + nums2[j] == -(c + d)，
                                         # 进而 nums1[i] + nums2[j] + c + d == 0。

                    count += sum_ab[target_sum] # 如果 target_sum 存在于 sum_ab 中，我们就将 count 增加 sum_ab[target_sum] 的值。
                                              # sum_ab[target_sum] 是 target_sum 在 nums1[i] + nums2[j] 的和中出现的次数。
                                              # 对于当前的 (c, d) 组合，它可以与所有使得 nums1[i] + nums2[j] 等于 target_sum 的 (i, j) 组合配对。
                                              # 因此，满足条件的组合数量就增加了 target_sum 出现的次数。

        return count # 返回最终的计数，即满足条件的元组 (i, j, k, l) 的数量。
```

**算法总结：**

1. **构建哈希表：** 遍历 `nums1` 和 `nums2` 的所有组合，计算它们的和，并将和以及出现次数存入一个字典。这个字典记录了所有可能的 `nums1[i] + nums2[j]` 的值及其频率。
2. **查找匹配项：** 遍历 `nums3` 和 `nums4` 的所有组合，计算它们的和。对于每一个 `nums3[k] + nums4[l]` 的和，计算其负值。在之前构建的哈希表中查找这个负值。如果找到，说明存在 `nums1[i] + nums2[j]` 等于这个负值的情况，从而满足 `nums1[i] + nums2[j] + nums3[k] + nums4[l] == 0`。
3. **累加计数：** 如果在哈希表中找到了目标负值，将计数器增加该负值在哈希表中记录的出现次数。这是因为每一个出现次数都代表着一个可以与当前 `(k, l)` 组合形成满足条件的四元组的 `(i, j)` 组合。

**为什么这样效率更高？**

- 构建哈希表需要 `O(n^2)` 的时间复杂度（遍历 `nums1` 和 `nums2`）。
- 查找匹配项需要 `O(n^2)` 的时间复杂度（遍历 `nums3` 和 `nums4`），并且在哈希表中查找一个 key 的平均时间复杂度是 `O(1)`。
- 总的时间复杂度是 `O(n^2) + O(n^2 * 1)` ≈ `O(n^2)`。
- 相比于暴力解法的 `O(n^4)`， `O(n^2)` 的效率要高得多，尤其当 `n` 较大时。

## [459. 重复的子字符串](https://leetcode.cn/problems/repeated-substring-pattern/)

给定一个非空的字符串 `s` ，检查是否可以通过由它的一个子串重复多次构成。

 

**示例 1:**

```
输入: s = "abab"
输出: true
解释: 可由子串 "ab" 重复两次构成。
```

**示例 2:**

```
输入: s = "aba"
输出: false
```

**示例 3:**

```
输入: s = "abcabcabcabc"
输出: true
解释: 可由子串 "abc" 重复四次构成。 (或子串 "abcabc" 重复两次构成。)
```

````python
class Solution:
    def repeatedSubstringPattern(self, s: str) -> bool:
        s_len = len(s)
        for l in range(1, s_len // 2 + 1):
            if s_len % l == 0:
                sub = s[:l]
                count = s_len // l
                repeat_sub = sub * count
                if repeat_sub == s:
                    return True
        return False
solution = Solution()
input_s = input("请输入字符串s：")
outout_bool = solution.repeatedSubstringPattern(input_s)
print(outout_bool)
````

**算法思路：**

这道题的核心是判断一个字符串是否由某个子串重复多次组成。基于这个特点，我们可以想到最直观的方法是尝试所有可能的重复子串，然后验证它们是否符合要求。

如果字符串 `s` 是由某个子串重复构成的，那么这个子串一定具备以下性质：

1.  **子串必须是 `s` 的一个前缀。** （比如 "abab" 是由 "ab" 构成，而 "ab" 是它的前缀）
2.  **子串的长度必须是 `s` 的长度的因数。** （比如长度为 6 的字符串不可能由长度为 4 的子串重复构成）
3.  **将这个子串重复多次后，必须正好等于 `s`。**

基于这些性质，我们可以设计如下算法：

1.  获取输入字符串 `s` 的长度 `n`。
2.  如果一个字符串由某个子串重复构成，那么这个子串的最大长度不会超过 `n/2`。因为如果子串长度大于 `n/2`，重复两次就会超过原字符串的长度。所以，我们只需要考虑长度从 1 到 `n/2` 的子串。
3.  遍历所有可能的子串长度 `l`，从 1 到 `n/2`。
4.  在每次遍历中，首先检查 `n` 是否能被当前的子串长度 `l` 整除。这是因为如果 `n` 不能被 `l` 整除，那么长度为 `l` 的子串无论重复多少次，其总长度都不可能等于 `n`。
5.  如果 `n` 可以被 `l` 整除，说明长度为 `l` 的子串有可能是构成 `s` 的重复子串。
6.  提取 `s` 的前 `l` 个字符作为候选子串 `sub`。
7.  计算需要重复的次数 `count = n // l`。
8.  将候选子串 `sub` 重复 `count` 次，生成一个新的字符串 `repeated_sub`。
9.  比较 `repeated_sub` 是否与原始字符串 `s` 相等。
10. 如果相等，说明我们找到了一个符合条件的重复子串，可以直接返回 `True`。
11. 如果循环结束，都没有找到符合条件的重复子串，说明 `s` 不能由任何子串重复构成，返回 `False`。

**Python 代码实现：**

```python
class Solution:
    def repeatedSubstringPattern(self, s: str) -> bool:
        # 获取字符串s的长度
        s_len = len(s)

        # 遍历所有可能的子串长度，从1到字符串长度的一半（包括一半）
        # range(start, stop, step) 生成一个序列，这里从1开始，到s_len // 2 结束（不包含 s_len // 2 + 1）
        for l in range(1, s_len // 2 + 1):
            # 检查字符串的总长度是否能被当前的子串长度整除
            # 如果不能整除，说明当前长度的子串无法重复构成整个字符串
            if s_len % l == 0:
                # 提取长度为 l 的前缀子串
                # s[:l] 是字符串切片语法，表示从字符串开头到索引 l-1 的部分
                sub = s[:l]
                # 计算需要重复的次数
                # s_len // l 是整除运算，得到重复的次数
                count = s_len // l
                # 将子串重复 count 次
                # sub * count 是字符串的乘法操作，将字符串 sub 重复 count 次
                repeat_sub = sub * count
                # 检查重复后的子串是否与原始字符串 s 相等
                if repeat_sub == s:
                    # 如果相等，说明找到了符合条件的重复子串，返回 True
                    return True

        # 如果循环结束，都没有找到符合条件的重复子串，返回 False
        return False

# 以下是用于测试的代码
solution = Solution()

# 使用 input() 获取用户输入的字符串
input_s = input("请输入字符串s：")

# 调用 repeatedSubstringPattern 方法，并将结果赋值给 output_bool
outout_bool = solution.repeatedSubstringPattern(input_s)

# 打印结果
print(outout_bool)
```
