## 54. 替换数字（第八期模拟笔试）

**题目描述**

给定一个字符串 s，它包含小写字母和数字字符，请编写一个函数，将字符串中的字母字符保持不变，而将每个数字字符替换为number。 例如，对于输入字符串 "a1b2c3"，函数应该将其转换为 "anumberbnumbercnumber"。

**输入描述**

输入一个字符串 s,s 仅包含小写字母和数字字符。

**输出描述**

打印一个新的字符串，其中每个数字字符都被替换为了number

**输入示例**

```
a1b2c3
```

**输出示例**

```
anumberbnumbercnumber
```

**提示信息**

数据范围：
1 <= s.length < 10000。

````python
class Solution:
    def exchangeNumber(self, s:str) -> str:
        result = ""
        for char in s:
            if '0' <= char <= '9':
                result += "number"
            else:
                result += char
        return result

solution = Solution()

input_str = input("请输入字符串s")
output = solution.exchangeNumber(input_str)
print(output)
````

## 55. 右旋字符串（第八期模拟笔试）

**题目描述**

字符串的右旋转操作是把字符串尾部的若干个字符转移到字符串的前面。给定一个字符串 s 和一个正整数 k，请编写一个函数，将字符串中的后面 k 个字符移到字符串的前面，实现字符串的右旋转操作。 

例如，对于输入字符串 "abcdefg" 和整数 2，函数应该将其转换为 "fgabcde"。

**输入描述**

输入共包含两行，第一行为一个正整数 k，代表右旋转的位数。第二行为字符串 s，代表需要旋转的字符串。

**输出描述**

输出共一行，为进行了右旋转操作后的字符串。

**输入示例**

```
2
abcdefg
```

**输出示例**

```
fgabcde
```

**提示信息**

数据范围：
1 <= k < 10000,
1 <= s.length < 10000;

````python
class Solution():
    def right_xuanzhuan(self, k: int, s: str) -> str:
        s_list = list(s)
        left = 0
        right = len(s_list) - k - 1
        while left < right:
            s_list[left], s_list[right] = s_list[right], s_list[left]
            left += 1
            right -= 1
        left = len(s_list) - k
        right = len(s_list) - 1
        while left < right:
            s_list[left], s_list[right] = s_list[right], s_list[left]
            left += 1
            right -= 1
        left = 0
        right = len(s_list) - 1
        while left < right:
            s_list[left], s_list[right] = s_list[right], s_list[left]
            left += 1
            right -= 1
        return "".join(s_list)

solution = Solution()
input_k = input("请输入正整数k")
input_k = int(input_k)
input_s = input("请输入字符串s")
output_s = solution.right_xuanzhuan(input_k, input_s)
print("输出为：", output_s)
````

**算法思路**

解决字符串右旋转问题有多种方法，其中一种非常巧妙且高效的方法是使用 **三次反转**。这种方法可以在原地修改，不需要额外的存储空间。

三次反转的步骤如下：

1. **反转整个字符串的前面部分（除最后 k 个字符）：** 将字符串的前 `n - k` 个字符进行反转，其中 `n` 是字符串的长度。
2. **反转整个字符串的后面部分（最后 k 个字符）：** 将字符串的后 `k` 个字符进行反转。
3. **反转整个字符串：** 将整个字符串进行反转。

经过这三次反转后，原字符串的后 `k` 个字符就会被移动到前面，而原来的前 `n - k` 个字符会移动到后面，实现了右旋转的效果。

**为什么三次反转可以实现右旋转？**

我们可以用一个例子来理解：假设字符串为 "abcdefg"，k=2。

1. **反转前 n-k = 5 个字符 ("abcde")：** "edcba" + "fg" = "edcbagf"
   现在，需要旋转的 "fg" 在后面，前面的部分是反转过的。
2. **反转后 k = 2 个字符 ("fg")：** "edcba" + "gf" = "edcbagf"
   现在，需要旋转的 "fg" 已经反转好了，等待被移到前面。前面的部分仍然是反转过的。
3. **反转整个字符串 ("edcbagf")：** "fgabcde"
   整个字符串的反转，相当于把原来在后面的 "fg" 整体移到了前面，而原来在前面的 "abcde" 的反转 "edcba" 经过整体反转后又变回了 "abcde"，并且移到了 "fg" 的后面。

最终得到了右旋转后的结果。

**代码实现 (Python)**

```python
class Solution():
    def right_xuanzhuan(self, k: int, s: str) -> str:
        """
        使用三次反转的方法实现字符串右旋转
        Args:
            k: 右旋转的位数
            s: 需要旋转的字符串
        Returns:
            右旋转后的字符串
        """
        # 将字符串转换为列表，方便进行原地修改
        s_list = list(s)
        n = len(s_list) # 获取列表长度

        # 如果k大于字符串长度，取余数，因为旋转长度为n的效果和不旋转一样
        k = k % n

        # 如果k为0，直接返回原字符串
        if k == 0:
            return s

        # 第一次反转：反转前 n - k 个字符
        # 范围是 [0, n - k - 1]
        left = 0
        right = n - k - 1
        while left < right:
            # 交换两个元素
            s_list[left], s_list[right] = s_list[right], s_list[left]
            left += 1
            right -= 1

        # 第二次反转：反转后 k 个字符
        # 范围是 [n - k, n - 1]
        left = n - k
        right = n - 1
        while left < right:
            # 交换两个元素
            s_list[left], s_list[right] = s_list[right], s_list[left]
            left += 1
            right -= 1

        # 第三次反转：反转整个字符串
        # 范围是 [0, n - 1]
        left = 0
        right = n - 1
        while left < right:
            # 交换两个元素
            s_list[left], s_list[right] = s_list[right], s_list[left]
            left += 1
            right -= 1

        # 将修改后的列表转换回字符串并返回
        return "".join(s_list)

# 获取用户输入
input_k = input("请输入正整数k: ")
input_k = int(input_k) # 将输入的字符串转换为整数

input_s = input("请输入字符串s: ")

# 创建 Solution 类的实例
solution = Solution()

# 调用右旋转方法并获取结果
output_s = solution.right_xuanzhuan(input_k, input_s)

# 打印输出结果
print("输出为：", output_s)
```

**Python 语法解释**

* **`class Solution():`**: 定义了一个名为 `Solution` 的类。在 Python 中，类是一种蓝图，用于创建对象。
* **`def right_xuanzhuan(self, k: int, s: str) -> str:`**: 在 `Solution` 类中定义了一个方法（函数）名为 `right_xuanzhuan`。
    * `self`: 在类的方法定义中，`self` 参数代表类的实例本身。通过 `self`，我们可以访问和修改对象的属性和调用其他方法。
    * `k: int`: 参数 `k` 的类型提示为整数。
    * `s: str`: 参数 `s` 的类型提示为字符串。
    * `-> str`: 方法的返回值类型提示为字符串。
* **`s_list = list(s)`**: 将输入的字符串 `s` 转换为一个列表 `s_list`。字符串在 Python 中是不可变的，而列表是可变的，这使得我们可以方便地对列表中的元素进行修改（例如交换位置）。
* **`n = len(s_list)`**: 使用内置函数 `len()` 获取列表 `s_list` 的长度，并将其赋值给变量 `n`。
* **`k = k % n`**: 这是一个重要的优化。如果旋转的位数 `k` 大于字符串的长度 `n`，实际的旋转效果与旋转 `k % n` 位是相同的。例如，右旋 "abcdefg" 7 位（长度为 7）的结果仍然是 "abcdefg"，相当于右旋 0 位。右旋 9 位相当于右旋 2 位（9 % 7 = 2）。
* **`if k == 0:`**: 如果 `k` 经过取余后为 0，说明不需要旋转，直接返回原字符串即可。
* **`left = 0`, `right = n - k - 1` 等:** 初始化用于反转操作的左索引 `left` 和右索引 `right`。在 Python 中，列表的索引从 0 开始。
* **`while left < right:`**: 循环条件。当左索引小于右索引时，说明需要进行交换的元素还在左边。
* **`s_list[left], s_list[right] = s_list[right], s_list[left]`**: 这是 Python 中简洁地交换两个变量值的方式。它会先评估等号右边的表达式，然后同时将结果赋值给等号左边的变量。
* **`left += 1`, `right -= 1`**: 在每次循环中，将左索引向右移动一位，将右索引向左移动一位，逐步缩小反转的范围。
* **`return "".join(s_list)`**: 使用字符串的 `join()` 方法将列表 `s_list` 中的所有元素连接成一个字符串。`""` 是一个空字符串，表示在连接元素时不需要任何分隔符。
* **`input("...")`**: Python 内置函数，用于获取用户从控制台输入的字符串。
* **`int(input_k)`**: 将输入的字符串 `input_k` 转换为整数类型。
* **`solution = Solution()`**: 创建 `Solution` 类的一个实例。通过这个实例，我们可以调用类中定义的方法。
* **`output_s = solution.right_xuanzhuan(input_k, input_s)`**: 调用 `solution` 实例的 `right_xuanzhuan` 方法，并将用户输入的 `k` 和 `s` 作为参数传递。方法的返回值（即右旋转后的字符串）被赋给了 `output_s` 变量。
* **`print("输出为：", output_s)`**: 使用 Python 内置函数 `print()` 将结果输出到控制台。

**总结**

三次反转法是一种高效且优雅的字符串右旋转方法，因为它在原地进行操作，空间复杂度为 O(1)。通过将字符串转换为列表，我们可以利用列表的可变性方便地进行元素交换，然后将修改后的列表转换回字符串即可得到结果。在实现过程中，需要注意索引的计算和边界条件，以及处理 `k` 大于字符串长度的情况。
