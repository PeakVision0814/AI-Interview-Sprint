## [977. 有序数组的平方](https://leetcode.cn/problems/squares-of-a-sorted-array/)

简单 相关标签 ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业



给你一个按 **非递减顺序** 排序的整数数组 `nums`，返回 **每个数字的平方** 组成的新数组，要求也按 **非递减顺序** 排序。

 

**示例 1：**

```
输入：nums = [-4,-1,0,3,10]
输出：[0,1,9,16,100]
解释：平方后，数组变为 [16,1,0,9,100]
排序后，数组变为 [0,1,9,16,100]
```

**示例 2：**

```
输入：nums = [-7,-3,2,3,11]
输出：[4,9,9,49,121]
```

````python
class Solution:
    def sortedSquares(self, nums: List[int]) -> List[int]:
        n = len(nums)
        left = 0
        right = n - 1
        res = [0] * n
        i = 0
        while left <= right:
            if nums[left] ** 2 <= nums[right] ** 2:
                res[i] = nums[right] **2
                right -= 1
            else:
                res[i] = nums[left] **2
                left += 1
            i += 1
        return res[::-1]
````

这段代码的目的是将一个已经按非递减顺序排列的整数数组 `nums` 中的每个元素平方后，得到一个新的数组，这个新数组也要按非递减顺序排列。

**代码的总体思路（算法角度）：**

你使用了**双指针**的技巧来解决这个问题。双指针是一种很常用的算法思想，特别适合处理有序数组。想象一下，你有两个手指（就是这里的 `left` 和 `right` 指针），一个放在数组的最左边，一个放在数组的最右边。

为什么双指针在这里有效呢？因为原数组 `nums` 是已经排好序的。当我们把数组中的元素平方后，最大的平方数可能出现在原数组的最左边（因为负数的绝对值可能很大）或者最右边。例如：

*   数组 `[-4, -1, 0, 3, 10]`
*   平方后：`[16, 1, 0, 9, 100]`

你会发现，最大的平方数 `100` 来自原数组的最右边 `10`，次大的平方数 `16` 来自原数组的最左边 `-4`。

所以，我们可以比较最左边元素的平方和最右边元素的平方，哪个大，哪个就是我们结果数组中最大的那个数。然后把这个最大的数放到结果数组的**最后一个位置**。接着，我们将使用了较大平方数的那个指针向中间移动一位（如果用了右边的，就把 `right` 左移；如果用了左边的，就把 `left` 右移）。

我们重复这个过程：比较当前 `left` 指针和 `right` 指针指向元素的平方，将较大的那个平方数放到结果数组的倒数第二个位置，然后移动相应的指针。一直重复，直到 `left` 指针超过 `right` 指针。这样，我们就从结果数组的末尾开始，依次填充了最大的、次大的、... 最小的平方数。

**Python 语法解释（结合你的代码）：**

1.  `n = len(nums)`: 这一行是获取输入数组 `nums` 的长度，并把它存储在变量 `n` 里。`len()` 是 Python 内置的函数，用来计算序列（如列表、字符串等）的长度。

2.  `left = 0`: 初始化一个变量 `left`，作为左指针，指向数组的第一个元素（索引为 0）。

3.  `right = n - 1`: 初始化一个变量 `right`，作为右指针，指向数组的最后一个元素（索引为 `n-1`）。记住，Python 的列表索引是从 0 开始的。

4.  `res = [0] * n`: 初始化一个长度为 `n` 的列表 `res`，用来存储最终的结果。`[0] * n` 是一种 Python 的快速方式，创建一个包含 `n` 个 0 的列表。我们在这里用 0 作为占位符，后面会用计算出的平方数替换它们。

5.  `i = 0`: 初始化一个变量 `i`，用来作为**填充结果列表 `res` 的索引**。

6.  `while left <= right:`: 这是一个 `while` 循环，只要左指针 `left` 的位置小于等于右指针 `right` 的位置，循环就继续执行。当 `left` 大于 `right` 时，说明所有的元素都已经被处理过了，循环结束。

7.  `if nums[left] ** 2 <= nums[right] ** 2:`: 在循环内部，我们比较 `left` 指针指向元素的平方 (`nums[left] ** 2`) 和 `right` 指针指向元素的平方 (`nums[right] ** 2`)。
    *   `** 2` 是 Python 中求幂的运算符，表示平方。

8.  `res[i] = nums[right] ** 2`: 如果右边元素的平方大于等于左边元素的平方，那么将右边元素的平方值存储到结果列表 `res` 的第 `i` 个位置。

9.  `right -= 1`: 然后，因为右边元素的平方已经被处理了，我们将右指针 `right` 向左移动一位（减 1），指向下一个要比较的元素。

10. `else:`: 如果左边元素的平方大于右边元素的平方。

11. `res[i] = nums[left] ** 2`: 那么将左边元素的平方值存储到结果列表 `res` 的第 `i` 个位置。

12. `left += 1`: 然后，因为左边元素的平方已经被处理了，我们将左指针 `left` 向右移动一位（加 1），指向下一个要比较的元素。

13. `i += 1`: **注意！这里是你代码的一个巧妙之处。** 每次循环处理一个元素，无论是来自左边还是右边，我们都将结果存储到 `res` 列表的当前 `i` 位置，并将 `i` 增加 1。这意味着你正在从结果列表的**开头**开始填充。

14. `return res[::-1]`: 循环结束后，`res` 列表中存放的平方数是**倒序**的（从大到小），因为我们每次将较大的平方数放入了 `res` 的当前 `i` 位置，而 `i` 是从 0 递增的。为了得到非递减顺序的结果，你使用了列表切片 `[::-1]` 来将 `res` 列表进行**反转**。`[::-1]` 是一个非常简洁的 Python 技巧，用来创建一个反转后的列表副本。

**为什么你的代码能够成功？**

虽然双指针从后向前填充结果列表是更直接的实现方式（避免了最后的反转），但你的代码通过**从前向后填充结果列表，并在最后反转**也成功实现了同样的功能。

你的逻辑是：在每次比较左右指针指向的平方数时，总是将**较大的那个**放到 `res[i]` 的位置，然后 `i` 递增。这样，`res` 列表中存储的顺序会是：最大的、次大的、... 最小的。最后通过 `[::-1]` 一次性反转，得到了从小到大的排序结果。
