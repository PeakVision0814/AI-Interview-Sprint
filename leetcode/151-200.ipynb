## [151. 反转字符串中的单词](https://leetcode.cn/problems/reverse-words-in-a-string/)

中等   相关标签   ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业



给你一个字符串 `s` ，请你反转字符串中 **单词** 的顺序。

**单词** 是由非空格字符组成的字符串。`s` 中使用至少一个空格将字符串中的 **单词** 分隔开。

返回 **单词** 顺序颠倒且 **单词** 之间用单个空格连接的结果字符串。

**注意**：输入字符串 `s`中可能会存在前导空格、尾随空格或者单词间的多个空格。返回的结果字符串中，单词间应当仅用单个空格分隔，且不包含任何额外的空格。

 

**示例 1：**

```
输入：s = "the sky is blue"
输出："blue is sky the"
```

**示例 2：**

```
输入：s = "  hello world  "
输出："world hello"
解释：反转后的字符串中不能存在前导空格和尾随空格。
```

**示例 3：**

```
输入：s = "a good   example"
输出："example good a"
解释：如果两个单词间有多余的空格，反转后的字符串需要将单词间的空格减少到仅有一个。
```

````python
class Solution():
    def reverseWords(self, s:str) -> str:
        new_s = s.split()
        left = 0
        right = len(new_s) - 1
        while left < right:
            new_s[left], new_s[right] = new_s[right], new_s[left]
            left += 1
            right -= 1
        return " ".join(new_s)
    
solition = Solution()
s_input = input("请输入字符串s：")
reveresd_s = solition.reverseWords(s_input)
print(reveresd_s)
````

**思路分析：**

这道题的核心是将字符串中的单词按照相反的顺序排列。同时需要处理输入字符串中可能存在的冗余空格（前导、尾随和单词间的多个空格），确保输出结果中单词间只有一个空格。

我们可以将问题分解为以下几个步骤：

1. **识别和提取单词：** 从原始字符串中找出所有的单词，并忽略冗余的空格。
2. **反转单词的顺序：** 将提取出来的单词按照相反的顺序排列。
3. **构建最终结果字符串：** 将反转后的单词用单个空格连接起来形成最终的字符串。

**算法实现：**

你的代码采用了非常简洁且高效的算法实现：

1. **利用 Python 的 `split()` 方法：**
   这一步巧妙地利用了 Python 字符串内置的 `split()` 方法。当不给 `split()` 方法传递参数时，它会默认按照所有的空白字符（包括空格、制表符、换行符等）进行分割，并且会自动忽略连续的空白字符。这样，原始字符串中的前导、尾随和单词间的多个空格就被自动处理掉了，我们直接得到了一个由单词组成的列表。

````python
class ListNode:
    """
    定义链表节点类
    """
    def __init__(self, x):
        self.val = x # 节点的值
        self.next = None # 指向下一个节点的指针，默认为 None

class Solution:
    """
    解决链表交点问题的类
    """
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> ListNode:
        """
        寻找两个链表的交点（使用双指针法）

        Args:
            headA: 第一个链表的头节点
            headB: 第二个链表的头节点

        Returns:
            如果找到交点，返回交点节点；否则返回 None
        """
        # 初始化两个指针，分别指向两个链表的头节点
        pA = headA
        pB = headB

        # 循环直到两个指针相遇 (即 pA == pB)
        # 如果存在交点，它们会在交点处相遇
        # 如果不存在交点，它们最终会同时到达 None
        while pA != pB:
            # 如果 pA 到达链表 A 的末尾 (None)，则从链表 B 的头节点开始继续遍历
            # 否则，pA 向后移动一个节点
            if pA is None:
                pA = headB
            else:
                pA = pA.next

            # 如果 pB 到达链表 B 的末尾 (None)，则从链表 A 的头节点开始继续遍历
            # 否则，pB 向后移动一个节点
            if pB is None:
                pB = headA
            else:
                pB = pB.next

        # 当循环结束时，pA 和 pB 指向同一个节点
        # 如果存在交点，它们就是交点
        # 如果不存在交点，它们都是 None
        return pA

````

**解题思路：双指针法**

这道题可以使用一种非常巧妙的 **双指针法** 来解决，而无需提前计算链表的长度。其核心思想是让两个指针分别遍历两个链表，当一个指针到达链表末尾时，将其重定向到另一个链表的头节点。通过这种方式，如果两个链表存在交点，两个指针最终一定会相遇在交点处。如果不存在交点，两个指针会同时到达各自遍历路径的末尾，即都变为 `None`，并在 `None` 处相遇。

**算法步骤：**

1.  **初始化指针:** 创建两个指针 `pA` 和 `pB`，分别指向链表 A 的头节点 `headA` 和链表 B 的头节点 `headB`。
2.  **同步移动和重定向:** 进入一个循环，条件是 `pA != pB` (当两个指针不相等时继续循环)。在循环内部：
    *   每次循环，将 `pA` 向后移动一个节点 (`pA = pA.next`)。如果 `pA` 在移动后变为了 `None` (到达链表 A 的末尾)，则将其重定向到链表 B 的头节点 `headB`。
    *   同样地，将 `pB` 向后移动一个节点 (`pB = pB.next`)。如果 `pB` 在移动后变为了 `None` (到达链表 B 的末尾)，则将其重定向到链表 A 的头节点 `headA`。
3.  **相遇判断:** 当 `pA == pB` 时，循环结束。
    *   如果相遇的节点是 `None`，说明两个链表没有交点。
    *   如果相遇的节点不是 `None`，那么这个节点就是两个链表的交点。
4.  **返回结果:** 返回最终 `pA` (或者 `pB`，因为它们相等) 指向的节点。

**为什么这个方法可行？**

假设链表 A 在交点前的长度是 `a`，链表 B 在交点前的长度是 `b`，交点后的共同部分的长度是 `c`。

*   当 `pA` 遍历完链表 A (长度为 `a+c`) 后，它会被重定向到链表 B 的头节点。然后它会继续遍历链表 B 的 `b+c` 部分。`pA` 总共走过的长度是 `(a+c) + (b+c)`。
*   当 `pB` 遍历完链表 B (长度为 `b+c`) 后，它会被重定向到链表 A 的头节点。然后它会继续遍历链表 A 的 `a+c` 部分。`pB` 总共走过的长度是 `(b+c) + (a+c)`。

可以看到，无论链表 A 和链表 B 的长度是否相等，如果它们有交点，两个指针最终都会在交点处相遇，因为它们走过的总路径长度是相同的 (`a + c + b` 或 `b + c + a`)，并且它们以相同的“速度”（每次移动一个节点）前进。

如果两个链表没有交点，那么它们各自遍历完自己的链表后，会被重定向到另一个链表的头节点。它们会继续遍历另一个链表，最终两个指针都会同时到达 `None`，并在 `None` 处相遇。

class ListNode:

    定义链表节点类
    """
    def __init__(self, x):
        self.val = x # 节点的值
        self.next = None # 指向下一个节点的指针，默认为 None

    """
    解决链表交点问题的类
    """
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> ListNode:

        寻找两个链表的交点（使用双指针法）

        Args:
            headA: 第一个链表的头节点
            headB: 第二个链表的头节点

        Returns:

            # 否则，pA 继续向后移动 (pA.next)
            pA = pA.next if pA is not None else headB

            # pB 移动到下一个节点
            # 如果 pB 已经是 None (到达链表 B 末尾)，则将其重定向到链表 A 的头节点 headA
            # 否则，pB 继续向后移动 (pB.next)
            pB = pB.next if pB is not None else headA

        # 当循环结束时，pA == pB
        # 如果存在交点，pA 和 pB 会指向交点节点

给定一个二叉树的 **根节点** `root`，想象自己站在它的右侧，按照从顶部到底部的顺序，返回从右侧所能看到的节点值。

 

**示例 1：**

**输入：**root = [1,2,3,null,5,null,4]

**输出：**[1,3,4]

**示例 4：**

**输入：**root = []

**输出：**[]

````python
from typing import Optional, List
import collections
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    def rightSideView(self, root: Optional[TreeNode]) -> List[int]:
        if not root:
            return []
        
        queue = collections.deque([root])
        right_view = []

        while queue:
            level_size = len(queue)

            for i in range(level_size):
                node = queue.popleft()

                if i == level_size - 1:
                    right_view.append(node.val)

                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)

        return right_view
````

这串代码使用了**广度优先搜索 (BFS)** 的方法来解决这个问题，并且巧妙地利用了 BFS 按层遍历的特性。

我们来逐行解释：

```python
class Solution:
    def rightSideView(self, root: Optional[TreeNode]) -> List[int]:
        # rightSideView 是这个类的一个方法，用来计算二叉树的右视图
        # self 表示类的实例自身
        # root: Optional[TreeNode] 表示这个方法接受一个叫做 root 的参数，它的类型是 TreeNode 或者 None (Optional)
        # -> List[int] 表示这个方法将返回一个整数列表

        if not root:
            # 如果根节点不存在（树是空的），直接返回一个空列表
            return []

        # 初始化一个双端队列（deque），并把根节点放进去
        # 双端队列在这里用作 BFS 的队列
        queue = collections.deque([root])

        # 初始化一个列表，用来存储每一层的最右边节点的值
        right_view = []

        # 开始 BFS 循环，只要队列不为空，就继续遍历
        while queue:
            # 获取当前层的节点数量
            # 在进入内层循环前获取当前层的节点数量很重要，
            # 因为在内层循环中我们会不断向队列添加下一层的节点，
            # 如果不提前获取，level_size 会变动，就无法确定当前层的范围了。
            level_size = len(queue)

            # 遍历当前层的所有节点
            # range(level_size) 会生成一个从 0 到 level_size-1 的序列
            for i in range(level_size):
                # 从队列的左边（队首）取出一个节点进行处理
                node = queue.popleft()

                # 判断当前节点是不是当前层的最右边节点
                # i 是当前层节点的索引，从 0 开始
                # level_size - 1 就是当前层最后一个节点的索引
                if i == level_size - 1:
                    # 如果是当前层的最右边节点，将其值添加到 right_view 列表中
                    right_view.append(node.val)

                # 将当前节点的左子节点添加到队列尾部，以便在下一轮循环中处理
                # 需要检查左子节点是否存在（不为 None）
                if node.left:
                    queue.append(node.left)

                # 将当前节点的右子节点添加到队列尾部，以便在下一轮循环中处理
                # 需要检查右子节点是否存在（不为 None）
                if node.right:
                    queue.append(node.right)

        # BFS 遍历完成后，right_view 列表就包含了每一层的最右边节点的值
        # 返回这个列表作为结果
        return right_view
```

**核心思想：**

这个算法的核心思想是利用 BFS 的**按层遍历**特性。它逐层地遍历二叉树。在遍历每一层的时候，它会记住当前层有多少个节点 (`level_size`)。然后，它遍历完当前层的所有节点，并且特别关注最后一个被处理的节点（也就是 `i == level_size - 1` 的时候）。由于是从左到右依次将子节点入队，所以当前层的最后一个节点一定是该层从右边看过去能看到的那个节点。

**Python 语法解释：**

*   **`class Solution:`**: 定义一个名为 `Solution` 的类。在 LeetCode 等平台，通常会将解题代码放在一个类里面。
*   **`def rightSideView(self, root: Optional[TreeNode]) -> List[int]:`**: 定义了类 `Solution` 的一个方法 `rightSideView`。
    *   `self`: 表示这个方法是针对 `Solution` 类的实例调用的。
    *   `root: Optional[TreeNode]`: 参数 `root` 的类型注解，表示 `root` 可以是一个 `TreeNode` 对象，也可以是 `None`。这是 Python 3.5+ 引入的类型提示，可以帮助我们理解代码，但不是强制的。`TreeNode` 通常是一个表示二叉树节点的类，包含 `val` (节点值)、`left` (左子节点)、`right` (右子节点) 等属性。
    *   `-> List[int]`: 返回值类型注解，表示这个方法将返回一个由整数组成的列表。
*   **`if not root:`**: 判断 `root` 是否为 `None`。在 Python 中，`None`、空列表、空字符串、0 等在布尔上下文中都为 `False`。所以 `not root` 就是判断 `root` 是否是“假的”，在这里就是判断 `root` 是否是 `None`。
*   **`collections.deque([root])`**: 使用 `collections` 模块创建了一个双端队列（deque）。`[root]` 是一个包含根节点的列表，作为初始元素传入双端队列。双端队列比普通列表在两端添加和删除元素更高效，因此非常适合作为 BFS 的队列。
*   **`while queue:`**: 循环条件。只要 `queue` 这个双端队列不是空的，循环就继续。当队列为空时，说明所有节点都已经被处理过了。
*   **`len(queue)`**: 返回队列中元素的数量。
*   **`for i in range(level_size):`**: 这是一个标准的 for 循环，`range(level_size)` 生成一个从 0 到 `level_size - 1` 的整数序列。
*   **`queue.popleft()`**: 从双端队列的左边（队首）移除并返回一个元素。
*   **`right_view.append(node.val)`**: 将 `node` 对象的 `val` 属性（即节点的值）添加到 `right_view` 列表的末尾。
*   **`if node.left:`**: 判断 `node` 的左子节点是否存在。
*   **`queue.append(node.left)`**: 将 `node` 的左子节点添加到双端队列的右边（队尾）。这是 BFS 中将下一层节点入队的操作。
*   **`if node.right:`**: 判断 `node` 的右子节点是否存在。
*   **`queue.append(node.right)`**: 将 `node` 的右子节点添加到双端队列的右边（队尾）。

总的来说，这段代码是一个非常经典的 BFS 遍历二叉树的模板，通过在每一层只记录最后一个访问到的节点，巧妙地解决了二叉树右视图的问题。
