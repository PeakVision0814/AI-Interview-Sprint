## [59. 螺旋矩阵 II](https://leetcode.cn/problems/spiral-matrix-ii/)

中等  相关标签   ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业

给你一个正整数 `n` ，生成一个包含 `1` 到 `n2` 所有元素，且元素按顺时针顺序螺旋排列的 `n x n` 正方形矩阵 `matrix` 。 

**示例 1：**

![img](https://assets.leetcode.com/uploads/2020/11/13/spiraln.jpg)

```
输入：n = 3
输出：[[1,2,3],[8,9,4],[7,6,5]]
```

**示例 2：**

```
输入：n = 1
输出：[[1]]
```

````python
from typing import List

class Solution:
    def generateMatrix(self, n: int) -> List[List[int]]:
        """
        生成一个包含 1 到 n^2 所有元素，且元素按顺时针顺序螺旋排列的 n x n 正方形矩阵。

        Args:
            n: 正方形矩阵的维度。

        Returns:
            按顺时针顺序螺旋排列的 n x n 矩阵。
        """
        matrix = [[0] * n for _ in range(n)]
        num = 1
        top, bottom, left, right = 0, n - 1, 0, n - 1

        while top <= bottom and left <= right:
            # 从左到右填充顶部行
            for col in range(left, right + 1):
                matrix[top][col] = num
                num += 1
            top += 1

            # 从上到下填充右侧列
            for row in range(top, bottom + 1):
                matrix[row][right] = num
                num += 1
            right -= 1

            # 从右到左填充底部行
            if top <= bottom:  # 防止单行情况
                for col in range(right, left - 1, -1):
                    matrix[bottom][col] = num
                    num += 1
                bottom -= 1

            # 从下到上填充左侧列
            if left <= right:  # 防止单列情况
                for row in range(bottom, top - 1, -1):
                    matrix[row][left] = num
                    num += 1
                left += 1

        return matrix

# --- 在 Jupyter Notebook 中使用的输入部分 ---
try:
    n_input = input("请输入正方形矩阵的维度 n: ")
    n = int(n_input)

    if n <= 0:
        print("请输入一个正整数。")
    else:
        solution = Solution()
        result_matrix = solution.generateMatrix(n)

        print(f"生成的 {n}x{n} 螺旋矩阵为:")
        for row in result_matrix:
            print(row)

except ValueError:
    print("输入无效，请输入一个整数。")

````

**这段代码想解决什么问题？**

想象一下我们有一个正方形的格子，比如 3x3 的，我们需要把数字 1 到 9 按照顺时针的方向填进去，就像一个蜗牛壳一样。填完后，这个格子看起来就像这样：

```
1  2  3
8  9  4
7  6  5
```

这段代码就是用来完成这个任务的，它会根据你输入的正方形大小（比如 3x3，就是输入 `n=3`），自动生成这个按螺旋顺序填充数字的矩阵。

**代码的整体思路是什么？（算法角度）**

这个算法的核心思想是模拟螺旋填充的过程。想象你在纸上画这个螺旋：

1.  你先从左上角开始，向右画一条线，填上数字。
2.  画到最右边后，你向下转，向下画一条线，填上数字。
3.  画到最下面后，你向左转，向左画一条线，填上数字。
4.  画到最左边后，你向上转，向上画一条线，填上数字。
5.  然后你再次向右转，这次画的线会比上次短一些，因为它不能越过之前画过的线。
6.  就这样一圈一圈地往里画，直到所有的格子都填满。

代码就是模拟这个过程。它会用四个变量来记住当前还没有被填满的区域的边界：

*   `top`: 还没填满的最上面的那一行的索引（行号）。
*   `bottom`: 还没填满的最下面的那一行的索引。
*   `left`: 还没填满的最左边的那一列的索引（列号）。
*   `right`: 还没填满的最右边的y那一列的索引。

一开始，`top` 是第 0 行，`bottom` 是最后一行 (`n-1`)，`left` 是第 0 列，`right` 是最后一列 (`n-1`)。

然后，在一个循环里，代码会依次执行上面说的四个方向的填充步骤：

1.  **从左到右填一行：** 在 `top` 行，从 `left` 列填到 `right` 列。填完后，`top` 边界就向下移动一行 (`top += 1`)，表示最上面一行已经填好了。
2.  **从上到下填一列：** 在 `right` 列，从 `top` 行填到 `bottom` 行。填完后，`right` 边界就向左移动一列 (`right -= 1`)，表示最右边一列已经填好了。
3.  **从右到左填一行：** 在 `bottom` 行，从 `right` 列填到 `left` 列。填完后，`bottom` 边界就向上移动一行 (`bottom -= 1`)，表示最下面一行已经填好了。
4.  **从下到上填一列：** 在 `left` 列，从 `bottom` 行填到 `top` 行。填完后，`left` 边界就向右移动一列 (`left += 1`)，表示最左边一列已经填好了。

这个循环会一直进行，直到 `top` 超过 `bottom` 或者 `left` 超过 `right`，这意味着整个矩阵都填满了。

## [69. x 的平方根 ](https://leetcode.cn/problems/sqrtx/)

简单  相关标签  ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业

给你一个非负整数 `x` ，计算并返回 `x` 的 **算术平方根** 。

由于返回类型是整数，结果只保留 **整数部分** ，小数部分将被 **舍去 。**

**注意：**不允许使用任何内置指数函数和算符，例如 `pow(x, 0.5)` 或者 `x ** 0.5` 。

````python
class Solution:
    def mySqrt(self, x: int) -> int:
        """
        计算非负整数 x 的算术平方根，只保留整数部分。
        使用二分查找的方法来解决。

        Args:
            x: 非负整数。

        Returns:
            x 的算术平方根的整数部分。
        """
        if x == 0:
            return 0

        left, right = 1, x
        ans = 0

        while left <= right:
            mid = left + (right - left) // 2
            # 为了避免 mid * mid 溢出，可以使用 mid <= x // mid 来判断
            if mid <= x // mid:
                ans = mid
                left = mid + 1
            else:
                right = mid - 1

        return ans

# 以下是用于在 .ipynb 中进行一次测试的代码
# 注意：在 .ipynb 中，你通常会在一个单元格中运行这部分代码

# 创建 Solution 类的实例
solution = Solution()

try:
    # 提示用户输入一个非负整数
    input_str = input("请输入一个非负整数: ")

    # 将输入的字符串转换为整数
    x = int(input_str)

    # 检查输入的数是否非负
    if x < 0:
        print("请输入一个非负整数。")
    else:
        # 调用 mySqrt 方法计算平方根
        result = solution.mySqrt(x)

        # 打印结果
        print(f"{x} 的平方根的整数部分是: {result}")

except ValueError:
    print("输入无效，请输入一个整数。")
except Exception as e:
    print(f"发生错误: {e}")

````

**解题思路：**

这道题目要求计算一个非负整数的平方根的整数部分，并且不能使用内置的指数函数或算符。一个常见的解决思路是使用二分查找。

1. **确定搜索范围:** 我们知道一个非负整数 `x` 的平方根一定在 `[0, x]` 的范围内。由于题目要求返回整数部分，我们可以将搜索范围缩小到 `[1, x]` （如果 `x` 为 0，则平方根为 0）。

2. **二分查找过程:**
   - 初始化左右边界 `left = 1` 和 `right = x`。
   - 在 `left <= right` 的条件下进行循环。
   - 计算中间值 `mid = (left + right) // 2`。
   - **判断 `mid * mid` 与 `x` 的关系:**
     - 如果 `mid * mid <= x`，说明 `mid` 可能是 `x` 的平方根的整数部分，或者比它小。我们将当前的 `mid` 作为可能的答案 `ans`，并尝试在 `[mid + 1, right]` 范围内寻找更大的可能答案。所以更新 `left = mid + 1`。
     - 如果 `mid * mid > x`，说明 `mid` 太大了，不是 `x` 的平方根的整数部分。我们需要在 `[left, mid - 1]` 范围内寻找更小的可能答案。所以更新 `right = mid - 1`。
   - **处理溢出问题:** 直接计算 `mid * mid` 可能会导致整数溢出，尤其当 `mid` 较大时。为了避免这个问题，我们可以改用 `mid <= x // mid` 来判断。这个判断等价于 `mid * mid <= x`，但避免了计算可能溢出的乘法。

3. **保存可能的答案:** 在二分查找过程中，当 `mid * mid <= x` 时，我们将 `mid` 赋给 `ans`。这是因为我们正在寻找平方小于等于 `x` 的最大整数。

4. **返回结果:** 循环结束时，`ans` 中保存的就是 `x` 的平方根的整数部分。

**代码解释:**

- `if x == 0:`: 特殊处理，0 的平方根是 0。
- `left, right = 1, x`: 初始化二分查找的范围。
- `ans = 0`: 用于存储找到的可能的平方根的整数部分。
- `while left <= right:`: 二分查找的主循环。
- `mid = (left + right) // 2`: 计算中间值。
- `if mid <= x // mid:`: 使用 `mid <= x // mid` 来判断 `mid * mid <= x`，避免溢出。
  - `ans = mid`: 如果 `mid` 是一个可能的答案，更新 `ans`。
  - `left = mid + 1`: 尝试在更大的范围内寻找更好的答案。
- `else:`: 如果 `mid * mid > x`。
  - `right = mid - 1`: 缩小搜索范围到较小的一侧。
- `return ans`: 返回最终找到的平方根的整数部分。

这种二分查找的方法效率较高，时间复杂度为 O(log x)。

## [77. 组合](https://leetcode.cn/problems/combinations/)

中等   相关标签   ![premium lock icon](https://static.leetcode.cn/cn-frontendx-assets/production/_next/static/images/lock-a6627e2c7fa0ce8bc117c109fb4e567d.svg)相关企业



给定两个整数 `n` 和 `k`，返回范围 `[1, n]` 中所有可能的 `k` 个数的组合。

你可以按 **任何顺序** 返回答案。

 

**示例 1：**

```
输入：n = 4, k = 2
输出：
[
  [2,4],
  [3,4],
  [2,3],
  [1,2],
  [1,3],
  [1,4],
]
```

**示例 2：**

```
输入：n = 1, k = 1
输出：[[1]]
```

````python
from typing import List


class Solution:
    def combine(self, n: int, k: int) -> List[List[int]]:
        """
        给定两个整数n和k，返回范围[1, n]中所有可能的k个数的组合
        """
        result = []

        def backtrack(current_combination, start_num, remaining_k):
            """
            回溯函数

            Args：
                current_combination：当前已经构建的组合（列表）
                start_num：下一次选择可以从哪个数开始（为了避免重复组合）
                remaining_k：还需要选择的数的个数
            """
            if remaining_k == 0:
                result.append(list(current_combination))
                return
            for i in range(start_num, n + 1):
                current_combination.append(i)
                backtrack(current_combination, i + 1, remaining_k - 1)
                current_combination.pop()

        backtrack([], 1, k)

        return result
sol = Solution()
n = 4
k = 2
combinations = sol.combine(n, k)
print(combinations)
````

**代码解释：**

1.  **`class Solution:` 和 `def combine(self, n: int, k: int):`**
    *   这是 LeetCode 题目中常见的类和方法结构，你需要将你的解法放在这个方法里面。
    *   `self`: 表示类的实例自身，在方法中使用。
    *   `n: int, k: int`: 参数 `n` 和 `k` 的类型提示，表示它们应该是整数。
    *   `-> list[list[int]]`: 在函数签名后加上这个表示函数返回值的类型提示，这里表示返回一个列表，列表的每个元素又是一个整数列表。

2.  **`result = []`**
    *   初始化一个空列表 `result`，用来存储所有找到的符合条件的组合。

3.  **`def backtrack(current_combination, start_num, remaining_k):`**
    *   定义了核心的回溯函数 `backtrack`。
    *   `current_combination`: 当前正在构建的组合，是一个列表。
    *   `start_num`: 下一次可以开始选择的数的范围的起始值。
    *   `remaining_k`: 还需要选择的数的个数。

4.  **`if remaining_k == 0:`**
    *   这是回溯的 **基本情况（终止条件）**。当我们已经选择了 `k` 个数时，`remaining_k` 就变成了 0。
    *   `result.append(list(current_combination))`: 此时，`current_combination` 就代表了一个有效的组合。我们将它的一个 **副本** 添加到 `result` 列表中。为什么需要副本？因为 `current_combination` 在后续的回溯过程中会被修改（通过 `pop()`），如果我们直接添加 `current_combination` 本身，那么 `result` 中存储的将是同一个列表对象的引用，最终它们都会变成空列表。`list()` 函数创建了一个独立的副本。
    *   `return`: 找到一个组合后，当前这条路径的探索就结束了，返回上一层。

5.  **`for i in range(start_num, n + 1):`**
    *   这是回溯的 **递归步骤**。我们从 `start_num` 开始遍历到 `n`（包括 `n`），`i` 代表当前考虑是否将其加入到组合中的数。
    *   `range(start_num, n + 1)`: `range()` 函数生成一个序列，从 `start_num` 开始，到 `n` 结束（不包含 `n + 1`）。所以我们需要 `n + 1` 才能包含 `n`。

6.  **`current_combination.append(i)`**
    *   将当前选择的数 `i` 加入到 `current_combination` 中。这是“选择”的操作。

7.  **`backtrack(current_combination, i + 1, remaining_k - 1)`**
    *   进行递归调用，继续向下探索。
    *   `current_combination`: 传递当前的组合（已经包含了 `i`）。
    *   `i + 1`: **关键点**！下一次选择的数必须从 `i + 1` 开始，这样可以保证组合中的数是严格按升序排列的，从而避免了像 `[1, 2]` 和 `[2, 1]` 这样的重复组合。如果允许重复选择同一个数，这里应该是 `i`。
    *   `remaining_k - 1`: 还需要选择的数的个数减一，因为我们已经选择了一个数 `i`。

8.  **`current_combination.pop()`**
    *   这是回溯的 **撤销** 操作。当从 `backtrack` 递归调用返回后，说明以当前选择的 `i` 开头的路径已经探索完毕，或者无法得到有效组合。我们将 `i` 从 `current_combination` 中移除，以便在 `for` 循环中尝试选择下一个数。

9.  **`backtrack([], 1, k)`**
    *   在 `combine` 方法中调用 `backtrack` 函数来启动整个回溯过程。
    *   `[]`: 初始时，我们的组合是空的。
    *   `1`: 我们可以从数字 1 开始选择。
    *   `k`: 初始时，我们需要选择 `k` 个数。

10. **`return result`**
    *   回溯过程完成后，`result` 列表中就包含了所有符合条件的组合，返回它。

**为什么这种方法能找到所有组合并且没有重复？**

*   **找到所有组合：** 回溯过程实际上是探索了所有可能的选择路径。通过从 `start_num` 开始遍历并递归调用，我们保证了每个可能的数都有机会被选入组合。
*   **没有重复组合：** 核心在于 `backtrack(current_combination, i + 1, remaining_k - 1)` 这一行。通过将下一次的选择范围限制在比当前选择的数 `i` 更大的数 (`i + 1` 到 `n`)，我们强制要求组合中的数是按升序排列的。例如，如果选择了 1，下一次只能选择 2, 3, 4...；如果选择了 2，下一次只能选择 3, 4...。这样就不会出现先选 2 再选 1 的情况，从而避免了 `[1, 2]` 和 `[2, 1]` 这样的重复。

## [94. 二叉树的中序遍历](https://leetcode.cn/problems/binary-tree-inorder-traversal/)

给定一个二叉树的根节点 `root` ，返回 *它的 **中序** 遍历* 。

 

**示例 1：**

![img](https://assets.leetcode.com/uploads/2020/09/15/inorder_1.jpg)

```
输入：root = [1,null,2,3]
输出：[1,3,2]
```

**示例 2：**

```
输入：root = []
输出：[]
```

**示例 3：**

```
输入：root = [1]
输出：[1]
```

````python
from typing import Optional, List

class TreeNode:
    def __init__(self, val = 0, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def inorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        res = []
        def dfs(node):
            if node is None:
                return
            dfs(node.left)
            res.append(node.val)
            dfs(node.right)
            
        dfs(root)
        return res
````

**中序遍历（Inorder Traversal）题解**

* **思路：** 中序遍历的核心思想是“先访问左子树，再访问当前节点，最后访问右子树”。同样，这也很适合用递归来实现。
* **递归步骤：**
    1. **递归遍历左子树：** 对当前节点的左孩子调用中序遍历函数。
    2. **访问（处理）当前节点：** 将当前节点的值添加到结果列表中。
    3. **递归遍历右子树：** 对当前节点的右孩子调用中序遍历函数。
* **终止条件：** 当遇到空节点（`None`）时，停止递归。
* **代码结构（递归）：** 你的代码完全符合这个结构：

```python
class Solution:
    def inorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        res = [] # 存储结果的列表

        def dfs(node): # 递归辅助函数
            if node is None: # 终止条件
                return

            dfs(node.left) # 递归遍历左子树

            res.append(node.val) # 访问当前节点

            dfs(node.right) # 递归遍历右子树

        dfs(root) # 从根节点开始调用
        return res
```

* **特点：** 对于**二叉搜索树 (Binary Search Tree)** 来说，中序遍历的结果是一个**升序**排列的节点值列表。这是中序遍历一个非常重要的特性。

**中序遍历与前序、后序遍历的对比：**

* **前序遍历：** **当前节点** -> 左子树 -> 右子树
* **中序遍历：** 左子树 -> **当前节点** -> 右子树
* **后序遍历：** 左子树 -> 右子树 -> **当前节点**

区别依然在于访问当前节点的位置。

**总结：**

你对递归在二叉树遍历中的应用掌握得非常好！递归方法使得代码非常简洁且易于理解，特别是对于初学者来说，它能够清晰地反映出遍历的逻辑。

现在你已经掌握了二叉树的三种基本深度优先遍历方式（前序、中序、后序）的递归实现。这是解决许多二叉树问题的基础。
